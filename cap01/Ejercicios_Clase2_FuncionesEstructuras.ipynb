```python
def validar_parentesis(s: str) -> bool:
    pares = {')': '(', ']': '[', '}': '{'}
    pila = []
    for char in s:
        if char in '([{':
            pila.append(char)

            yield elem

def top_k_frecuentes(texto: str, k: int) -> list:
    palabras = re.findall(r'\b\w+\b', texto.lower())
    counter = collections.Counter(palabras)
    ordenado = sorted(counter.items(), key=lambda x: (-x[1], x[0]))
    return ordenado[:k]


# ─── Desafío: MiniStack ──────────────────────────────────────────────────────
class PilaVacia(Exception): pass
class CapacidadExcedida(Exception): pass

class MiniStack:
    def __init__(self, maxsize=None):
        self._data = []
        self._maxsize = maxsize

    def push(self, valor):
        if self._maxsize is not None and len(self._data) >= self._maxsize:
            raise CapacidadExcedida(f"Capacidad máxima {self._maxsize} alcanzada")
        self._data.append(valor)

    def pop(self):
        if not self._data:
            raise PilaVacia("La pila está vacía")
        return self._data.pop()

    def peek(self):
        if not self._data:
            raise PilaVacia("La pila está vacía")
        return self._data[-1]

    def __len__(self):
        return len(self._data)

    def __bool__(self):
        return len(self._data) > 0

    def __iter__(self):
        return reversed(self._data)

    def __repr__(self):
        return f"MiniStack({list(reversed(self._data))})"
```

````python
# --- Visualización: MiniStack — comportamiento LIFO ---
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def visualizar_pila(operaciones):
    """Muestra el estado de la pila tras cada operación."""
    pila_estado = []
    estados = []
    etiquetas = []
    for op, val in operaciones:
        if op == 'push':
            pila_estado.append(val)
            etiquetas.append(f"push({val})")
        elif op == 'pop' and pila_estado:
            pila_estado.pop()
            etiquetas.append(f"pop()")
        estados.append(list(pila_estado))

    n = len(estados)
    fig, axes = plt.subplots(1, n, figsize=(n * 1.6, 5))
    if n == 1: axes = [axes]

    max_h = max((len(s) for s in estados), default=1)
    for ax, estado, etiq in zip(axes, estados, etiquetas):
        ax.set_xlim(0, 1); ax.set_ylim(0, max_h + 1)
        ax.axis('off'); ax.set_title(etiq, fontsize=8, pad=3)
        for i, v in enumerate(estado):
            color = 'steelblue' if i < len(estado) - 1 else 'coral'
            rect = mpatches.FancyBboxPatch((0.1, i + 0.05), 0.8, 0.85,
                                           boxstyle="round,pad=0.05",
                                           fc=color, ec='white', lw=1.5)
            ax.add_patch(rect)
            ax.text(0.5, i + 0.5, str(v), ha='center', va='center',
                    fontsize=11, color='white', fontweight='bold')
        if estado:
            ax.text(0.5, len(estado) + 0.3, '← tope', ha='center',
                    fontsize=7, color='gray')

    plt.suptitle('Evolución de MiniStack (rojo = tope actual)', fontsize=10, y=1.02)
    plt.tight_layout(); plt.show()

ops = [('push', 10), ('push', 20), ('push', 30), ('pop', None), ('push', 40)]
visualizar_pila(ops)
````

## 📊 Análisis empírico de complejidad

Verificamos experimentalmente que `top_k_frecuentes` tiene complejidad O(n log n).

````python
import time
import random
import string

def texto_aleatorio(n_palabras, vocab_size=50):
    """Genera texto con `n_palabras` palabras de vocabulario fijo."""
    vocab = [''.join(random.choices(string.ascii_lowercase, k=5))
             for _ in range(vocab_size)]
    return ' '.join(random.choices(vocab, k=n_palabras))

tamaños = [1_000, 5_000, 10_000, 50_000, 100_000, 500_000]
tiempos = []

for n in tamaños:
    texto = texto_aleatorio(n)
    t0 = time.perf_counter()
    for _ in range(3):              # 3 repeticiones para estabilidad
        top_k_frecuentes(texto, 10)
    t1 = time.perf_counter()
    tiempos.append((t1 - t0) / 3 * 1000)   # ms

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

# Escala lineal
axes[0].plot(tamaños, tiempos, 'o-', color='steelblue', linewidth=2)
axes[0].set_xlabel("N palabras"); axes[0].set_ylabel("Tiempo (ms)")
axes[0].set_title("top_k_frecuentes — escala lineal")
axes[0].grid(alpha=0.3)

# Escala log-log
axes[1].loglog(tamaños, tiempos, 's-', color='coral', linewidth=2)
axes[1].set_xlabel("N palabras (log)"); axes[1].set_ylabel("Tiempo ms (log)")
axes[1].set_title("Escala log-log → pendiente ≈ 1 implica O(n)")
axes[1].grid(alpha=0.3, which='both')

plt.tight_layout(); plt.show()
print("Nota: Counter + sorted ~ O(n log k) donde k = palabras únicas")
````
