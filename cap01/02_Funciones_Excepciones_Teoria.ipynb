# 🔷 Sección 1: Funciones en Python

## 🎯 Introducción

> *"Functions are first-class objects in Python, meaning they can be assigned to variables, passed as parameters, and returned from other functions."*  
> — Goodrich, Tamassia & Goldwasser, §1.5

**Analogía:** Una función es como una **receta de cocina**:
- Tiene un nombre claro (`def nombre`)
- Acepta ingredientes con cantidades (parámetros con defaults)
- Puede recibir ingredientes a gusto (`*args`, `**kwargs`)
- Devuelve el plato terminado (`return`)
- Viene con instrucciones escritas (docstring)

## 📊 Anatomía de una función

```python
def nombre(param1, param2=default, *args, **kwargs):
    """Docstring: qué hace, args, returns, raises."""
    # cuerpo
    return resultado
```

| Componente | Descripción |
|-----------|-------------|
| `param1` | Parámetro posicional obligatorio |
| `param2=default` | Parámetro con valor por defecto (opcional) |
| `*args` | Tupla de argumentos posicionales extra |
| `**kwargs` | Dict de argumentos keyword extra |
| docstring | Documentación inmediata; accesible con `help()` |

````python
# ── Parámetros y defaults ─────────────────────────────────────────────────────
def potencia(base, exponente=2):
    """
    Calcula base^exponente usando multiplicaciones.
    
    Args:
        base: número base
        exponente: entero no negativo (default 2)
    
    Returns:
        base ** exponente (int o float)
    
    Raises:
        ValueError: si exponente < 0
    """
    if exponente < 0:
        raise ValueError("exponente debe ser >= 0")
    resultado = 1
    for _ in range(exponente):
        resultado *= base
    return resultado

print("potencia(5):       ", potencia(5))       # usa default exp=2
print("potencia(2, 10):   ", potencia(2, 10))   # exp explícito
print("potencia(3, 0):    ", potencia(3, 0))    # caso base

# ── *args y **kwargs ──────────────────────────────────────────────────────────
def suma(*numeros):
    """Suma un número arbitrario de argumentos. §1.5"""
    total = 0
    for n in numeros:
        total += n
    return total

print("\nsuma(1,2,3):          ", suma(1, 2, 3))
print("suma(1,2,3,4,5):      ", suma(1, 2, 3, 4, 5))
print("suma():               ", suma())           # retorna 0

def describir(**atributos):
    """Muestra atributos key=value arbitrarios."""
    return ", ".join(f"{k}={v}" for k, v in atributos.items())

print("\ndescribir(nombre='Pila', capacidad=100, llena=False):")
print(" ", describir(nombre="Pila", capacidad=100, llena=False))

# ── Funciones como objetos (first-class) ──────────────────────────────────────
print("\n--- Funciones como objetos ---")
def aplicar(func, valores):
    """Aplica func a cada elemento de valores."""
    return [func(v) for v in valores]

cuadrado = lambda x: x * x     # función anónima
print("cuadrados:", aplicar(cuadrado, [1, 2, 3, 4, 5]))
print("abs       :", aplicar(abs, [-3, 5, -1, 2]))
````

### 🔍 Scope: regla LEGB

Python resuelve nombres siguiendo la regla **LEGB**:

| Nivel | Descripción | Ejemplo |
|-------|-------------|---------|
| **L**ocal | Variables dentro de la función actual | `x = 1` dentro de `def f():` |
| **E**nclosing | Funciones anidadas (closure) | la `x` de la función exterior |
| **G**lobal | Nivel de módulo | variable declarada fuera de todas las funciones |
| **B**uilt-in | Nombres de Python (`len`, `print`, `True`…) | siempre disponibles |

> ⚠️ **Gotcha:** si asignás una variable dentro de una función, Python la trata como **local** aunque exista una global con el mismo nombre. Usar `global x` o `nonlocal x` para modificar la del scope exterior (raro en código limpio; preferir pasar como argumento).

# 🔷 Sección 2: Entrada y Salida

## 🎯 Introducción

Python ofrece `print()` altamente configurable e `input()` para leer texto del usuario.  
En Jupyter Notebooks usamos principalmente `print` para mostrar resultados de algoritmos.

## 📊 Opciones de `print()`

| Argumento | Default | Descripción |
|-----------|---------|-------------|
| `sep` | `' '` | Separador entre argumentos |
| `end` | `'\n'` | Carácter al final |
| `file` | `sys.stdout` | Stream de salida |
| `flush` | `False` | Forzar vaciado del buffer |

````python
# ── print() avanzado ─────────────────────────────────────────────────────────
print("a", "b", "c")                      # sep por defecto: espacio
print("a", "b", "c", sep="|")             # separador personalizado
print("Sin salto", end=" → ")
print("misma línea")

# ── f-strings (Python 3.6+) ───────────────────────────────────────────────────
print("\n--- f-strings ---")
nombre = "Python"
version = 3.12
pi = 3.14159265358979

print(f"Lenguaje: {nombre}, versión: {version}")
print(f"pi con 4 decimales:  {pi:.4f}")
print(f"pi en notación científica: {pi:.3e}")
print(f"entero con ceros: {42:05d}")          # 00042
print(f"expresión: {2 ** 10 = }")             # muestra la expresión y su valor

# ── Formato tabular ────────────────────────────────────────────────────────────
print("\n--- Tabla de cuadrados y cubos ---")
print(f"{'n':>4} {'n²':>6} {'n³':>8}")
print("-" * 20)
for n in range(1, 8):
    print(f"{n:>4} {n**2:>6} {n**3:>8}")

# ── Lectura segura de entero ──────────────────────────────────────────────────
def leer_entero(prompt, minimo=None, maximo=None):
    """
    Lee un entero desde stdin con validación.
    (Simulado aquí con un valor fijo para el notebook)
    """
    # En un script real: valor = input(prompt)
    valor_simulado = "7"
    try:
        n = int(valor_simulado)
    except ValueError:
        raise ValueError(f"'{valor_simulado}' no es un entero válido")
    if minimo is not None and n < minimo:
        raise ValueError(f"{n} es menor que el mínimo {minimo}")
    if maximo is not None and n > maximo:
        raise ValueError(f"{n} es mayor que el máximo {maximo}")
    return n

valor = leer_entero("Ingresá un número del 1 al 10: ", 1, 10)
print(f"\nValor leído: {valor}")
````

# 🔷 Sección 3: Manejo de Excepciones

## 🎯 Introducción

> *"Python uses exceptions to signal that an error or unexpected condition has occurred."*  
> — Goodrich, Tamassia & Goldwasser, §1.7

En estructuras de datos es fundamental distinguir entre:
- **Precondición violada** (culpa del llamador) → `raise ValueError`, `TypeError`
- **Estado inválido** (culpa de la estructura) → excepción propia, p.ej. `PilaVaciaError`

## 📊 Jerarquía de excepciones relevantes

```
BaseException
└── Exception
    ├── ArithmeticError
    │   └── ZeroDivisionError
    ├── LookupError
    │   ├── IndexError
    │   └── KeyError
    ├── TypeError
    ├── ValueError
    ├── AttributeError
    ├── RuntimeError
    │   └── RecursionError
    └── StopIteration   ← importante para iteradores
```

## 📋 Sintaxis completa `try/except/else/finally`

| Bloque | Cuándo se ejecuta |
|--------|-------------------|
| `try` | Siempre — código que puede fallar |
| `except E` | Si ocurrió excepción de tipo `E` |
| `else` | Solo si NO ocurrió ninguna excepción |
| `finally` | **Siempre**, haya o no excepción (liberación de recursos) |

````python
# ── try / except / else / finally básico ─────────────────────────────────────
print("=== try/except/else/finally ===\n")

def dividir(a, b):
    """División segura con manejo de excepciones."""
    try:
        resultado = a / b
    except ZeroDivisionError as e:
        print(f"  [except] No se puede dividir por cero: {e}")
        return None
    except TypeError as e:
        print(f"  [except] Tipos inválidos: {e}")
        return None
    else:
        print(f"  [else]   {a} / {b} = {resultado:.4f}")
        return resultado
    finally:
        print(f"  [finally] llamada dividir({a}, {b}) completada")

dividir(10, 3)
print()
dividir(5, 0)
print()
dividir("5", 2)

# ── raise ──────────────────────────────────────────────────────────────────────
print("\n=== raise con mensaje descriptivo ===")
def busqueda_binaria(data, objetivo):
    """Búsqueda binaria. Requiere data ordenada."""
    if not isinstance(data, list):
        raise TypeError(f"Se esperaba list, se recibió {type(data).__name__}")
    if data != sorted(data):
        raise ValueError("La lista debe estar ordenada para búsqueda binaria")
    
    lo, hi = 0, len(data) - 1
    while lo <= hi:
        mid = (lo + hi) // 2
        if data[mid] == objetivo:
            return mid
        elif data[mid] < objetivo:
            lo = mid + 1
        else:
            hi = mid - 1
    return -1

data_ord = [1, 3, 5, 7, 9, 11, 13]
print(f"buscar 7 en {data_ord}: índice {busqueda_binaria(data_ord, 7)}")
print(f"buscar 6 en {data_ord}: índice {busqueda_binaria(data_ord, 6)}")

try:
    busqueda_binaria([3, 1, 2], 2)  # desordenada → ValueError
except ValueError as e:
    print(f"ValueError capturado: {e}")
````

````python
# ── Excepciones propias para estructuras de datos ─────────────────────────────
print("=== Excepciones propias ===\n")

class EstructuraVaciaError(Exception):
    """Se lanza cuando se intenta operar sobre una estructura de datos vacía."""
    pass

class CapacidadExcedidaError(Exception):
    """Se lanza cuando la estructura alcanzó su capacidad máxima."""
    def __init__(self, capacidad):
        super().__init__(f"Capacidad máxima alcanzada: {capacidad} elementos")
        self.capacidad = capacidad

# Pila simple que usa excepciones propias
class PilaSimple:
    """Pila (LIFO) con capacidad máxima opcional."""
    def __init__(self, capacidad=None):
        self._datos = []
        self._capacidad = capacidad

    def push(self, valor):
        if self._capacidad and len(self._datos) >= self._capacidad:
            raise CapacidadExcedidaError(self._capacidad)
        self._datos.append(valor)

    def pop(self):
        if self.esta_vacia():
            raise EstructuraVaciaError("No se puede hacer pop en pila vacía")
        return self._datos.pop()

    def peek(self):
        if self.esta_vacia():
            raise EstructuraVaciaError("No se puede hacer peek en pila vacía")
        return self._datos[-1]

    def esta_vacia(self):
        return len(self._datos) == 0

    def __len__(self):
        return len(self._datos)

# Demostración
pila = PilaSimple(capacidad=3)
for v in [10, 20, 30]:
    pila.push(v)
    print(f"  push({v}) → tamaño={len(pila)}")

try:
    pila.push(40)   # excede capacidad
except CapacidadExcedidaError as e:
    print(f"  CapacidadExcedidaError: {e}")

while not pila.esta_vacia():
    print(f"  pop() → {pila.pop()}")

try:
    pila.pop()      # pila vacía
except EstructuraVaciaError as e:
    print(f"  EstructuraVaciaError: {e}")
````

### 🧪 Pruebas con unittest

````python
class TestFuncionesExcepciones(unittest.TestCase):

    # ── potencia ──────────────────────────────────────────────────────────────
    def test_potencia_cuadrado(self):
        self.assertEqual(potencia(5), 25)       # default exp=2

    def test_potencia_explicito(self):
        self.assertEqual(potencia(2, 10), 1024)

    def test_potencia_cero(self):
        self.assertEqual(potencia(99, 0), 1)    # cualquier base^0 = 1

    def test_potencia_negativo_lanza(self):
        with self.assertRaises(ValueError):
            potencia(5, -1)

    # ── suma variadic ─────────────────────────────────────────────────────────
    def test_suma_vacia(self):
        self.assertEqual(suma(), 0)

    def test_suma_varios(self):
        self.assertEqual(suma(1, 2, 3, 4, 5), 15)

    # ── busqueda_binaria ─────────────────────────────────────────────────────
    def test_binaria_encontrado(self):
        self.assertEqual(busqueda_binaria([1,3,5,7,9], 7), 3)

    def test_binaria_no_encontrado(self):
        self.assertEqual(busqueda_binaria([1,3,5,7,9], 4), -1)

    def test_binaria_desordenada_lanza(self):
        with self.assertRaises(ValueError):
            busqueda_binaria([3, 1, 2], 2)

    # ── PilaSimple ────────────────────────────────────────────────────────────
    def test_pila_vacia_lanza(self):
        p = PilaSimple()
        with self.assertRaises(EstructuraVaciaError):
            p.pop()

    def test_pila_capacidad_lanza(self):
        p = PilaSimple(capacidad=2)
        p.push(1); p.push(2)
        with self.assertRaises(CapacidadExcedidaError):
            p.push(3)

    def test_pila_push_pop(self):
        p = PilaSimple()
        for v in [10, 20, 30]:
            p.push(v)
        self.assertEqual(p.pop(), 30)   # LIFO
        self.assertEqual(len(p), 2)

unittest.main(argv=[''], verbosity=2, exit=False)
````

### 🔍 Funciones con type hints (anotaciones de tipo)

````python
# Type hints: no imponen tipos en tiempo de ejecución, pero mejoran legibilidad
# y permite análisis estático con herramientas como mypy
from typing import Optional, List

def maximo(valores: List[int | float]) -> Optional[int | float]:
    """
    Retorna el máximo de una lista.
    
    Args:
        valores: lista de números (no vacía)
    
    Returns:
        El máximo valor, o None si la lista está vacía
    """
    if not valores:
        return None
    m = valores[0]
    for v in valores[1:]:
        if v > m:
            m = v
    return m

print("maximo([3,1,4,1,5,9,2,6]):", maximo([3,1,4,1,5,9,2,6]))
print("maximo([]):               ", maximo([   ]))
````

````python
# ── Patrones de excepción en DS&A ─────────────────────────────────────────────
print("=== Patrón: EAFP vs LBYL ===\n")

# LBYL — Look Before You Leap (estilo C)
def get_lbyl(d: dict, clave: str):
    if clave in d:          # verificación primero
        return d[clave]
    return None

# EAFP — Easier to Ask Forgiveness than Permission (estilo Python)
def get_eafp(d: dict, clave: str):
    try:
        return d[clave]
    except KeyError:
        return None

datos = {"a": 1, "b": 2}
print(f"LBYL: {get_lbyl(datos, 'a')}, {get_lbyl(datos, 'z')}")
print(f"EAFP: {get_eafp(datos, 'a')}, {get_eafp(datos, 'z')}")
print("\n→ Ambos estilos son válidos. EAFP es más pythónico cuando la")
print("  excepción es rara; LBYL cuando la verificación es barata.")
````

````python
# ── raise ... from: encadenamiento de excepciones ────────────────────────────
print("=== raise from: encadenamiento de excepciones ===\n")

class ConfigError(Exception):
    """Error de configuración del sistema."""
    pass

def cargar_config(ruta: str) -> dict:
    """Carga configuración desde un archivo (simulado)."""
    try:
        # Simulamos un FileNotFoundError
        if ruta.endswith(".missing"):
            raise FileNotFoundError(f"Archivo no encontrado: {ruta}")
        return {"clave": "valor"}
    except FileNotFoundError as e:
        raise ConfigError(f"No se pudo inicializar la configuración") from e

try:
    cargar_config("config.missing")
except ConfigError as e:
    print(f"ConfigError: {e}")
    print(f"Causa original: {e.__cause__}")
````

## 📊 Resumen

### 🎯 Conceptos Clave

| Concepto | Descripción |
|----------|-------------|
| **Funciones first-class** | Pueden asignarse a variables, pasarse y retornarse |
| **Parámetros por defecto** | Evalúan **una sola vez** en la definición (⚠️ gotcha con mutables) |
| **`*args` / `**kwargs`** | Capturan argumentos extras como tupla / dict |
| **LEGB** | Orden de búsqueda de nombres: Local → Enclosing → Global → Built-in |
| **EAFP** | Estilo pythónico: intentar y capturar en vez de verificar primero |
| **`try/except/else/finally`** | Flujo completo de manejo de errores |
| **Excepciones propias** | Subclases de `Exception` para dar semántica a errores de la estructura |

### ⚠️ Gotcha clásico: default mutable

```python
# ¡Incorrecto! La lista [] se crea UNA sola vez
def agregar(x, lista=[]):
    lista.append(x)
    return lista

print(agregar(1))  # [1]
print(agregar(2))  # [1, 2]  ← ¡no es [2]!

# Correcto: usar None y crear la lista adentro
def agregar_ok(x, lista=None):
    if lista is None:
        lista = []
    lista.append(x)
    return lista
```

### 🔄 Aplicaciones en DS&A

- **`try/except`** se usa en `pop()` y `peek()` de pilas/colas para detectar estructura vacía
- **Excepciones propias** dan mensajes claros al usuario de la estructura (`EstructuraVaciaError`)
- **Type hints** mejoran la documentación de métodos de ADTs sin costo en runtime
