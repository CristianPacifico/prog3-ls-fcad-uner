### 📈 Visualización del algoritmo de Euclides

````python
def euclides_pasos(a: int, b: int) -> list:
    """Retorna la lista de pares (a, b) por los que pasa el algoritmo."""
    pasos = [(a, b)]
    while b != 0:
        a, b = b, a % b
        pasos.append((a, b))
    return pasos

# Trazar pasos para (1071, 462)
pasos = euclides_pasos(1071, 462)
print("Pasos del algoritmo de Euclides para mcd(1071, 462):")
for i, (a, b) in enumerate(pasos):
    if b != 0:
        print(f"  paso {i}: mcd({a}, {b}) = mcd({b}, {a} mod {b} = {a % b})")
    else:
        print(f"  paso {i}: mcd({a}, 0) → resultado = {a}")

fig, ax = plt.subplots(figsize=(8, 4))
valores_a = [p[0] for p in pasos]
valores_b = [p[1] for p in pasos]
ax.plot(valores_a, 'b-o', label='a', linewidth=2)
ax.plot(valores_b, 'r-s', label='b', linewidth=2)
ax.set_xlabel("Paso")
ax.set_ylabel("Valor")
ax.set_title("Convergencia del Algoritmo de Euclides — mcd(1071, 462)")
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

# ── Ej 1: conteo_frecuencias ────────────────────────────────────────────────
def conteo_frecuencias(texto):
    freq = {}
    for c in texto:
        if c.isalpha():
            c = c.lower()
            freq[c] = freq.get(c, 0) + 1
    return freq

# ── Ej 2: operaciones_conjunto ──────────────────────────────────────────────
def operaciones_conjunto(lst1, lst2):
    s1, s2 = set(lst1), set(lst2)
    return (s1 & s2, s1 | s2, s1 - s2, s1 ^ s2)

# ── Ej 3: comprimir_rlc ─────────────────────────────────────────────────────
def comprimir_rlc(s):
    if not s:
        return ""
    resultado = ""
    char_actual, conteo = s[0], 1
    for c in s[1:]:
        if c == char_actual:
            conteo += 1
        else:
            resultado += char_actual + str(conteo)
            char_actual, conteo = c, 1
    return resultado + char_actual + str(conteo)

def descomprimir_rlc(s):
    resultado = ""
    i = 0
    while i < len(s):
        char = s[i]
        i += 1
        num_str = ""
        while i < len(s) and s[i].isdigit():
            num_str += s[i]
            i += 1
        resultado += char * int(num_str)
    return resultado

# ── Ej 4: mcd y mcm ─────────────────────────────────────────────────────────
def mcd_euclides(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def mcm(a, b):
    if a == 0 or b == 0:
        return 0
    return abs(a * b) // mcd_euclides(a, b)

# ── Desafío: evaluar_rpn ─────────────────────────────────────────────────────
def evaluar_rpn(expresion):
    if not expresion.strip():
        raise ValueError("Expresión vacía")
    ops = {'+': lambda a,b: a+b, '-': lambda a,b: a-b,
           '*': lambda a,b: a*b, '/': lambda a,b: a/b,
           '//': lambda a,b: a//b, '%': lambda a,b: a%b,
           '**': lambda a,b: a**b}
    stack = []
    for token in expresion.split():
        if token in ops:
            if len(stack) < 2:
                raise ValueError(f"Stack insuficiente para operador '{token}'")
            b, a = stack.pop(), stack.pop()
            stack.append(ops[token](a, b))
        else:
            try:
                stack.append(float(token) if '.' in token else int(token))
            except ValueError:
                raise ValueError(f"Token inválido: '{token}'")
    if len(stack) != 1:
        raise ValueError("Expresión mal formada")
    return stack[0]

````python
# ── Bonus: comprimir_rlc con itertools.groupby (solución elegante) ─────────────
import itertools

def comprimir_rlc_groupby(s: str) -> str:
    """Solución elegante usando itertools.groupby."""
    return "".join(c + str(sum(1 for _ in grupo))
                   for c, grupo in itertools.groupby(s))

print("Solución con groupby:")
for s in ["aaabbc", "abcde", "aabbccddee"]:
    manual  = comprimir_rlc(s) if s else ""
    elegant = comprimir_rlc_groupby(s)
    match   = "✅" if manual == elegant else f"❌ (manual={manual})"
    print(f"  '{s}' → '{elegant}' {match}")
````
