# 🔷 Sección 1: El Modelo de Objetos de Python

## 🎯 Introducción

> *"Python is a purely object-oriented language, and each object has three traits: an **identity**, a **type**, and a **value**."*  
> — Goodrich, Tamassia & Goldwasser, §1.2

**Analogía:** Pensá en un objeto Python como una persona:
- 🆔 **Identidad** (`id`): número de DNI — único, no cambia  
- 🏷️ **Tipo** (`type`): profesión — define qué operaciones puede hacer  
- 📄 **Valor**: su contenido actual — puede o no cambiar según el objeto

En Python, **todo** es un objeto: números, texto, funciones, clases, módulos.

## 🔑 Conceptos Clave

| Concepto | Descripción | Función/Operador |
|----------|-------------|-----------------|
| Identidad | Dirección en memoria (única por objeto en su vida) | `id(obj)` |
| Tipo | Clase del objeto, define operaciones válidas | `type(obj)` |
| Igualdad de identidad | ¿Son el **mismo** objeto en memoria? | `a is b` |
| Igualdad de valor | ¿Tienen el **mismo** valor? | `a == b` |
| Mutabilidad | ¿Puede cambiar su valor sin cambiar su identidad? | — |

````python
# ── Identidad, tipo y valor ──────────────────────────────────────────────────
x = 42
print(f"valor:     {x}")
print(f"tipo:      {type(x)}")
print(f"identidad: {id(x)}")

# is vs ==
a = [1, 2, 3]
b = [1, 2, 3]   # mismo VALOR, diferente objeto
c = a            # mismo OBJETO (alias)

print("\n--- is vs == ---")
print(f"a == b → {a == b}   (mismo valor)")
print(f"a is b → {a is b}   (¿mismo objeto?)")
print(f"a is c → {a is c}   (c es alias de a)")
print(f"id(a)={id(a)}, id(b)={id(b)}, id(c)={id(c)}")

# ── Mutabilidad ───────────────────────────────────────────────────────────────
print("\n--- Mutabilidad ---")
# list es MUTABLE: cambia su valor sin cambiar su identidad
lista = [10, 20, 30]
id_antes = id(lista)
lista.append(40)
print(f"list  mutable:   id antes={id_antes}, id después={id(lista)}, igual={id_antes == id(lista)}")

# tuple es INMUTABLE: no se puede modificar
tupla = (10, 20, 30)
id_antes = id(tupla)
# tupla.append(40)  → AttributeError
tupla = tupla + (40,)  # crea un NUEVO objeto
print(f"tuple inmutable: id antes={id_antes}, id después={id(tupla)}, igual={id_antes == id(tupla)}")

# str también es INMUTABLE
s = "hola"
id_antes = id(s)
s += " mundo"   # s apunta a un NUEVO objeto
print(f"str   inmutable: id antes={id_antes}, id después={id(s)}, igual={id_antes == id(s)}")
````

### ⚠️ Aliasing y el riesgo de los objetos mutables

El **aliasing** ocurre cuando dos variables apuntan al **mismo** objeto. Con objetos mutables, modificar uno modifica el otro porque son el mismo.

```python
a = [1, 2, 3]
b = a           # alias — mismo objeto
b.append(99)
print(a)        # → [1, 2, 3, 99]  ← ¡a cambió!
```

**Cómo evitarlo:**
- `copy.copy(obj)` — copia superficial (shallow): copia el contenedor, no los objetos internos
- `copy.deepcopy(obj)` — copia profunda (deep): copia recursivamente todo

> **⚡ Implicancia para estructuras de datos:** al pasar listas a funciones, Python pasa la referencia al objeto, no una copia. Si la función modifica la lista, el cambio es visible afuera. Tenerlo en cuenta al implementar estructuras.

# 🔷 Sección 2: Tipos Built-in de Python

## 🎯 Introducción

Python ofrece un rico conjunto de **tipos integrados**. Cada tipo define:
- Las operaciones válidas sobre sus instancias
- La semántica de esas operaciones
- La complejidad típica de cada operación

## 📊 Tipos principales y complejidades clave

| Tipo | Mutable | Ordenado | Ejemplo literal | Uso típico en DS&A |
|------|---------|----------|-----------------|---------------------|
| `int` | No | — | `42`, `-7` | contadores, índices |
| `float` | No | — | `3.14` | pesos, probabilidades |
| `bool` | No | — | `True`, `False` | flags, condiciones |
| `str` | No | Sí | `"hola"` | etiquetas, tokens |
| `list` | **Sí** | Sí | `[1,2,3]` | arrays dinámicos (cap 5) |
| `tuple` | No | Sí | `(1,2,3)` | registros inmutables |
| `dict` | **Sí** | (desde 3.7) | `{"a":1}` | mapas / tablas hash (cap 10) |
| `set` | **Sí** | No | `{1,2,3}` | conjuntos, deduplicación |
| `frozenset` | No | No | `frozenset({1,2})` | claves de dict con conjuntos |

````python
# ── Números ───────────────────────────────────────────────────────────────────
print("=== Tipos Numéricos ===")
print(f"int:     {42 + 7}   {type(42)}")
print(f"float:   {3.14 * 2:.4f}   {type(3.14)}")
print(f"bool:    {True and False}   {type(True)} (bool es subclase de int)")
print(f"int//float: {7 // 2} (división entera)  {7 % 2} (módulo)  {2**10} (potencia)")

# ── str ───────────────────────────────────────────────────────────────────────
print("\n=== str — secuencia inmutable de caracteres ===")
s = "Estructuras de Datos"
print(f"len: {len(s)}")             # O(1) — almacenado
print(f"[0]: {s[0]}")              # O(1) — acceso por índice
print(f"[-1]: {s[-1]}")            # O(1) — índice negativo
print(f"[0:12]: {s[0:12]}")        # O(k) — slice de largo k
print(f"'Datos' in s: {'Datos' in s}")  # O(n) — subcadena

# ── list ──────────────────────────────────────────────────────────────────────
print("\n=== list — array dinámico mutable ===")
lst = [3, 1, 4, 1, 5, 9, 2, 6]
print(f"list:      {lst}")
print(f"lst[2]:    {lst[2]}   O(1)")
lst.append(7)
print(f"append(7): {lst}  O(1) amortizado")
lst.sort()
print(f"sort():    {lst}  O(n log n)")
print(f"2 in lst:  {2 in lst}  O(n)")

# ── dict ──────────────────────────────────────────────────────────────────────
print("\n=== dict — tabla hash mutable ===")
d = {"uno": 1, "dos": 2, "tres": 3}
print(f"d['dos']:    {d['dos']}   O(1) promedio")
d["cuatro"] = 4
print(f"inserción:   {d}   O(1) promedio")
print(f"'tres' in d: {'tres' in d}   O(1) promedio")

# ── set ───────────────────────────────────────────────────────────────────────
print("\n=== set — conjunto con hash ===")
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7}
print(f"unión:        {a | b}")
print(f"intersección: {a & b}")
print(f"diferencia:   {a - b}")
print(f"3 in a:       {3 in a}   O(1) promedio")
````

### 📈 Complejidades de operaciones en contenedores

Comparemos visualmente:  `list` vs `dict/set` para búsqueda por valor.

````python
import random, timeit

sizes = [100, 500, 1_000, 5_000, 10_000, 50_000]
t_list, t_set, t_dict = [], [], []

for n in sizes:
    data = list(range(n))
    random.shuffle(data)
    target = data[n // 2]   # valor que existe (caso promedio)

    lst    = data[:]
    s      = set(data)
    d      = {v: True for v in data}

    reps = max(1, 10_000 // n)   # más reps para n pequeños

    t_list.append(timeit.timeit(lambda: target in lst, number=reps) / reps * 1e6)
    t_set.append( timeit.timeit(lambda: target in s,   number=reps) / reps * 1e6)
    t_dict.append(timeit.timeit(lambda: target in d,   number=reps) / reps * 1e6)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

ax1.plot(sizes, t_list, 'b-o', label='list  (O(n))', linewidth=2)
ax1.plot(sizes, t_set,  'g-s', label='set   (O(1))', linewidth=2)
ax1.plot(sizes, t_dict, 'r-^', label='dict  (O(1))', linewidth=2)
ax1.set_xlabel('Tamaño n')
ax1.set_ylabel('Tiempo (µs)')
ax1.set_title('Búsqueda: list vs set vs dict')
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2.plot(sizes, t_list, 'b-o', label='list  (O(n))', linewidth=2)
ax2.plot(sizes, t_set,  'g-s', label='set   (O(1))', linewidth=2)
ax2.plot(sizes, t_dict, 'r-^', label='dict  (O(1))', linewidth=2)
ax2.set_xlabel('Tamaño n')
ax2.set_ylabel('Tiempo (µs) — escala log')
ax2.set_title('Misma comparación (escala logarítmica)')
ax2.set_yscale('log')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
print("→ list crece linealmente; set y dict permanecen casi constantes.")
````

````python
# ── Copia shallow vs deep ────────────────────────────────────────────────────
print("=== Copia superficial (shallow) vs profunda (deep) ===\n")

original = [[1, 2], [3, 4], [5, 6]]

shallow = copy.copy(original)     # nuevo contenedor, mismos objetos internos
deep    = copy.deepcopy(original) # copia recursiva completa

# Modificar un elemento interno
original[0].append(99)

print(f"original: {original}")
print(f"shallow:  {shallow}  ← ¡también cambió! (comparte lista interna)")
print(f"deep:     {deep}   ← independiente ✅")

print()
# Para listas 1D (sin anidamiento), copy() y deepcopy() son equivalentes:
a = [10, 20, 30]
b = copy.copy(a)
a.append(40)
print(f"a: {a}")
print(f"b (shallow de lista plana): {b}  ← independiente ✅")
````

# 🔷 Sección 3: Operadores y Control de Flujo

## 🎯 Introducción

Los operadores de Python y las estructuras de control son las herramientas con las que construimos **todos los algoritmos**.

## 📊 Tabla de Operadores (Goodrich §1.3)

| Categoría | Operadores | Ejemplo | Resultado |
|-----------|-----------|---------|-----------|
| Aritméticos | `+`, `-`, `*`, `/`, `//`, `%`, `**` | `7 // 2` | `3` |
| Relacionales | `<`, `<=`, `>`, `>=`, `==`, `!=` | `3 != 4` | `True` |
| Lógicos | `and`, `or`, `not` | `not False` | `True` |
| Bitwise | `&`, `\|`, `^`, `~`, `<<`, `>>` | `5 & 3` | `1` |
| Asignación | `=`, `+=`, `-=`, `*=`, `/=`, `//=`, `%=` | `x += 1` | — |
| Identidad | `is`, `is not` | `x is None` | bool |
| Pertenencia | `in`, `not in` | `3 in [1,2,3]` | `True` |

### ⚠️ Precedencia (de mayor a menor)
`**` → `+x -x ~x` → `* / // %` → `+ -` → `<< >>` → `&` → `^` → `|` → comparaciones → `not` → `and` → `or`

**Regla práctica:** ante duda, usar paréntesis.

````python
# ── Operadores ────────────────────────────────────────────────────────────────
print("=== Operadores ===")
print(f"División:        7 / 2  = {7 / 2}    (float)")
print(f"División entera: 7 // 2 = {7 // 2}    (int, hacia -∞)")
print(f"Módulo:          7 % 2  = {7 % 2}    (resto)")
print(f"Potencia:        2**10  = {2**10}")
print(f"Bitwise AND:     5 & 3  = {5 & 3}    (101 & 011 = 001)")
print(f"Shift izquierda: 1 << 4 = {1 << 4}   (= 2^4)")
print()

# ── Condicionales ─────────────────────────────────────────────────────────────
print("=== if / elif / else ===")

def clasificar_nota(nota):
    """Clasifica una nota numérica. §1.4.1"""
    if nota >= 90:
        return "Sobresaliente"
    elif nota >= 70:
        return "Aprobado"
    elif nota >= 50:
        return "Regular"
    else:
        return "Desaprobado"

for n in [95, 75, 55, 30]:
    print(f"  nota {n}: {clasificar_nota(n)}")

# ── while ─────────────────────────────────────────────────────────────────────
print("\n=== while — búsqueda lineal ===")
def busqueda_lineal(lista, objetivo):
    """Búsqueda lineal con while. Retorna índice o -1."""
    i = 0
    while i < len(lista):
        if lista[i] == objetivo:
            return i
        i += 1
    return -1

data = [4, 7, 2, 9, 1, 5]
print(f"  buscar 9 en {data}: índice {busqueda_lineal(data, 9)}")
print(f"  buscar 8 en {data}: índice {busqueda_lineal(data, 8)}")

# ── for + break + continue ────────────────────────────────────────────────────
print("\n=== for con break, continue y else ===")
nums = [3, 6, 2, 8, 1, 7, 4]
objetivo = 8

for i, v in enumerate(nums):
    if v % 2 != 0:
        continue           # omitir impares
    if v == objetivo:
        print(f"  Encontrado {objetivo} en índice {i}")
        break
else:
    # El else se ejecuta si el for termina SIN break
    print(f"  {objetivo} no encontrado entre los pares")

# for sobre dict
print("\n=== Iteración sobre dict ===")
frecuencias = {"a": 3, "b": 1, "c": 5}
for clave, valor in frecuencias.items():
    print(f"  '{clave}': {'█' * valor} ({valor})")
````

### 🧪 Pruebas con unittest

````python
class TestModeloObjetos(unittest.TestCase):

    # ── Identidad vs igualdad ─────────────────────────────────────────────────
    def test_is_vs_igual(self):
        a = [1, 2, 3]
        b = [1, 2, 3]
        self.assertEqual(a, b)       # mismo valor
        self.assertIsNot(a, b)       # diferente objeto

    def test_alias(self):
        a = [1, 2, 3]
        c = a
        self.assertIs(a, c)          # mismo objeto
        c.append(4)
        self.assertEqual(a, [1, 2, 3, 4])  # cambio visible por alias

    # ── Mutabilidad ────────────────────────────────────────────────────────────
    def test_list_mutable(self):
        lst = [1, 2, 3]
        id_orig = id(lst)
        lst.append(4)
        self.assertEqual(id(lst), id_orig)   # mismo objeto
        self.assertEqual(lst[-1], 4)

    def test_str_inmutable(self):
        s = "hola"
        id_orig = id(s)
        s += " mundo"
        self.assertNotEqual(id(s), id_orig)  # nuevo objeto
        self.assertEqual(s, "hola mundo")

    # ── Tipos ──────────────────────────────────────────────────────────────────
    def test_bool_es_int(self):
        self.assertIsInstance(True, int)
        self.assertEqual(True + True + False, 2)

    def test_busqueda_lineal(self):
        self.assertEqual(busqueda_lineal([4,7,2,9], 9), 3)
        self.assertEqual(busqueda_lineal([4,7,2,9], 5), -1)

unittest.main(argv=[''], verbosity=2, exit=False)
````

````python
# ── Shallow vs deep copy: verificación ───────────────────────────────────────
print("=== Verificación shallow vs deep copy ===\n")
import copy

# Lista 2D
original = [[1, 2], [3, 4]]
shallow  = copy.copy(original)
deep     = copy.deepcopy(original)

original[0][0] = 99

print(f"Después de original[0][0] = 99:")
print(f"  original: {original}")
print(f"  shallow:  {shallow}   ← cambió (comparte referencia interna)")
print(f"  deep:     {deep}    ← no cambió ✅")

# Interesante: el interning de enteros pequeños
print("\n=== Interning de enteros pequeños ===")
a = 256
b = 256
print(f"a=256, b=256  → a is b: {a is b}   (CPython internea -5..256)")
a = 1000
b = 1000
# En módulo interactivo puede variar; en script suelen no ser el mismo obj.
print(f"a=1000, b=1000 → a == b: {a == b}, a is b: {a is b}  (objetos distintos posibles)")
````

````python
# ── Patrones de control útiles en DS&A ───────────────────────────────────────
print("=== Patrones de control comunes en algoritmos ===\n")

# Patrón 1: Búsqueda con flag
def contiene_par(lst):
    """Retorna True si la lista contiene algún número par."""
    for x in lst:
        if x % 2 == 0:
            return True   # early exit — no recorre todo
    return False

print("Patrón 1 — early exit:")
print(f"  [1,3,5,4,7] tiene par: {contiene_par([1,3,5,4,7])}")
print(f"  [1,3,5,7]   tiene par: {contiene_par([1,3,5,7])}")

# Patrón 2: Acumulador con dict (frecuencias)
def frecuencias(iterable):
    """Cuenta ocurrencias de cada elemento. O(n)."""
    conteo = {}
    for elem in iterable:
        conteo[elem] = conteo.get(elem, 0) + 1
    return conteo

texto = "anagramas"
print(f"\nPatrón 2 — acumulador con dict:")
print(f"  frecuencias('{texto}'): {frecuencias(texto)}")

# Patrón 3: Dos punteros (preview de cap 5)
def es_palindromo(s):
    """Verifica palíndromo con dos punteros. O(n) tiempo, O(1) espacio."""
    lo, hi = 0, len(s) - 1
    while lo < hi:
        if s[lo] != s[hi]:
            return False
        lo += 1
        hi -= 1
    return True

print(f"\nPatrón 3 — dos punteros:")
for palabra in ["radar", "level", "python", "racecar"]:
    print(f"  es_palindromo('{palabra}'): {es_palindromo(palabra)}")
````

## 📊 Resumen

### 🎯 Conceptos Clave

| Concepto | Descripción |
|----------|-------------|
| **Todo es objeto** | Cada valor en Python tiene id, type y value |
| **`is` vs `==`** | `is` → mismo objeto; `==` → mismo valor |
| **Mutabilidad** | list, dict, set son mutables; int, str, tuple son inmutables |
| **Aliasing** | Dos variables apuntando al mismo objeto mutable → riesgo de efectos no deseados |
| **Copia** | `copy.copy` → shallow; `copy.deepcopy` → deep |

### 📈 Complejidades a memorizar

| Tipo | Operación | Complejidad |
|------|-----------|-------------|
| `list` | acceso `lst[i]`, `len(lst)` | O(1) |
| `list` | `append`, `pop()` | O(1) amortizado |
| `list` | búsqueda `x in lst`, `insert(i,v)` | O(n) |
| `list` | `sort()` | O(n log n) |
| `dict` / `set` | búsqueda, inserción, eliminación | O(1) promedio |
| `str` | concatenación `s + t` | O(len(s) + len(t)) |

### 🔄 Aplicaciones en DS&A

- **Mutabilidad** es clave para implementar estructuras que cambian (listas enlazadas, pilas, colas)
- **dict y set** son la base del Capítulo 10 (Maps, Hash Tables)
- **list** es la base del Capítulo 5 (Array-Based Sequences)
