## Ejercicio 1 — Ordenar funciones por crecimiento 🟢

**Tarea:** Ordenar las siguientes funciones de **menor a mayor** tasa de crecimiento asintótico:

$$f_1(n) = n^2 \quad f_2(n) = 2^n \quad f_3(n) = n \log n \quad f_4(n) = n^3 \quad f_5(n) = \log n \quad f_6(n) = n \quad f_7(n) = 1 \quad f_8(n) = n!$$

**Pistas:**
- Recordá la jerarquía: constante ≪ logarítmica ≪ lineal ≪ n·log n ≪ polinomial ≪ exponencial ≪ factorial
- Para distinguir $n^2$ de $n^3$: evaluá en n=100: $10^4$ vs $10^6$
- $n!$ crece más rápido que $2^n$ para $n > 2$

````python
# Ejercicio 1: solución y verificación numérica
orden_correcto = ["O(1)", "O(log n)", "O(n)", "O(n log n)", "O(n²)", "O(n³)", "O(2ⁿ)", "O(n!)"]
print("Orden de menor a mayor crecimiento:")
for i, f in enumerate(orden_correcto, 1):
    print(f"  {i}. {f}")

n = 10
valores = {
    "O(1)":       1,
    "O(log n)":   math.log2(n),
    "O(n)":       n,
    "O(n log n)": n * math.log2(n),
    "O(n²)":      n**2,
    "O(n³)":      n**3,
    "O(2ⁿ)":      2**n,
    "O(n!)":      math.factorial(n),
}
print(f"\nValores para n={n}:")
for nombre, val in sorted(valores.items(), key=lambda x: x[1]):
    print(f"  {nombre:<14} = {val:>15,.0f}")

n_vals = np.linspace(1, 30, 200)
fig, ax = plt.subplots(figsize=(10, 5))
funciones_plot = [
    ("O(1)",       np.ones_like(n_vals),          'gray'),
    ("O(log n)",   np.log2(n_vals),                'blue'),
    ("O(n)",       n_vals,                         'green'),
    ("O(n log n)", n_vals * np.log2(n_vals),        'orange'),
    ("O(n²)",      n_vals**2,                      'red'),
]
for label, y, color in funciones_plot:
    ax.plot(n_vals, y, label=label, color=color, linewidth=2)
ax.set_xlim(1, 30); ax.set_ylim(0, 900)
ax.set_xlabel('n'); ax.set_ylabel('f(n)')
ax.set_title('Comparación de funciones de crecimiento')
ax.legend(); plt.tight_layout(); plt.show()
````

# Fragmento D  (len(data) = n)

````python
def fragmento_a(n):
    """O(n): un bucle simple, O(1) por iteración."""
    resultado = 0
    for i in range(n):
        resultado += i * i
    return resultado

def fragmento_b(n):
    """O(n²): dos bucles anidados independientes."""
    resultado = 0
    for i in range(n):
        for j in range(n):
            resultado += i + j
    return resultado

def fragmento_c(n):
    """O(log n): i se duplica cada iteración -> log₂(n) pasos."""
    resultado = 0; i = 1
    while i < n:
        resultado += i; i *= 2
    return resultado

def fragmento_d(data):
    """O(n log n): sorted() en el else domina para n >= 10."""
    if len(data) < 10:
        return sum(data)
    else:
        return sorted(data)[0]

print("Conteo iteraciones fragmento_c (O(log n)):")
for n in [2, 4, 8, 16, 64, 1024]:
    pasos = 0; i = 1
    while i < n:
        pasos += 1; i *= 2
    print(f"  n={n:>5}: {pasos} pasos  (log₂ = {math.log2(n):.1f})")

print("\nRespuestas:")
print("  Fragmento A: O(n)        — bucle simple")
print("  Fragmento B: O(n²)       — dos bucles anidados independientes")
print("  Fragmento C: O(log n)    — i se duplica: log₂(n) pasos")
print("  Fragmento D: O(n log n)  — sorted() en el else domina")
````

## Ejercicio 3 — `prefix_average` desde cero 🟡

Implementá **las dos versiones** de promedios prefijos del libro:

- `mi_prefix_avg1(S)` — versión naïve con doble bucle: **O(n²)**
- `mi_prefix_avg2(S)` — versión optimizada con suma acumulada: **O(n)**

**Especificación:** dado `S = [s₀, s₁, …, s_{n-1}]`, devolver `A` donde:
$$A[j] = \frac{s_0 + s_1 + \cdots + s_j}{j+1}$$

**Pistas:**
- En `mi_prefix_avg1`: dos bucles — externo en `j`, interno acumula `S[0..j]`
- En `mi_prefix_avg2`: una variable `total` actualizada con `total += S[j]`

````python
def mi_prefix_avg1(S):
    """Versión O(n²): doble bucle, recomputa la suma cada vez."""
    n = len(S)
    A = [0.0] * n
    for j in range(n):
        total = 0
        for i in range(j + 1):
            total += S[i]
        A[j] = total / (j + 1)
    return A

def mi_prefix_avg2(S):
    """Versión O(n): suma acumulada, una sola pasada."""
    n = len(S)
    A = [0.0] * n
    total = 0
    for j in range(n):
        total += S[j]
        A[j] = total / (j + 1)
    return A

S = [10, 20, 30, 40, 50]
a1, a2 = mi_prefix_avg1(S), mi_prefix_avg2(S)
print(f"S  = {S}")
print(f"A1 = {a1}")
print(f"A2 = {a2}")
print(f"Iguales: {a1 == a2}")
````

### Benchmark y visualización

````python
tamanios = [200, 500, 1000, 2000, 4000]
t1_list, t2_list = [], []

print(f"{'n':>6} {'avg1 O(n²) ms':>15} {'avg2 O(n) ms':>14} {'speedup':>9}")
print("-" * 48)
for n in tamanios:
    S = list(range(n))
    t1 = timeit.timeit(lambda: mi_prefix_avg1(S), number=10) / 10 * 1000
    t2 = timeit.timeit(lambda: mi_prefix_avg2(S), number=100) / 100 * 1000
    t1_list.append(t1); t2_list.append(t2)
    print(f"{n:>6} {t1:>15.3f} {t2:>14.4f} {t1/max(t2, 1e-9):>8.1f}x")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(tamanios, t1_list, 'r-o', label='prefix_avg1  O(n²)', linewidth=2)
axes[0].plot(tamanios, t2_list, 'g-o', label='prefix_avg2  O(n)',  linewidth=2)
axes[0].set_xlabel('n'); axes[0].set_ylabel('Tiempo (ms)')
axes[0].set_title('Escala lineal'); axes[0].legend()

axes[1].loglog(tamanios, t1_list, 'r-o', label='prefix_avg1  O(n²)', linewidth=2)
axes[1].loglog(tamanios, t2_list, 'g-o', label='prefix_avg2  O(n)',  linewidth=2)
axes[1].set_xlabel('n'); axes[1].set_ylabel('Tiempo (ms)')
axes[1].set_title('Log-log (pendiente = orden)'); axes[1].legend()

plt.suptitle('prefix_average: O(n²) vs O(n)', fontweight='bold')
plt.tight_layout(); plt.show()

import warnings
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    slope1 = np.polyfit(np.log(tamanios), np.log(t1_list), 1)[0]
    slope2 = np.polyfit(np.log(tamanios), np.log(t2_list), 1)[0]
print(f"\nPendiente log-log avg1: {slope1:.2f}  (esperado ~2.0)")
print(f"Pendiente log-log avg2: {slope2:.2f}  (esperado ~1.0)")
````

## Ejercicio 4 — `element_uniqueness` 🟡

**Problema:** dado un arreglo, determinar si todos sus elementos son distintos.

Implementá **dos versiones**:
- `unique1(S)` — O(n²): doble bucle, compara todos los pares `(S[j], S[k])` con `j < k`
- `unique2(S)` — O(n log n): ordenar y comparar elementos adyacentes

**¿Por qué `unique2` es O(n log n) y no O(n)?** Porque `sorted()` tarda O(n log n), que domina el recorrido O(n) posterior.

````python
def unique1(S):
    """O(n²): compara todos los pares (S[j], S[k]) con j < k."""
    n = len(S)
    for j in range(n):
        for k in range(j + 1, n):
            if S[j] == S[k]:
                return False
    return True

def unique2(S):
    """O(n log n): ordena y revisa adyacentes."""
    temp = sorted(S)
    for j in range(1, len(temp)):
        if temp[j] == temp[j - 1]:
            return False
    return True

casos = [
    ([1, 2, 3, 4, 5],  True,  "todos distintos"),
    ([1, 2, 3, 2, 5],  False, "tiene duplicado"),
    ([1],              True,  "un solo elemento"),
    ([5, 5],           False, "dos iguales"),
]
print(f"{'unique1':>8} {'unique2':>8}  caso")
for S, esperado, desc in casos:
    r1, r2 = unique1(S), unique2(S)
    ok = "OK" if r1 == r2 == esperado else "FAIL"
    print(f"  {str(r1):>6} {str(r2):>8}  {ok}  {desc}")

print("\nBenchmark (peor caso — sin duplicados):")
print(f"{'n':>8} {'unique1 ms':>12} {'unique2 ms':>12}")
for n in [200, 1000, 3000]:
    S = list(range(n))
    t1 = timeit.timeit(lambda: unique1(S), number=20) / 20 * 1000
    t2 = timeit.timeit(lambda: unique2(S), number=200) / 200 * 1000
    print(f"{n:>8,} {t1:>12.3f} {t2:>12.4f}")
````

## ⭐ Desafío — `three_way_disjoint`: tres versiones 🔴🔴

**Problema:** dados tres arreglos A, B, C (cada uno de tamaño n, sin repeticiones internas), determinar si son **mutuamente disjuntos**.

| Versión | Complejidad | Estrategia |
|---------|-------------|-----------|
| `disjoint_v1` | O(n³) | Triple bucle — fuerza bruta |
| `disjoint_v2` | O(n²) | Doble bucle + guardia: buscar en C solo si `a == b` |
| `disjoint_v3` | O(n²) con set | Convertir C a `set` para lookups O(1) |

**Pregunta clave:** ¿Por qué `disjoint_v2` es O(n²) aunque tiene **tres** bucles `for`?

> **Pista:** El bucle sobre C solo corre cuando `a == b`, lo cual ocurre O(n) veces en O(n²) pares.

````python
def disjoint_v1(A, B, C):
    """O(n³): fuerza bruta."""
    for a in A:
        for b in B:
            for c in C:
                if a == b == c:
                    return False
    return True

def disjoint_v2(A, B, C):
    """O(n²): guardia — el bucle sobre C solo corre cuando a==b."""
    for a in A:
        for b in B:
            if a == b:
                for c in C:
                    if a == c:
                        return False
    return True

def disjoint_v3(A, B, C):
    """O(n²): igual que v2 pero lookup en set -> O(1)."""
    C_set = set(C)
    for a in A:
        for b in B:
            if a == b and a in C_set:
                return False
    return True

A1=[1,2,3]; B1=[4,5,6]; C1=[7,8,9]
A2=[1,2,3]; B2=[2,4,5]; C2=[6,7,2]
for f in (disjoint_v1, disjoint_v2, disjoint_v3):
    print(f"{f.__name__}: disjunto={f(A1,B1,C1)}  con_interseccion={f(A2,B2,C2)}")
````

### Benchmark de las tres versiones

````python
print(f"{'n':>6} {'v1 O(n³) ms':>14} {'v2 O(n²) ms':>14} {'v3 O(n²) ms':>14}")
print("-" * 54)
for n in [30, 60, 100, 200]:
    A = list(range(n)); B = list(range(n, 2*n)); C = list(range(2*n, 3*n))
    t1 = timeit.timeit(lambda: disjoint_v1(A,B,C), number=5)/5*1000
    t2 = timeit.timeit(lambda: disjoint_v2(A,B,C), number=50)/50*1000
    t3 = timeit.timeit(lambda: disjoint_v3(A,B,C), number=50)/50*1000
    print(f"{n:>6} {t1:>14.2f} {t2:>14.3f} {t3:>14.3f}")

print("\nPor qué disjoint_v2 es O(n²) con 3 loops:")
print("El loop sobre C corre SOLO cuando a==b.")
print("Con A, B sin duplicados: a==b ocurre O(n) veces en O(n²) pares.")
print("-> Total: O(n²) + O(n)*O(n) = O(n²).")
````

## 🧪 Suite de Tests

````python
class TestPrefixAverage(unittest.TestCase):
    def test_iguales(self):
        S = list(range(1, 11))
        self.assertEqual(mi_prefix_avg1(S), mi_prefix_avg2(S))
    def test_primer_elemento(self):
        S = [7, 2, 9]
        self.assertAlmostEqual(mi_prefix_avg2(S)[0], 7.0)
    def test_promedio_correcto(self):
        S = [1, 3, 5]
        a = mi_prefix_avg2(S)
        self.assertAlmostEqual(a[2], 3.0)

class TestUnique(unittest.TestCase):
    def test_unicos(self):
        self.assertTrue(unique1([1,2,3]))
        self.assertTrue(unique2([1,2,3]))
    def test_duplicado(self):
        self.assertFalse(unique1([1,2,1]))
        self.assertFalse(unique2([1,2,1]))
    def test_un_elemento(self):
        self.assertTrue(unique1([5]))
        self.assertTrue(unique2([5]))

class TestDisjoint(unittest.TestCase):
    def setUp(self):
        self.A=[1,2,3]; self.B=[4,5,6]; self.C=[7,8,9]
        self.A2=[1,2,3]; self.B2=[2,4,5]; self.C2=[6,7,2]
    def test_v1_disjoint(self):
        self.assertTrue(disjoint_v1(self.A, self.B, self.C))
    def test_v2_disjoint(self):
        self.assertTrue(disjoint_v2(self.A, self.B, self.C))
    def test_v3_disjoint(self):
        self.assertTrue(disjoint_v3(self.A, self.B, self.C))
    def test_v1_not_disjoint(self):
        self.assertFalse(disjoint_v1(self.A2, self.B2, self.C2))
    def test_v2_not_disjoint(self):
        self.assertFalse(disjoint_v2(self.A2, self.B2, self.C2))
    def test_v3_not_disjoint(self):
        self.assertFalse(disjoint_v3(self.A2, self.B2, self.C2))

all_suites = unittest.TestSuite([
    unittest.TestLoader().loadTestsFromTestCase(TestPrefixAverage),
    unittest.TestLoader().loadTestsFromTestCase(TestUnique),
    unittest.TestLoader().loadTestsFromTestCase(TestDisjoint),
])
result = unittest.TextTestRunner(verbosity=2).run(all_suites)
print(f"\n{'OK' if result.wasSuccessful() else 'FAILED'} — {result.testsRun} tests")
````

## 📚 Síntesis: patrones aprendidos

| Patrón | Complejidad | Ejemplo en este notebook |
|--------|------------|--------------------------|
| Bucle simple hasta n | O(n) | `mi_prefix_avg2`, parte lineal de `unique2` |
| Dos bucles anidados independientes | O(n²) | `unique1` |
| i se duplica/divide | O(log n) | `fragmento_c` |
| Ordenar + recorrer | O(n log n) | `unique2` |
| Suma acumulada evita recomputar | O(n) en vez de O(n²) | `mi_prefix_avg2` vs `mi_prefix_avg1` |
| Guardia antes del bucle interno | Reduce factor oculto | `disjoint_v2` |
| Set para lookup | O(1) por consulta | `disjoint_v3` |

### El "truco de la suma acumulada"
```python
# Costoso: recomputa S[0..j] en cada iteración -> O(n²)
for j in range(n):
    total = sum(S[:j+1])

# Eficiente: suma corriente -> O(n)
acum = 0
for j in range(n):
    acum += S[j]
```
Este patrón aparece en: prefix_average, running totals, sliding window.

````python
# Visualización final: pendientes en escala log-log
ns = [100, 300, 700, 1500, 3000]
algs = [
    ('prefix_avg1 O(n²)',  lambda n: mi_prefix_avg1(list(range(n))),  5,   'red'),
    ('prefix_avg2 O(n)',   lambda n: mi_prefix_avg2(list(range(n))),  50,  'green'),
    ('unique1 O(n²)',      lambda n: unique1(list(range(n))),          5,   'orange'),
    ('unique2 O(n log n)', lambda n: unique2(list(range(n))),          50,  'blue'),
]
fig, ax = plt.subplots(figsize=(9, 5))
for label, func, reps, color in algs:
    tiempos = [timeit.timeit(lambda f=func, n=n: f(n), number=reps)/reps*1000 for n in ns]
    ax.loglog(ns, tiempos, '-o', label=label, color=color, linewidth=2)
ax.set_xlabel('n (escala log)'); ax.set_ylabel('Tiempo ms (escala log)')
ax.set_title('Confirmación empírica de complejidad (log-log)')
ax.legend(); plt.tight_layout(); plt.show()
print("En escala log-log: pendiente ~2 -> O(n²), pendiente ~1 -> O(n)")
````
