## 🟢 Ejercicio 1 — Clase `Rectangulo`

### Enunciado

Implementá la clase `Rectangulo` que represente un rectángulo definido por su ancho y alto. La clase debe cumplir los siguientes requisitos:

**Constructor:**
- `Rectangulo(ancho, alto)` — lanza `ValueError` si alguno es negativo o cero

**Propiedades (solo lectura):**
- `ancho`, `alto`, `area`, `perimetro`

**Métodos especiales:**
- `__repr__` → `Rectangulo(ancho=4, alto=3)`
- `__str__`  → `Rectángulo 4×3 (área=12, perímetro=14)`
- `__eq__`   → True si tienen la misma área
- `__lt__`   → True si `self.area < other.area` (usar `@functools.total_ordering`)

**Método extra:**
- `es_cuadrado()` → True si `ancho == alto`

### Esqueleto

```python
@functools.total_ordering
class Rectangulo:
    def __init__(self, ancho, alto):
        # TODO: validar y guardar
        ...
    @property
    def area(self): ...
    # ... completar
```

````python
@functools.total_ordering
class Rectangulo:
    """Rectángulo definido por ancho y alto."""

    def __init__(self, ancho: float, alto: float):
        if ancho <= 0 or alto <= 0:
            raise ValueError("ancho y alto deben ser positivos")
        self._ancho = float(ancho)
        self._alto  = float(alto)

    @property
    def ancho(self) -> float:    return self._ancho

    @property
    def alto(self) -> float:     return self._alto

    @property
    def area(self) -> float:     return self._ancho * self._alto

    @property
    def perimetro(self) -> float: return 2 * (self._ancho + self._alto)

    def es_cuadrado(self) -> bool:
        return self._ancho == self._alto

    def __repr__(self):
        return f"Rectangulo(ancho={self._ancho}, alto={self._alto})"

    def __str__(self):
        return (f"Rectángulo {self._ancho:.0f}×{self._alto:.0f} "
                f"(área={self.area:.2f}, perímetro={self.perimetro:.2f})")

    def __eq__(self, other):
        if not isinstance(other, Rectangulo): return NotImplemented
        return abs(self.area - other.area) < 1e-9

    def __lt__(self, other):
        if not isinstance(other, Rectangulo): return NotImplemented
        return self.area < other.area


# ── Demo ─────────────────────────────────────────────────────────────────────
r1 = Rectangulo(4, 3)
r2 = Rectangulo(6, 2)
r3 = Rectangulo(5, 5)

print(repr(r1))
print(str(r1))
print(f"r1.area = {r1.area}")
print(f"r1.perimetro = {r1.perimetro}")
print(f"r3.es_cuadrado() = {r3.es_cuadrado()}")
print(f"r1 == r2 = {r1 == r2}")   # 12 == 12 -> True
print(f"r1 < r3  = {r1 < r3}")    # 12 < 25  -> True
print(f"r3 > r1  = {r3 > r1}")    # generado por @total_ordering

try:
    Rectangulo(-1, 5)
except ValueError as e:
    print(f"Error esperado: {e}")
````

## 🟢 Ejercicio 2 — Clase `PilaOOP`

### Enunciado

Implementá una clase `PilaOOP` (pila LIFO) que **envuelva** internamente una lista Python.

**Excepción propia:**
- `PilaVacia(Exception)` — lanzada cuando se hace `pop()` o `peek()` en una pila vacía

**Constructor:**
- `PilaOOP()` — pila vacía

**Métodos:**
- `push(valor)` — agrega al tope
- `pop()` → retorna y elimina el tope (lanza `PilaVacia` si vacía)
- `peek()` → retorna el tope sin eliminar (lanza `PilaVacia` si vacía)

**Métodos especiales:**
- `__len__()` → número de elementos
- `__bool__()` → True si no está vacía
- `__repr__()` → `PilaOOP([1, 2, 3])` (tope en la derecha)
- `__iter__()` → itera del fondo al tope

### 💡 Pista

La pila usa una lista interna. El **tope** es el último elemento de la lista → `lista[-1]`.  
Todas las operaciones (`push`, `pop`, `peek`) son O(1) con este diseño.

````python
class PilaVacia(Exception):
    """Excepción lanzada al operar sobre una pila vacía."""
    pass


class PilaOOP:
    """Pila LIFO basada en list."""

    def __init__(self):
        self._datos = []

    def push(self, valor) -> None:
        self._datos.append(valor)

    def pop(self):
        if not self._datos:
            raise PilaVacia("No se puede hacer pop: la pila está vacía")
        return self._datos.pop()

    def peek(self):
        if not self._datos:
            raise PilaVacia("No se puede hacer peek: la pila está vacía")
        return self._datos[-1]

    def __len__(self) -> int:
        return len(self._datos)

    def __bool__(self) -> bool:
        return len(self._datos) > 0

    def __iter__(self):
        return iter(self._datos)

    def __repr__(self):
        return f"PilaOOP({self._datos!r})"


# ── Demo ─────────────────────────────────────────────────────────────────────
p = PilaOOP()
print(f"Pila vacía: {p}  | bool: {bool(p)}")

for v in [10, 20, 30, 40]:
    p.push(v)

print(f"Después de push(10,20,30,40): {p}")
print(f"peek: {p.peek()}")
print(f"pop:  {p.pop()}")
print(f"Después de pop: {p}")
print(f"len:  {len(p)}")
print(f"for v in pila: {list(p)}")

try:
    PilaOOP().pop()
except PilaVacia as e:
    print(f"\nError esperado: {e}")
````

### 🧪 Tests — Ejercicio 2

````python
class TestPilaOOP(unittest.TestCase):

    def test_push_pop(self):
        p = PilaOOP()
        p.push(1); p.push(2); p.push(3)
        self.assertEqual(p.pop(), 3)
        self.assertEqual(p.pop(), 2)

    def test_peek(self):
        p = PilaOOP()
        p.push(42)
        self.assertEqual(p.peek(), 42)
        self.assertEqual(len(p), 1)   # peek no elimina

    def test_len_vacia(self):
        self.assertEqual(len(PilaOOP()), 0)

    def test_bool_falsy_cuando_vacia(self):
        p = PilaOOP()
        self.assertFalse(p)
        p.push(1)
        self.assertTrue(p)

    def test_pop_vacia_lanza_error(self):
        with self.assertRaises(PilaVacia):
            PilaOOP().pop()

    def test_peek_vacia_lanza_error(self):
        with self.assertRaises(PilaVacia):
            PilaOOP().peek()

    def test_iter(self):
        p = PilaOOP()
        for v in [1, 2, 3]:
            p.push(v)
        self.assertEqual(list(p), [1, 2, 3])


unittest.TextTestRunner(verbosity=2).run(
    unittest.TestLoader().loadTestsFromTestCase(TestPilaOOP)
)
````

## 🟡 Ejercicio 3 — Clase `Polinomio`

### Enunciado

Implementá `Polinomio` para representar un polinomio de una variable real.

**Constructor:** `Polinomio(*coeficientes)` donde el primer coeficiente es el de mayor grado.  
Por ejemplo, `Polinomio(3, 0, -2, 1)` representa 3x³ + 0x² - 2x + 1.

**Métodos requeridos:**
- `grado` (property) → grado del polinomio
- `evaluar(x)` → valor numérico en x (usa la Regla de Horner para eficiencia)
- `__add__(other)` → suma de dos polinomios (retorna `Polinomio`)
- `__mul__(escalar: float)` → producto por escalar (retorna `Polinomio`)
- `__repr__` → `Polinomio(3, 0, -2, 1)`
- `__str__` → representación algebraica: `3x³ - 2x + 1`

### 💡 Pistas
- La **Regla de Horner**: `3x³ - 2x + 1 = ((3)*x + 0)*x + (-2))*x + 1`  
  → bucle: `resultado = resultado * x + coef` (más eficiente que calcular potencias)
- Para la suma, alinear los coeficientes por el lado derecho (grado menor)

````python
class Polinomio:
    """
    Polinomio de una variable real.
    Los coeficientes van de mayor a menor grado.
    Polinomio(3, -2, 1)  ->  3x² - 2x + 1
    """

    def __init__(self, *coeficientes):
        if not coeficientes:
            coeficientes = (0,)
        # Eliminar ceros líderes (salvo el caso del polinomio cero)
        coefs = list(coeficientes)
        while len(coefs) > 1 and coefs[0] == 0:
            coefs.pop(0)
        self._coefs = tuple(float(c) for c in coefs)

    @property
    def grado(self) -> int:
        """Grado del polinomio (len - 1)."""
        return len(self._coefs) - 1

    def evaluar(self, x: float) -> float:
        """Regla de Horner: O(n) en lugar de O(n²)."""
        resultado = 0.0
        for c in self._coefs:
            resultado = resultado * x + c
        return resultado

    def __add__(self, other: 'Polinomio') -> 'Polinomio':
        """Suma alineando coeficientes por el lado derecho (mayor índice = menor grado)."""
        # Alinear usando itertools o zip con padding
        a, b = list(self._coefs), list(other._coefs)
        diff = len(a) - len(b)
        if diff > 0:
            b = [0.0] * diff + b
        elif diff < 0:
            a = [0.0] * (-diff) + a
        return Polinomio(*(x + y for x, y in zip(a, b)))

    def __mul__(self, escalar: float) -> 'Polinomio':
        """Multiplicación por escalar."""
        return Polinomio(*(c * escalar for c in self._coefs))

    def __rmul__(self, escalar: float) -> 'Polinomio':
        return self.__mul__(escalar)

    def __eq__(self, other):
        if not isinstance(other, Polinomio): return NotImplemented
        return self._coefs == other._coefs

    def __repr__(self):
        return f"Polinomio({', '.join(str(c) for c in self._coefs)})"

    def __str__(self):
        """Representación algebraica legible: 3x³ - 2x + 1"""
        superscripts = str.maketrans('0123456789', '⁰¹²³⁴⁵⁶⁷⁸⁹')
        if not any(self._coefs):
            return "0"
        partes = []
        n = len(self._coefs)
        for i, c in enumerate(self._coefs):
            if c == 0:
                continue
            exp = n - 1 - i
            coef_str = '' if abs(c) == 1 and exp > 0 else str(int(c) if c == int(c) else c)
            signo = '+' if c > 0 else '-'
            coef_abs = str(int(abs(c)) if abs(c) == int(abs(c)) else abs(c))
            if abs(c) == 1 and exp > 0:
                coef_abs = ''
            if exp == 0:
                var = coef_abs if coef_abs else '1'
            elif exp == 1:
                var = f"{coef_abs}x"
            else:
                var = f"{coef_abs}x{str(exp).translate(superscripts)}"
            partes.append((signo, var))
        if not partes:
            return "0"
        resultado = partes[0][1] if partes[0][0] == '+' else f"-{partes[0][1]}"
        for signo, var in partes[1:]:
            resultado += f" {signo} {var}"
        return resultado


# ── Demo ─────────────────────────────────────────────────────────────────────
p1 = Polinomio(3, 0, -2, 1)   # 3x³ - 2x + 1
p2 = Polinomio(1, 2)           # x + 2

print(f"p1 = {p1}  ({repr(p1)})")
print(f"p2 = {p2}  ({repr(p2)})")
print(f"p1.grado = {p1.grado}")
print(f"p1.evaluar(2) = {p1.evaluar(2)}")   # 3(8) - 2(2) + 1 = 21
print(f"p1 + p2 = {p1 + p2}")
print(f"2 * p1 = {2 * p1}")

# Verificar Horner: 3x³ - 2x + 1 con x=2 -> 3(8) - 4 + 1 = 21
print(f"\nVerificación: 3(2)³ - 2(2) + 1 = {3*8 - 2*2 + 1}")
````

## 🟡 Ejercicio 4 — Clase `ColaPrioridad`

### Enunciado

Implementá `ColaPrioridad` que gestione elementos con prioridad numérica (menor número = mayor prioridad, como en una cola de emergencias: prioridad 1 es más urgente que 3).

**Constructor:** `ColaPrioridad()` — cola vacía

**Métodos:**
- `insertar(item, prioridad: int)` — agrega el elemento
- `extraer_min()` → `(item, prioridad)` — extrae el de menor prioridad (lanza `IndexError` si vacía)
- `ver_min()` → `(item, prioridad)` — inspecciona sin extraer
- `__len__()`, `__bool__()`, `__repr__()`

### 💡 Pista
Internamente, guardar una lista de tuplas `(prioridad, item)`. Para `extraer_min()`, buscá el mínimo con `min()` sobre la lista — no es la implementación más eficiente (eso se verá en Cap. 9 con heaps), pero es correcta y simple.

````python
class ColaPrioridad:
    """
    Cola de prioridad mínima simple.
    Internamente: lista de tuplas (prioridad, item).
    Complejidad: insertar O(1), extraer_min O(n).
    (Cap. 9 introduce heaps para O(log n).)
    """

    def __init__(self):
        self._datos = []   # lista de (prioridad, item)

    def insertar(self, item, prioridad: int) -> None:
        """Agrega item con la prioridad dada."""
        self._datos.append((prioridad, item))

    def extraer_min(self):
        """
        Extrae y retorna (item, prioridad) de menor prioridad.
        Lanza IndexError si la cola está vacía.
        """
        if not self._datos:
            raise IndexError("extraer_min: la cola está vacía")
        min_elem = min(self._datos)   # min por prioridad (primer elemento de la tupla)
        self._datos.remove(min_elem)
        prioridad, item = min_elem
        return item, prioridad

    def ver_min(self):
        """Inspecciona sin extraer. Lanza IndexError si vacía."""
        if not self._datos:
            raise IndexError("ver_min: la cola está vacía")
        prioridad, item = min(self._datos)
        return item, prioridad

    def __len__(self) -> int:
        return len(self._datos)

    def __bool__(self) -> bool:
        return len(self._datos) > 0

    def __repr__(self):
        ordenado = sorted(self._datos)
        items = ', '.join(f"({p}: {i!r})" for p, i in ordenado)
        return f"ColaPrioridad([{items}])"


# ── Demo ─────────────────────────────────────────────────────────────────────
cq = ColaPrioridad()
print(f"Vacía: {cq} | len={len(cq)} | bool={bool(cq)}")

# Simular una sala de emergencias
for paciente, prioridad in [("Carlos", 3), ("Ana", 1), ("María", 2), ("Luis", 1)]:
    cq.insertar(paciente, prioridad)
    print(f"  Insertado {paciente!r} (prio={prioridad})")

print(f"\nCola: {cq}")
print(f"Ver mín: {cq.ver_min()}")

print("\nAtendiendo en orden de prioridad:")
while cq:
    paciente, prio = cq.extraer_min()
    print(f"  [{prio}] Atendiendo a {paciente}")

try:
    cq.extraer_min()
except IndexError as e:
    print(f"\nError esperado: {e}")
````

## 🔴🔴 Desafío — Clase `Matriz2x2`

````python
# ── Esqueleto (intentá antes de ver la solución) ─────────────────────────────

class Matriz2x2:
    """
    Matriz 2x2 con operaciones algebraicas.
    
    Constructor: Matriz2x2(a, b, c, d)  representa  |a b|
                                                      |c d|
    Operaciones a implementar:
      - determinante() -> float
      - __add__(other)  -> Matriz2x2
      - __mul__(other)  -> Matriz2x2 si other es Matriz2x2;
                           Matriz2x2 si other es escalar (int/float)
      - __repr__() , __str__() , __eq__()
    """

    def __init__(self, a, b, c, d):
        self._m = [[float(a), float(b)],
                   [float(c), float(d)]]

    def __getitem__(self, idx):
        return self._m[idx]   # permite m[0][1], m[1][0], etc.

    def determinante(self) -> float:
        a, b, c, d = self._m[0][0], self._m[0][1], self._m[1][0], self._m[1][1]
        return a * d - b * c

    def __add__(self, other: 'Matriz2x2') -> 'Matriz2x2':
        if not isinstance(other, Matriz2x2): return NotImplemented
        a = self._m
        b = other._m
        return Matriz2x2(a[0][0]+b[0][0], a[0][1]+b[0][1],
                         a[1][0]+b[1][0], a[1][1]+b[1][1])

    def __mul__(self, other):
        if isinstance(other, (int, float)):     # escalar
            a = self._m
            return Matriz2x2(a[0][0]*other, a[0][1]*other,
                             a[1][0]*other, a[1][1]*other)
        if isinstance(other, Matriz2x2):        # producto matricial
            a, b = self._m, other._m
            return Matriz2x2(
                a[0][0]*b[0][0] + a[0][1]*b[1][0],
                a[0][0]*b[0][1] + a[0][1]*b[1][1],
                a[1][0]*b[0][0] + a[1][1]*b[1][0],
                a[1][0]*b[0][1] + a[1][1]*b[1][1]
            )
        return NotImplemented

    def __rmul__(self, escalar):
        return self.__mul__(escalar)

    def __eq__(self, other):
        if not isinstance(other, Matriz2x2): return NotImplemented
        return self._m == other._m

    def __repr__(self):
        a = self._m
        return f"Matriz2x2({a[0][0]}, {a[0][1]}, {a[1][0]}, {a[1][1]})"

    def __str__(self):
        a = self._m
        return (f"|{a[0][0]:6.2f}  {a[0][1]:6.2f}|\n"
                f"|{a[1][0]:6.2f}  {a[1][1]:6.2f}|")
````

### Demostración y verificación

````python
A = Matriz2x2(1, 2, 3, 4)
B = Matriz2x2(5, 6, 7, 8)
I = Matriz2x2(1, 0, 0, 1)   # identidad

print("=== Matrices ===")
print(f"A =\n{A}\n")
print(f"B =\n{B}\n")

print(f"det(A) = {A.determinante()}")   # 1*4 - 2*3 = -2
print(f"det(I) = {I.determinante()}")   # 1

print(f"\nA + B =\n{A + B}\n")
print(f"A * B =\n{A * B}\n")
print(f"2 * A =\n{2 * A}\n")
print(f"A * I =\n{A * I}\n")   # debe ser igual a A

print(f"A * I == A: {A * I == A}")

# Propiedad: A * I = A
print(f"\nrepr(A): {repr(A)}")

# Tests rápidos
class TestMatriz(unittest.TestCase):

    def test_determinante(self):
        self.assertAlmostEqual(Matriz2x2(1, 2, 3, 4).determinante(), -2.0)
        self.assertAlmostEqual(Matriz2x2(1, 0, 0, 1).determinante(), 1.0)

    def test_suma(self):
        A = Matriz2x2(1, 2, 3, 4)
        B = Matriz2x2(5, 6, 7, 8)
        self.assertEqual(A + B, Matriz2x2(6, 8, 10, 12))

    def test_mul_escalar(self):
        A = Matriz2x2(1, 2, 3, 4)
        self.assertEqual(A * 2, Matriz2x2(2, 4, 6, 8))

    def test_mul_matricial(self):
        A = Matriz2x2(1, 2, 3, 4)
        I = Matriz2x2(1, 0, 0, 1)
        self.assertEqual(A * I, A)

    def test_producto_especifico(self):
        A = Matriz2x2(1, 2, 3, 4)
        B = Matriz2x2(5, 6, 7, 8)
        # A*B = |1*5+2*7  1*6+2*8| = |19  22|
        #       |3*5+4*7  3*6+4*8|   |43  50|
        self.assertEqual(A * B, Matriz2x2(19, 22, 43, 50))


unittest.TextTestRunner(verbosity=2).run(
    unittest.TestLoader().loadTestsFromTestCase(TestMatriz)
)
````

## 📋 Síntesis de Conceptos

### Patrones aplicados en esta clase

| Ejercicio | Conceptos clave | Métodos especiales |
|-----------|----------------|-------------------|
| `Rectangulo` | `@property`, `@total_ordering`, validación en constructor | `__repr__`, `__str__`, `__eq__`, `__lt__` |
| `PilaOOP` | Wrapper de colección, excepción propia | `__len__`, `__bool__`, `__iter__`, `__repr__` |
| `Polinomio` | Algoritmo de Horner, suma de secuencias | `__add__`, `__mul__`, `__rmul__`, `__repr__`, `__str__` |
| `ColaPrioridad` | Diseño ADT, `min()` sobre lista de tuplas | `__len__`, `__bool__`, `__repr__` |
| `Matriz2x2` | Álgebra lineal, despacho por tipo en `__mul__` | `__add__`, `__mul__`, `__rmul__`, `__eq__`, `__str__`, `__repr__` |

### Lecciones de diseño

1. **Validar en `__init__`**: el contrato de la clase se establece en la construcción
2. **`@total_ordering`**: define sólo `__eq__` + uno de los comparadores, obtenés el resto gratis
3. **Excepción propia**: más descriptiva que usar `ValueError` genérico
4. **Retornar `NotImplemented`** (no raise) en operadores binarios para que Python pruebe el otro operando
5. **Wrapper vs. herencia**: `PilaOOP` envuelve una lista en vez de heredar de ella → mayor control de la interfaz
