```python
# ============================================================
# Ejercicio 3: Skip List — Visualización y Búsqueda
# ============================================================
# Implementa una Skip List simplificada con trazado visual

class SkipNode:
    """Nodo de una Skip List."""
    def __init__(self, key, value, height):
        self.key = key
        self.value = value
        self.next = [None] * height   # next[i] = siguiente nodo en nivel i

class SkipList:
    """Skip List simplificada para ilustrar el concepto.
    
    - Máximo MAX_LEVEL niveles
    - Probabilidad p=0.5 de subir al siguiente nivel
    - Búsqueda O(log n) esperado
    """
    MAX_LEVEL = 4
    P = 0.5
    
    def __init__(self):
        # Sentinel head con clave -infinito
        self._head = SkipNode(float('-inf'), None, self.MAX_LEVEL)
        self._level = 0   # nivel máximo actual
    
    def _random_level(self):
        """Genera nivel aleatorio: 0 con prob 0.5, 1 con prob 0.25, etc."""
        lvl = 0
        while random.random() < self.P and lvl < self.MAX_LEVEL - 1:
            lvl += 1
        return lvl
    
    def search(self, k, trace=False):
        """Busca clave k. Retorna valor o None. Opcionalmente traza la búsqueda."""
        current = self._head
        path = []
        
        for i in range(self._level, -1, -1):
            while current.next[i] and current.next[i].key < k:
                current = current.next[i]
                if trace: path.append((i, current.key))
        
        current = current.next[0]
        if current and current.key == k:
            if trace:
                print(f"  Búsqueda de {k}: camino={path} → ENCONTRADO({current.value})")
            return current.value
        if trace:
            print(f"  Búsqueda de {k}: camino={path} → NO ENCONTRADO")
        return None
    
    def insert(self, k, v):
        """Inserta (k, v) en la skip list."""
        update = [None] * self.MAX_LEVEL
        current = self._head
        
        for i in range(self._level, -1, -1):
            while current.next[i] and current.next[i].key < k:
                current = current.next[i]
            update[i] = current
        
        current = current.next[0]
        if current and current.key == k:
            current.value = v   # actualizar
        else:
            new_level = self._random_level()
            if new_level > self._level:
                for i in range(self._level + 1, new_level + 1):
                    update[i] = self._head
                self._level = new_level
            
            new_node = SkipNode(k, v, new_level + 1)
            for i in range(new_level + 1):
                new_node.next[i] = update[i].next[i]
                update[i].next[i] = new_node
    
    def to_list(self):
        """Retorna lista de (key, value) en orden."""
        result = []
        current = self._head.next[0]
        while current:
            result.append((current.key, current.value))
            current = current.next[0]
        return result
    
    def show(self):
        """Muestra la skip list nivel por nivel."""
        nodes = self.to_list()
        keys = [k for k, _ in nodes]
        print(f"SkipList ({len(nodes)} elementos, nivel máx={self._level}):")
        for lvl in range(self._level, -1, -1):
            line = f"  L{lvl}: HEAD"
            current = self._head.next[lvl]
            while current:
                line += f" → {current.key}"
                current = current.next[lvl]
            print(line)


# ---- Demo ----
print("=== Skip List Demo ===")
random.seed(42)
sl = SkipList()
valores = [3, 6, 7, 9, 12, 17, 19, 21, 25, 26]
for v in valores:
    sl.insert(v, f"val_{v}")

sl.show()
print()
print(f"Elementos en orden: {[k for k,_ in sl.to_list()]}")
print()
print("Búsquedas con traza:")
sl.search(17, trace=True)
sl.search(12, trace=True)
sl.search(8, trace=True)   # no existe
```

```python
# ============================================================
# Ejercicio 5: Sistema de Inventario con Rangos de Precio
# ============================================================
# Implementa un sistema de inventario que permita búsquedas eficientes
# por rango de precio usando SortedTableMap

# Copiar SortedTableMap completo para este ejercicio
class SortedTableMap(MapBase):
    def __init__(self): self._table = []
    def __len__(self): return len(self._table)
    def __iter__(self):
        for item in self._table: yield item._key
    def _find_index(self, k, low, high):
        if high < low: return high + 1
        mid = (low + high) // 2
        if k == self._table[mid]._key: return mid
        elif k < self._table[mid]._key: return self._find_index(k, low, mid-1)
        else: return self._find_index(k, mid+1, high)
    def __getitem__(self, k):
        j = self._find_index(k, 0, len(self._table)-1)
        if j < len(self._table) and self._table[j]._key == k: return self._table[j]._value
        raise KeyError(repr(k))
    def __setitem__(self, k, v):
        j = self._find_index(k, 0, len(self._table)-1)
        if j < len(self._table) and self._table[j]._key == k: self._table[j]._value = v
        else: self._table.insert(j, self._Item(k, v))
    def __delitem__(self, k):
        j = self._find_index(k, 0, len(self._table)-1)
        if j == len(self._table) or self._table[j]._key != k: raise KeyError(repr(k))
        self._table.pop(j)
    def find_range(self, start, stop):
        j = 0 if start is None else self._find_index(start, 0, len(self._table)-1)
        while j < len(self._table):
            if stop is not None and self._table[j]._key >= stop: break
            yield (self._table[j]._key, self._table[j]._value)
            j += 1
    def find_min(self):
        if not self._table: return None
        return (self._table[0]._key, self._table[0]._value)
    def find_max(self):
        if not self._table: return None
        return (self._table[-1]._key, self._table[-1]._value)


class Inventario:
    """Sistema de inventario con búsqueda eficiente por rango de precio.
    
    Usa SortedTableMap con clave = precio para soportar find_range.
    OJO: varios productos pueden tener el mismo precio → usar lista de productos.
    
    Implementa:
    - agregar_producto(nombre, precio, stock)
    - productos_en_rango(precio_min, precio_max) → lista de productos
    - producto_mas_barato() → (nombre, precio)
    - producto_mas_caro() → (nombre, precio)
    - buscar_por_nombre(nombre) → dict o None
    """
    
    def __init__(self):
        # TU CÓDIGO AQUÍ
        # Pista: self._por_precio = SortedTableMap() donde valores son listas
        #        self._por_nombre = dict() para búsquedas por nombre
        pass
    
    def agregar_producto(self, nombre, precio, stock):
        # TU CÓDIGO AQUÍ
        pass
    
    def productos_en_rango(self, precio_min, precio_max):
        """Retorna lista de dicts con productos cuyo precio está en [min, max]."""
        # TU CÓDIGO AQUÍ
        pass
    
    def producto_mas_barato(self):
        # TU CÓDIGO AQUÍ
        pass
    
    def producto_mas_caro(self):
        # TU CÓDIGO AQUÍ
        pass
    
    def buscar_por_nombre(self, nombre):
        # TU CÓDIGO AQUÍ
        pass


# ---- Tests ----
print("Testeando Inventario con SortedTableMap...")
try:
    inv = Inventario()
    productos = [
        ("Notebook HP", 850_000, 5),
        ("Mouse Logitech", 12_500, 30),
        ("Teclado Mecánico", 45_000, 15),
        ("Monitor 24"", 320_000, 8),
        ("Auriculares Sony", 89_000, 12),
        ("Webcam HD", 28_000, 20),
        ("Hub USB", 8_000, 50),
        ("SSD 1TB", 95_000, 10),
        ("Impresora Epson", 185_000, 6),
        ("Tablet Samsung", 420_000, 4),
    ]
    for nombre, precio, stock in productos:
        inv.agregar_producto(nombre, precio, stock)
    
    # Test búsqueda por rango
    rango = inv.productos_en_rango(20_000, 100_000)
    nombres_rango = {p['nombre'] for p in rango}
    esperados = {'Teclado Mecánico', 'Webcam HD', 'Auriculares Sony', 'SSD 1TB'}
    assert nombres_rango == esperados, f"rango falló: {nombres_rango}"
    
    # Test extremos
    barato = inv.producto_mas_barato()
    caro = inv.producto_mas_caro()
    assert barato[0] == 'Hub USB' or barato[1] == 8_000, f"más barato falló: {barato}"
    assert caro[0] == 'Notebook HP' or caro[1] == 850_000, f"más caro falló: {caro}"
    
    # Test búsqueda por nombre
    p = inv.buscar_por_nombre("Monitor 24\"")
    assert p is not None and p['precio'] == 320_000, "buscar_por_nombre falló"
    
    print("  ✅ Todos los tests pasaron")
    print(f"\n  Productos entre $20.000 y $100.000:")
    for p in sorted(rango, key=lambda x: x['precio']):
        print(f"    {p['nombre']:25s} ${p['precio']:>10,}  stock: {p['stock']}")
    print(f"\n  Más barato: {inv.producto_mas_barato()}")
    print(f"  Más caro:   {inv.producto_mas_caro()}")
except Exception as e:
    print(f"  ❌ Error: {e}")
```
