## 🌟 **¿Qué son las Secuencias?**

### 🏠 **Analogía del Mundo Real**

Imagina que tienes diferentes tipos de **contenedores** en tu casa:

- 📚 **Estantería** (list): Puedes agregar, quitar y reorganizar libros
- 🎬 **DVD collection** (tuple): Una vez organizada, no cambias el orden 
- 📝 **Cartel** (str): Las letras están fijas, no puedes modificarlas

Todos son **contenedores ordenados**, pero con diferentes reglas de modificación.

### 🔬 **Definición Técnica**

Una **secuencia** en Python es una colección **ordenada** de elementos donde:

- ✅ Los elementos tienen un **orden específico**
- ✅ Se puede acceder por **índice** (0, 1, 2, ...)
- ✅ Soporta **slicing** (cortes)
- ✅ Tiene **longitud** definida
- ✅ Es **iterable**

**Los 3 tipos principales en Python:**
1. 📊 **`list`** - Mutable, heterogénea
2. 📦 **`tuple`** - Inmutable, heterogénea  
3. 📝 **`str`** - Inmutable, homogénea (caracteres)

````python
# 🧪 Ejemplo: Características comunes de las secuencias

# Crear una de cada tipo
mi_lista = [1, 2, 3, 'Python', True]
mi_tupla = (1, 2, 3, 'Python', True)
mi_string = "Hola Python"

print("📊 Longitudes:")
print(f"Lista: {len(mi_lista)}")
print(f"Tupla: {len(mi_tupla)}")
print(f"String: {len(mi_string)}")

print("\n🎯 Acceso por índice:")
print(f"Lista[0]: {mi_lista[0]}")
print(f"Tupla[0]: {mi_tupla[0]}")
print(f"String[0]: {mi_string[0]}")

print("\n✂️ Slicing:")
print(f"Lista[1:3]: {mi_lista[1:3]}")
print(f"Tupla[1:3]: {mi_tupla[1:3]}")
print(f"String[1:5]: {mi_string[1:5]}")
````

## 📊 **Lists - Las Secuencias Mutables**

### 🔄 **Características Principales**

- ✅ **Mutables**: Se pueden modificar después de crear
- ✅ **Heterogéneas**: Pueden contener diferentes tipos de datos
- ✅ **Dinámicas**: Crecen y decrecen automáticamente
- ✅ **Ordenadas**: Mantienen el orden de inserción

### 🏗️ **Operaciones Fundamentales**

````python
# 🧪 Creación y operaciones básicas en listas

# Diferentes formas de crear listas
lista_vacia = []
lista_numeros = [1, 2, 3, 4, 5]
lista_mixta = [1, "dos", 3.0, True, None]
lista_repetida = [0] * 5  # [0, 0, 0, 0, 0]

print("📋 Listas creadas:")
print(f"Vacía: {lista_vacia}")
print(f"Números: {lista_numeros}")
print(f"Mixta: {lista_mixta}")
print(f"Repetida: {lista_repetida}")
````

````python
# 🔧 Operaciones de modificación

estudiantes = ["Ana", "Luis", "María"]
print(f"📚 Lista inicial: {estudiantes}")

# Agregar elementos
estudiantes.append("Carlos")  # Al final
print(f"➕ Después de append: {estudiantes}")

estudiantes.insert(1, "José")  # En posición específica
print(f"📌 Después de insert: {estudiantes}")

# Eliminar elementos
estudiantes.remove("Luis")  # Por valor
print(f"➖ Después de remove: {estudiantes}")

ultimo = estudiantes.pop()  # Quitar último
print(f"🎯 Eliminado: {ultimo}, Lista: {estudiantes}")

# Modificar elementos
estudiantes[0] = "Andrea"
print(f"✏️ Después de modificar: {estudiantes}")
````

## 📦 Tuplas - Las Secuencias Inmutables

### 🔒 Características Principales

- ✅ **Inmutables**: No se pueden modificar después de crear
- ✅ **Heterogéneas**: Diferentes tipos de datos
- ✅ **Eficientes**: Menos memoria que las listas
- ✅ **Hashables**: Se pueden usar como claves de diccionario

### 🎯 ¿Cuándo usar tuples?

- 📍 **Coordenadas**: `punto = (10, 20)`
- 📅 **Fechas**: `fecha = (2025, 9, 13)`
- 🎨 **Colores RGB**: `color = (255, 128, 0)`
- 📊 **Datos que no cambian**: `configuracion = ("localhost", 8080, True)`

````python
# 🧪 Creación y uso de tuplas

# Diferentes formas de crear tuplas
tupla_vacia = ()
tupla_numeros = (1, 2, 3, 4, 5)
tupla_sin_parentesis = 1, 2, 3  # También válido
tupla_un_elemento = (42,)  # ¡Nota la coma!

print("📦 Tuplas creadas:")
print(f"Vacía: {tupla_vacia}")
print(f"Números: {tupla_numeros}")
print(f"Sin paréntesis: {tupla_sin_parentesis}")
print(f"Un elemento: {tupla_un_elemento}")

# Verificar tipos
print(f"\n🔍 Tipos:")
print(f"type(tupla_numeros): {type(tupla_numeros)}")
print(f"type(tupla_sin_parentesis): {type(tupla_sin_parentesis)}")
````

````python
# 🎯 Aplicaciones prácticas de tuplas

# 1. Intercambio de variables
a, b = 10, 20
print(f"Antes: a={a}, b={b}")
a, b = b, a  # ¡Intercambio con tuplas!
print(f"Después: a={a}, b={b}")

# 2. Retorno múltiple de funciones
def obtener_nombre_edad():
    return "Ana", 22  # Retorna una tupla

nombre, edad = obtener_nombre_edad()
print(f"\n👤 Datos: {nombre}, {edad} años")

# 3. Coordenadas y puntos
punto_a = (0, 0)
punto_b = (3, 4)

# Calcular distancia
import math
distancia = math.sqrt((punto_b[0] - punto_a[0])**2 + (punto_b[1] - punto_a[1])**2)
print(f"\n📐 Distancia entre {punto_a} y {punto_b}: {distancia:.2f}")
````

### ⚠️ Inmutabilidad de Tuplas

¿Qué significa que las tuplas son **inmutables**?

````python
# 🧪 Probando la inmutabilidad

mi_tupla = (1, 2, 3, "Python")
print(f"Tupla original: {mi_tupla}")

# Esto funcionará (acceso de lectura)
print(f"Elemento 0: {mi_tupla[0]}")
print(f"Slice [1:3]: {mi_tupla[1:3]}")

# ❌ Esto dará error (descomenta para probar)
# mi_tupla[0] = 999  # TypeError!
# mi_tupla.append("nuevo")  # AttributeError!

print("\n✅ La tupla sigue intacta después de intentos de modificación")
````

## 📝 Strings - Secuencias de Caracteres

### 🔤 Características Principales

- ✅ **Inmutables**: Como las tuplas
- ✅ **Homogéneas**: Solo caracteres
- ✅ **Unicode**: Soporte completo para caracteres internacionales
- ✅ **Métodos ricos**: Muchas operaciones de texto

### 🎨 Strings como Secuencias

````python
# 🧪 Strings como secuencias

mensaje = "Python es genial! 🐍"
print(f"📝 Mensaje: {mensaje}")
print(f"📏 Longitud: {len(mensaje)}")

# Acceso por índice
print(f"\n🎯 Primer carácter: '{mensaje[0]}'")
print(f"🎯 Último carácter: '{mensaje[-1]}'")

# Slicing
print(f"\n✂️ Primeras 6 letras: '{mensaje[:6]}'")
print(f"✂️ Últimas 8 caracteres: '{mensaje[-8:]}'")
print(f"✂️ Cada segundo carácter: '{mensaje[::2]}'")

# Iteración
print("\n🔄 Caracteres únicos:")
caracteres_unicos = set(mensaje.lower())
print(f"Total únicos: {len(caracteres_unicos)}")
````

````python
# 🔧 Métodos útiles de strings

texto = "  Estructuras de Datos con Python  "
print(f"📝 Original: '{texto}'")

# Limpieza y formato
print(f"🧹 strip(): '{texto.strip()}'")
print(f"📈 upper(): '{texto.upper()}'")
print(f"📉 lower(): '{texto.lower()}'")
print(f"🎯 title(): '{texto.title()}'")

# Búsqueda y reemplazo
print(f"\n🔍 'Python' en texto: {texto.find('Python')}")
print(f"🔄 replace(): '{texto.replace('Python', '🐍')}'")

# División y unión
palabras = texto.strip().split()
print(f"\n📝 Palabras: {palabras}")
print(f"🔗 join(): '{'-'.join(palabras)}'")
````

## 📊 Comparación de Secuencias

### 📋 Tabla Comparativa

| Característica | 📊 **List** | 📦 **Tuple** | 📝 **String** |
|----------------|-------------|--------------|---------------|
| **Mutabilidad** | ✅ Mutable | ❌ Inmutable | ❌ Inmutable |
| **Tipos de datos** | 🌈 Heterogénea | 🌈 Heterogénea | 📝 Homogénea (char) |
| **Sintaxis** | `[1, 2, 3]` | `(1, 2, 3)` | `"123"` |
| **Uso de memoria** | 🐌 Más | ⚡ Menos | ⚡ Menos |
| **Velocidad** | 🐌 Lenta | ⚡ Rápida | ⚡ Rápida |
| **Hashable** | ❌ No | ✅ Sí | ✅ Sí |
| **Métodos** | 🔧 Muchos | 📦 Pocos | 📝 Muchos |

````python
# 🧪 Comparando rendimiento

import timeit
import sys

# Crear datos de prueba
datos = [1, 2, 3, 4, 5] * 1000

# Convertir a diferentes tipos
mi_lista = list(datos)
mi_tupla = tuple(datos)
mi_string = ''.join(map(str, datos[:10]))  # Solo algunos para string

print("📊 Uso de memoria (bytes):")
print(f"Lista:  {sys.getsizeof(mi_lista):,}")
print(f"Tupla:  {sys.getsizeof(mi_tupla):,}")
print(f"String: {sys.getsizeof(mi_string):,}")

print("\n⚡ Velocidad de acceso (por elemento):")
print("(Menores números = más rápido)")

# Medir acceso por índice
tiempo_lista = timeit.timeit(lambda: mi_lista[100], number=100000)
tiempo_tupla = timeit.timeit(lambda: mi_tupla[100], number=100000)

print(f"Lista:  {tiempo_lista:.6f} segundos")
print(f"Tupla:  {tiempo_tupla:.6f} segundos")
print(f"Tupla es {tiempo_lista/tiempo_tupla:.1f}x más rápida")
````

## 🎯 Cuándo Usar Cada Tipo

### 📊 Lists - Usa cuando...

- 🔄 Necesites **modificar** el contenido
- 📈 El tamaño **cambie dinámicamente**
- 🔧 Requieras **métodos** como append, remove, sort
- 📝 Implementes **algoritmos** que modifiquen datos

**Ejemplos:** carritos de compra, listas de tareas, datos temporales

### 📦 Tuples - Usa cuando...

- 🔒 Los datos sean **inmutables** por naturaleza
- ⚡ Necesites **mejor rendimiento**
- 🗝️ Requieras **keys** de diccionario
- 📊 Retornes **múltiples valores** de funciones

**Ejemplos:** coordenadas, configuraciones, fechas, registros

### 📝 Strings - Usa cuando...

- 📖 Manejes **texto** y caracteres
- 🔍 Necesites **operaciones de texto** avanzadas
- 🌐 Trabajes con **entrada/salida** de usuarios
- 📄 Proceses **archivos de texto**

**Ejemplos:** mensajes, nombres, URLs, contenido de archivos

## 🎯 Resumen y Conceptos Clave

### ✅ Lo que aprendimos

1. **🐍 Python tiene 3 secuencias principales:**
   
   - `list`: mutable, heterogénea, dinámica
   - `tuple`: inmutable, heterogénea, eficiente  
   - `str`: inmutable, homogénea, rica en métodos

2. **📊 Todas las secuencias soportan:**
   
   - Indexing: `seq[i]`
   - Slicing: `seq[start:end:step]`
   - Length: `len(seq)`
   - Iteration: `for item in seq`

3. **🎯 Criterios de selección:**
   
   - **¿Necesitas modificar?** → List
   - **¿Datos fijos y rendimiento?** → Tuple  
   - **¿Trabajo con texto?** → String
