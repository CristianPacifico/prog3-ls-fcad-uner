```python
# 🧪 Experimento: Comparar uso de memoria

def experimento_memoria():
    """Compara uso de memoria entre diferentes estructuras"""
    
    # Datos de prueba: números del 1 al 1000
    n = 1000
    datos = list(range(n))
    
    # Crear diferentes estructuras
    estructuras = {
        'list': datos,
        'tuple': tuple(datos),
        'array.array': array.array('i', datos),  # signed int
        'string': ''.join(str(x) for x in range(10))  # Ejemplo más pequeño
    }
    
    print(f"📊 Comparación de memoria para {n:,} enteros:")
    print("Estructura    | Memoria (bytes) | Por elemento | Overhead")
    print("-" * 60)
    
    # Tu código aquí:
    # 1. Para cada estructura, mide la memoria total
    # 2. Calcula memoria por elemento
    # 3. Calcula overhead comparado con array.array (más eficiente)
    
    for nombre, estructura in estructuras.items():
        if nombre == 'string':
            continue  # Skip string para esta comparación numérica
        
        memoria = medir_memoria_total(estructura)
        por_elemento = memoria / len(estructura)
        
        print(f"{nombre:12} | {memoria:12,} | {por_elemento:10.1f} | ")
    
    # Experimento adicional: Strings
    print("\n📝 Memoria de strings:")
    strings = {
        'str corto': "Hola",
        'str largo': "Hola" * 250,
        'list chars': list("Hola" * 250)
    }
    
    for nombre, obj in strings.items():
        memoria = sys.getsizeof(obj)
        por_char = memoria / len(obj) if len(obj) > 0 else 0
        print(f"{nombre:12}: {memoria:5} bytes ({por_char:.1f} por char)")

experimento_memoria()

print("\n🤔 Preguntas de análisis:")
print("1. ¿Por qué list usa más memoria que array.array?")
print("2. ¿Cuándo preferirías tuple sobre list?")
print("3. ¿Qué estructura usarías para 1M de números float?")
```
