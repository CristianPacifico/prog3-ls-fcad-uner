### 🔬 **Conceptos Fundamentales**

**Big O Notation** describe cómo crece el tiempo de ejecución con el tamaño de entrada:

- ✅ **O(1)**: Tiempo constante - ¡Excelente!
- ✅ **O(log n)**: Tiempo logarítmico - Muy bueno
- ⚠️ **O(n)**: Tiempo lineal - Aceptable
- ⚠️ **O(n log n)**: Tiempo loglineal - Razonable para algoritmos complejos
- ❌ **O(n²)**: Tiempo cuadrático - Problemático para n grande

**Tipos de análisis:**
- **Worst-case**: El peor escenario posible
- **Average-case**: El caso promedio esperado
- **Amortized**: Costo promedio a lo largo de muchas operaciones

```python
# 🧪 Visualización de complejidades

import math

def calcular_complejidades(n):
    """Calcula diferentes complejidades para visualización"""
    return {
        'O(1)': 1,
        'O(log n)': math.log2(n) if n > 0 else 0,
        'O(n)': n,
        'O(n log n)': n * math.log2(n) if n > 0 else 0,
        'O(n²)': n * n
    }

# Comparar para diferentes tamaños
tamaños = [1, 10, 100, 1000, 10000]

print("📊 Comparación de complejidades:")
print("n      | O(1) | O(log n) | O(n)    | O(n log n) | O(n²)")
print("-" * 65)

for n in tamaños:
    comp = calcular_complejidades(n)
    print(f"{n:6} | {comp['O(1)']:4.0f} | {comp['O(log n)']:8.1f} | "
          f"{comp['O(n)']:7.0f} | {comp['O(n log n)']:10.0f} | "
          f"{comp['O(n²)']:8.0f}")

print("\n💡 Observa cómo crece cada complejidad con n")
print("🚀 O(1) y O(log n) escalan muy bien")
print("⚠️ O(n²) se vuelve prohibitivo rápidamente")
```

## 📊 **Tabla de Complejidades - Secuencias**

### 🎯 **Operaciones Fundamentales**

| Operación | 📊 **List** | 📦 **Tuple** | 📝 **String** | 🧱 **Array** |
|-----------|-------------|--------------|---------------|--------------|
| **len(s)** | O(1) | O(1) | O(1) | O(1) |
| **s[i]** (acceso) | O(1) | O(1) | O(1) | O(1) |
| **s[i] = x** (asignación) | O(1) | ❌ | ❌ | O(1) |
| **s.append(x)** | O(1)* | ❌ | ❌ | O(1)* |
| **s.insert(i, x)** | O(n) | ❌ | ❌ | O(n) |
| **s.pop()** | O(1)* | ❌ | ❌ | O(1)* |
| **s.pop(i)** | O(n) | ❌ | ❌ | O(n) |
| **del s[i]** | O(n) | ❌ | ❌ | O(n) |
| **x in s** | O(n) | O(n) | O(n) | O(n) |
| **s.count(x)** | O(n) | O(n) | O(n) | O(n) |
| **s.index(x)** | O(n) | O(n) | O(n) | O(n) |
| **s + t** (concatenación) | O(k) | O(k) | O(k) | O(k) |
| **s * k** (repetición) | O(kn) | O(kn) | O(kn) | O(kn) |
| **s[i:j]** (slicing) | O(k) | O(k) | O(k) | O(k) |

*donde k = tamaño del resultado, n = tamaño de la secuencia*  
**\* = Amortizado**

```python
# 🧪 Verificación empírica de complejidades

import timeit
import random

def medir_operacion(setup_code, test_code, sizes, repetitions=1000):
    """Mide tiempo de ejecución para diferentes tamaños"""
    resultados = []
    
    for n in sizes:
        # Preparar contexto
        setup = setup_code.format(n=n)
        
        # Medir tiempo
        tiempo = timeit.timeit(
            stmt=test_code,
            setup=setup,
            number=repetitions
        )
        
        resultados.append((n, tiempo))
    
    return resultados

# Test 1: Acceso por índice (debería ser O(1))
print("🎯 Test 1: Acceso por índice s[i]")
print("Tamaño | Tiempo (ms) | Factor")
print("-" * 35)

sizes = [1000, 5000, 10000, 50000]
resultados_acceso = medir_operacion(
    setup_code="lista = list(range({n})); import random",
    test_code="lista[random.randint(0, len(lista)-1)]",
    sizes=sizes,
    repetitions=10000
)

tiempo_base = resultados_acceso[0][1]
for n, tiempo in resultados_acceso:
    factor = tiempo / tiempo_base
    print(f"{n:6} | {tiempo*1000:10.3f} | {factor:5.2f}x")

print("\n💡 Si es O(1), el factor debería mantenerse ~constante")
```

```python
# Test 2: Búsqueda lineal (debería ser O(n))
print("🔍 Test 2: Búsqueda 'x in s'")
print("Tamaño | Tiempo (ms) | Factor")
print("-" * 35)

resultados_busqueda = medir_operacion(
    setup_code="lista = list(range({n}))",
    test_code="({n}//2) in lista",  # Buscar elemento en el medio
    sizes=sizes,
    repetitions=1000
)

tiempo_base = resultados_busqueda[0][1]
tamaño_base = resultados_busqueda[0][0]

for n, tiempo in resultados_busqueda:
    factor_tiempo = tiempo / tiempo_base
    factor_tamaño = n / tamaño_base
    print(f"{n:6} | {tiempo*1000:10.3f} | {factor_tiempo:5.2f}x")

print("\n💡 Si es O(n), el factor debería crecer linealmente con el tamaño")
print(f"Factor de tamaño vs factor de tiempo debería ser similar")
```

```python
# Test 3: Inserción al inicio (debería ser O(n))
print("📌 Test 3: Inserción al inicio s.insert(0, x)")
print("Tamaño | Tiempo (ms) | Factor")
print("-" * 35)

# Tamaños más pequeños para inserción (es costosa)
sizes_insert = [1000, 2000, 4000, 8000]

resultados_insert = medir_operacion(
    setup_code="lista = list(range({n}))",
    test_code="lista.insert(0, -1); lista.pop(0)",  # Insert y remove para no crecer
    sizes=sizes_insert,
    repetitions=100
)

tiempo_base = resultados_insert[0][1]
for n, tiempo in resultados_insert:
    factor = tiempo / tiempo_base
    print(f"{n:6} | {tiempo*1000:10.3f} | {factor:5.2f}x")

print("\n💡 Si es O(n), el factor debería duplicarse cuando el tamaño se duplica")
```

## 🔄 Operaciones Específicas de Python Lists

### 📊 Complejidades Detalladas

**Operaciones O(1) - Rápidas:**

- `len(lista)` - Mantiene contador interno
- `lista[i]` - Acceso directo por índice
- `lista[i] = x` - Asignación directa
- `lista.append(x)` - Agregar al final (amortizado)
- `lista.pop()` - Eliminar último (amortizado)

**Operaciones O(n) - Lentas:**

- `lista.insert(i, x)` - Debe desplazar elementos
- `lista.pop(i)` - Debe desplazar elementos
- `x in lista` - Búsqueda lineal
- `lista.remove(x)` - Buscar y desplazar
- `lista.sort()` - O(n log n) en realidad

```python
# 🧪 Comparando operaciones rápidas vs lentas

import timeit

def benchmark_operaciones(n=10000):
    """Compara tiempos de diferentes operaciones"""
    
    # Preparar datos
    lista = list(range(n))
    
    operaciones = {
        # Operaciones O(1)
        'len(lista)': lambda: len(lista),
        'lista[500]': lambda: lista[500],
        'lista.append(999)': lambda: [lista.append(999), lista.pop()][1],  # Append y remove
        
        # Operaciones O(n)
        'lista.insert(0, 999)': lambda: [lista.insert(0, 999), lista.pop(0)][1],
        '999 in lista': lambda: 999 in lista,
        'lista.count(500)': lambda: lista.count(500),
    }
    
    print(f"⚡ Benchmark de operaciones (n={n:,}):")
    print("Operación" + " " * 20 + "| Tiempo (μs) | Complejidad")
    print("-" * 55)
    
    resultados = []
    
    for nombre, operacion in operaciones.items():
        tiempo = timeit.timeit(operacion, number=1000)
        tiempo_us = tiempo * 1000  # Convertir a microsegundos
        resultados.append((nombre, tiempo_us))
    
    # Ordenar por tiempo
    resultados.sort(key=lambda x: x[1])
    
    for nombre, tiempo in resultados:
        complejidad = "O(1)" if tiempo < 10 else "O(n)"
        print(f"{nombre:30} | {tiempo:10.2f} | {complejidad}")
    
    return resultados

resultados = benchmark_operaciones()

print("\n🎯 Observaciones:")
print("- Las operaciones O(1) son consistentemente rápidas")
print("- Las operaciones O(n) toman más tiempo")
print("- La diferencia es más notable con listas grandes")
```

## 🎭 Casos de Uso y Optimizaciones

### 🎯 Escenario 1: Processing Pipeline

¿Cómo optimizar un pipeline que procesa elementos secuencialmente?

```python
# 🧪 Comparando estrategias de procesamiento

import time

def estrategia_naive(datos):
    """Estrategia ineficiente: insertar al inicio"""
    cola = []
    
    # Simular llegada de datos
    for dato in datos:
        cola.insert(0, dato)  # O(n) - ¡Malo!
    
    # Procesar
    resultado = []
    while cola:
        item = cola.pop()  # O(1) - Bien
        resultado.append(item * 2)
    
    return resultado

def estrategia_optimizada(datos):
    """Estrategia eficiente: append + reverse"""
    cola = []
    
    # Simular llegada de datos
    for dato in datos:
        cola.append(dato)  # O(1) amortizado - ¡Bien!
    
    # Invertir una vez al final
    cola.reverse()  # O(n) - Pero solo una vez
    
    # Procesar
    resultado = []
    while cola:
        item = cola.pop()  # O(1) - Bien
        resultado.append(item * 2)
    
    return resultado

# Comparar rendimiento
datos_test = list(range(1000))

print("🏁 Comparando estrategias de procesamiento:")

# Estrategia naive
start = time.time()
resultado1 = estrategia_naive(datos_test)
tiempo_naive = time.time() - start

# Estrategia optimizada
start = time.time()
resultado2 = estrategia_optimizada(datos_test)
tiempo_opt = time.time() - start

print(f"Estrategia naive:      {tiempo_naive:.4f} segundos")
print(f"Estrategia optimizada: {tiempo_opt:.4f} segundos")
print(f"Mejora: {tiempo_naive/tiempo_opt:.1f}x más rápida")

# Verificar que ambas dan el mismo resultado
print(f"\n✅ Resultados iguales: {resultado1 == resultado2}")

print("\n💡 Lección: Evita operaciones O(n) en loops cuando sea posible")
```

### 🎯 Escenario 2: Búsqueda Frecuente

¿Qué hacer cuando necesitas buscar elementos frecuentemente?

```python
# 🧪 Comparando estructuras para búsqueda

import random
import timeit

def benchmark_busqueda(n_elementos, n_busquedas):
    """Compara list vs set para búsquedas frecuentes"""
    
    # Crear datos
    datos = list(range(n_elementos))
    random.shuffle(datos)
    
    # Estructura 1: List (búsqueda O(n))
    lista = datos.copy()
    
    # Estructura 2: Set (búsqueda O(1) promedio)
    conjunto = set(datos)
    
    # Elementos a buscar
    buscar = [random.randint(0, n_elementos-1) for _ in range(n_busquedas)]
    
    # Benchmark list
    def buscar_en_lista():
        for item in buscar:
            _ = item in lista
    
    # Benchmark set
    def buscar_en_set():
        for item in buscar:
            _ = item in conjunto
    
    tiempo_lista = timeit.timeit(buscar_en_lista, number=1)
    tiempo_set = timeit.timeit(buscar_en_set, number=1)
    
    return tiempo_lista, tiempo_set

# Probar con diferentes tamaños
casos = [
    (100, 1000),
    (1000, 1000),
    (10000, 1000)
]

print("🔍 Benchmark: List vs Set para búsquedas")
print("Elementos | Búsquedas | List (ms) | Set (ms) | Mejora")
print("-" * 55)

for n_elem, n_busq in casos:
    tiempo_list, tiempo_set = benchmark_busqueda(n_elem, n_busq)
    mejora = tiempo_list / tiempo_set
    
    print(f"{n_elem:8} | {n_busq:8} | {tiempo_list*1000:8.2f} | "
          f"{tiempo_set*1000:7.2f} | {mejora:5.1f}x")

print("\n💡 Lección: Si haces muchas búsquedas, usa set en lugar de list")
print("⚠️ Pero set no mantiene orden ni permite duplicados")
```

## 🎯 Guía de Decisión: ¿Qué Estructura Usar?

### 📊 Matriz de Decisión

| Patrón de Uso | 📊 **Recomendación** | 🎯 **Alternativa** | 💡 **Razón** |
|---------------|----------------------|-------------------|---------------|
| **Acceso por índice frecuente** | `list` | `tuple` si inmutable | O(1) en ambos |
| **Append frecuente** | `list` | `deque` si también pop(0) | O(1) amortizado |
| **Insert/delete al inicio** | `deque` | - | O(1) vs O(n) |
| **Búsqueda frecuente** | `set` | `dict` si key-value | O(1) vs O(n) |
| **Datos numéricos grandes** | `array.array` | `numpy.array` | Menos memoria |
| **String processing** | `str` methods | `list` para mutabilidad | Optimizado |
| **Mantener orden + únicos** | `dict` (Python 3.7+) | - | Insertion order |
| **LIFO (stack)** | `list` | - | append/pop eficientes |
| **FIFO (queue)** | `deque` | `queue.Queue` | popleft/append O(1) |

### 🚩 Red Flags - Evita estos patrones

- ❌ `list.insert(0, x)` en loops → Usa `deque.appendleft()`
- ❌ `x in large_list` repetidamente → Usa `set` o `dict`
- ❌ `list + list` en loops → Usa `list.extend()` o pre-allocate
- ❌ `string += string` en loops → Usa `''.join(list)`
- ❌ `del list[0]` repetidamente → Usa `deque.popleft()`

```python
# 🧪 Demostración de anti-patrones vs buenas prácticas

import timeit
from collections import deque

def demo_antipatrones():
    """Demuestra anti-patrones comunes y sus soluciones"""
    
    n = 1000
    
    print("❌ vs ✅ Anti-patrones y buenas prácticas:")
    print("-" * 50)
    
    # 1. String concatenation
    def string_concatenation_bad():
        result = ""
        for i in range(n):
            result += str(i)  # ❌ Crea nueva string cada vez
        return result
    
    def string_concatenation_good():
        parts = []
        for i in range(n):
            parts.append(str(i))  # ✅ Append es O(1)
        return ''.join(parts)  # ✅ Join es eficiente
    
    tiempo_bad = timeit.timeit(string_concatenation_bad, number=10)
    tiempo_good = timeit.timeit(string_concatenation_good, number=10)
    
    print(f"String concatenation:")
    print(f"  ❌ += : {tiempo_bad:.4f}s")
    print(f"  ✅ join: {tiempo_good:.4f}s")
    print(f"  Mejora: {tiempo_bad/tiempo_good:.1f}x")
    
    # 2. Queue operations
    def queue_with_list():
        queue = []
        for i in range(n):
            queue.append(i)
        
        while queue:
            queue.pop(0)  # ❌ O(n) cada vez
    
    def queue_with_deque():
        queue = deque()
        for i in range(n):
            queue.append(i)
        
        while queue:
            queue.popleft()  # ✅ O(1)
    
    tiempo_list = timeit.timeit(queue_with_list, number=10)
    tiempo_deque = timeit.timeit(queue_with_deque, number=10)
    
    print(f"\nQueue operations:")
    print(f"  ❌ list.pop(0): {tiempo_list:.4f}s")
    print(f"  ✅ deque.popleft(): {tiempo_deque:.4f}s")
    print(f"  Mejora: {tiempo_list/tiempo_deque:.1f}x")
    
    # 3. Membership testing
    data = list(range(n))
    data_set = set(data)
    
    def membership_list():
        for i in range(0, n, 10):
            _ = i in data  # ❌ O(n) cada vez
    
    def membership_set():
        for i in range(0, n, 10):
            _ = i in data_set  # ✅ O(1) promedio
    
    tiempo_list_member = timeit.timeit(membership_list, number=100)
    tiempo_set_member = timeit.timeit(membership_set, number=100)
    
    print(f"\nMembership testing:")
    print(f"  ❌ x in list: {tiempo_list_member:.4f}s")
    print(f"  ✅ x in set: {tiempo_set_member:.4f}s")
    print(f"  Mejora: {tiempo_list_member/tiempo_set_member:.1f}x")

demo_antipatrones()

print("\n💡 Recuerda: Pequeños cambios en estructuras pueden dar grandes mejoras!")
```

## 🎯 Resumen y Conceptos Clave

### ✅ Lo que aprendimos

1. **⚡ Complejidad temporal es crucial:**

   - O(1): Excelente para operaciones frecuentes
   - O(n): Aceptable para operaciones ocasionales
   - O(n²): Evitar en datos grandes

2. **📊 Operaciones clave en secuencias:**

   - **Rápidas (O(1))**: len(), [i], append(), pop()
   - **Lentas (O(n))**: insert(0), pop(0), in, remove()
   - **Moderadas**: slicing, concatenation (depende del tamaño)

3. **🎯 Principios de optimización:**

   - Conocer las complejidades de tus operaciones
   - Elegir la estructura según el patrón de uso
   - Evitar operaciones O(n) en loops cuando sea posible
   - Medir antes de optimizar

4. **🔧 Herramientas de optimización:**

   - `collections.deque` para queues
   - `set` para membership testing
   - `array.array` para datos numéricos
   - `''.join()` para string concatenation

5. **📈 Escalabilidad:**

   - Las diferencias se amplifican con datos grandes
   - O(1) vs O(n) es la diferencia entre escalable y no escalable
   - El análisis amortizado explica por qué algunas operaciones "costosas" son aceptables
