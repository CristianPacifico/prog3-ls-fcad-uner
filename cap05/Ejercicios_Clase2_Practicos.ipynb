## 🏋️ **Ejercicio 2: Amortized Analysis in Practice** (20 min)

### 📝 **Problema:**
Como consultor de rendimiento, necesitas demostrar empíricamente que el análisis amortizado realmente funciona en la práctica. Tu cliente está escéptico sobre que las operaciones "caras" de resize se compensen con las operaciones "baratas".

### 🎯 **Tu tarea:**
1. **Implementar** un tracker de costos detallado
2. **Demostrar** que el costo amortizado converge a O(1)
3. **Visualizar** el comportamiento de diferentes factores de crecimiento
4. **Analizar** el trade-off memoria vs velocidad

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import deque

class AmortizedTracker:
    """Tracker para análisis detallado de costos amortizados"""
    
    def __init__(self, growth_factor: float = 2.0):
        self.growth_factor = growth_factor
        self.operations = []
        self.resize_operations = []
        
        # Estado del array
        self.capacity = 1
        self.size = 0
        
        # Métricas acumulativas
        self.total_copies = 0
        self.total_operations = 0
    
    def append_operation(self):
        """Simula una operación append y registra costos"""
        self.total_operations += 1
        
        # ¿Necesita resize?
        if self.size >= self.capacity:
            # Resize operation - costo O(n)
            old_capacity = self.capacity
            self.capacity = int(self.capacity * self.growth_factor)
            copies_made = self.size  # Copiar elementos existentes
            self.total_copies += copies_made
            
            # Registrar resize
            self.resize_operations.append({
                'operation_number': self.total_operations,
                'old_capacity': old_capacity,
                'new_capacity': self.capacity,
                'elements_copied': copies_made,
                'cost': copies_made + 1  # Copiar + insertar
            })
            
            operation_cost = copies_made + 1
        else:
            # Operación normal - costo O(1)
            operation_cost = 1
        
        # Incrementar size
        self.size += 1
        
        # Registrar operación
        amortized_cost = (self.total_copies + self.total_operations) / self.total_operations
        
        self.operations.append({
            'operation_number': self.total_operations,
            'actual_cost': operation_cost,
            'amortized_cost': amortized_cost,
            'size': self.size,
            'capacity': self.capacity,
            'load_factor': self.size / self.capacity
        })
    
    def run_simulation(self, n_operations: int):
        """Ejecuta simulación de n operaciones append"""
        for _ in range(n_operations):
            self.append_operation()
    
    def get_analysis(self) -> dict:
        """Retorna análisis estadístico"""
        if not self.operations:
            return {}
        
        actual_costs = [op['actual_cost'] for op in self.operations]
        amortized_costs = [op['amortized_cost'] for op in self.operations]
        
        return {
            'total_operations': self.total_operations,
            'total_copies': self.total_copies,
            'resize_count': len(self.resize_operations),
            'final_amortized_cost': amortized_costs[-1],
            'max_actual_cost': max(actual_costs),
            'avg_actual_cost': sum(actual_costs) / len(actual_costs),
            'theoretical_bound': self.growth_factor / (self.growth_factor - 1),  # Para factor > 1
            'final_capacity': self.capacity,
            'final_size': self.size,
            'memory_efficiency': self.size / self.capacity
        }

# Tu código aquí: Experimenta con diferentes factores
print("🔬 Análisis Amortizado Empírico")
print("=" * 40)

growth_factors = [1.5, 2.0, 3.0]
n_ops = 1000

for factor in growth_factors:
    print(f"\n📊 Growth Factor: {factor}")
    print("-" * 30)
    
    tracker = AmortizedTracker(factor)
    tracker.run_simulation(n_ops)
    analysis = tracker.get_analysis()
    
    print(f"Total operations: {analysis['total_operations']:,}")
    print(f"Total copies: {analysis['total_copies']:,}")
    print(f"Resize count: {analysis['resize_count']}")
    print(f"Final amortized cost: {analysis['final_amortized_cost']:.3f}")
    print(f"Theoretical bound: {analysis['theoretical_bound']:.3f}")
    print(f"Memory efficiency: {analysis['memory_efficiency']:.1%}")
```

```python
# 📊 Visualización del comportamiento amortizado

def plot_amortized_analysis():
    """Crea visualizaciones del análisis amortizado"""
    
    # Tu código aquí: Crear visualizaciones
    
    # Simulación detallada
    tracker = AmortizedTracker(growth_factor=2.0)
    tracker.run_simulation(100)  # Menos operaciones para visualización clara
    
    # Extraer datos
    operations = [op['operation_number'] for op in tracker.operations]
    actual_costs = [op['actual_cost'] for op in tracker.operations]
    amortized_costs = [op['amortized_cost'] for op in tracker.operations]
    capacities = [op['capacity'] for op in tracker.operations]
    sizes = [op['size'] for op in tracker.operations]
    
    # Crear subplots
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    # 1. Costos por operación
    ax1.plot(operations, actual_costs, 'r-', alpha=0.7, label='Costo actual')
    ax1.plot(operations, amortized_costs, 'b-', linewidth=2, label='Costo amortizado')
    ax1.axhline(y=2.0, color='g', linestyle='--', alpha=0.7, label='Bound teórico (2.0)')
    ax1.set_xlabel('Operación #')
    ax1.set_ylabel('Costo')
    ax1.set_title('Costos Actual vs Amortizado')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 2. Crecimiento de capacidad
    ax2.plot(operations, sizes, 'b-', label='Size')
    ax2.plot(operations, capacities, 'r-', label='Capacity')
    ax2.fill_between(operations, sizes, capacities, alpha=0.2, color='yellow', label='Unused space')
    ax2.set_xlabel('Operación #')
    ax2.set_ylabel('Elementos')
    ax2.set_title('Size vs Capacity')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 3. Histogram de costos
    ax3.hist(actual_costs, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
    ax3.axvline(x=np.mean(actual_costs), color='red', linestyle='--', label=f'Media: {np.mean(actual_costs):.2f}')
    ax3.set_xlabel('Costo de operación')
    ax3.set_ylabel('Frecuencia')
    ax3.set_title('Distribución de Costos')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # 4. Load factor over time
    load_factors = [op['load_factor'] for op in tracker.operations]
    ax4.plot(operations, load_factors, 'g-', linewidth=2)
    ax4.axhline(y=0.5, color='r', linestyle='--', alpha=0.7, label='50% threshold')
    ax4.set_xlabel('Operación #')
    ax4.set_ylabel('Load Factor')
    ax4.set_title('Factor de Carga')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Análisis textual
    analysis = tracker.get_analysis()
    print("\n📈 Análisis de Convergencia:")
    print(f"- Costo amortizado final: {analysis['final_amortized_cost']:.3f}")
    print(f"- Bound teórico: {analysis['theoretical_bound']:.3f}")
    print(f"- Diferencia: {abs(analysis['final_amortized_cost'] - analysis['theoretical_bound']):.3f}")
    print(f"- Convergencia: {'✅ SÍ' if analysis['final_amortized_cost'] <= analysis['theoretical_bound'] + 0.1 else '❌ NO'}")

# Ejecutar visualización
try:
    plot_amortized_analysis()
except ImportError:
    print("📊 Matplotlib no disponible - mostrando análisis numérico solamente")
    
    # Análisis sin gráficos
    tracker = AmortizedTracker(growth_factor=2.0)
    tracker.run_simulation(1000)
    analysis = tracker.get_analysis()
    
    print("\n📊 Análisis Numérico:")
    for key, value in analysis.items():
        if isinstance(value, float):
            print(f"{key}: {value:.3f}")
        else:
            print(f"{key}: {value:,}")
```

## 🏋️ **Ejercicio 4: Performance Profiling** (10 min)

### 📝 **Problema:**
Tu jefe te pide optimizar una función crítica del sistema que está siendo lenta. Necesitas usar profiling para identificar exactamente dónde están los cuellos de botella.

### 🎯 **Tu tarea:**
1. **Perfilar** una función con múltiples ineficiencias
2. **Identificar** los cuellos de botella principales
3. **Optimizar** basándote en los datos del profiler
4. **Validar** las mejoras con métricas

```python
import cProfile
import pstats
import io
from functools import wraps
import time

# Función problemática para optimizar
def inefficient_data_processor(data):
    """Función con múltiples ineficiencias - TU TAREA: optimizar"""
    
    # Ineficiencia 1: String concatenation en loop
    result_string = ""
    for item in data:
        result_string += str(item) + ","  # ❌ O(n²)
    
    # Ineficiencia 2: Búsqueda lineal repetitiva
    filtered_data = []
    target_values = [100, 200, 300, 400, 500]
    for item in data:
        if item in target_values:  # ❌ O(n) por cada item
            filtered_data.append(item)
    
    # Ineficiencia 3: Operaciones redundantes
    processed_data = []
    for item in filtered_data:
        # Cálculos costosos repetidos
        squared = item ** 2
        sqrt_val = squared ** 0.5  # ❌ Redundante con item
        final_val = sqrt_val * 2
        processed_data.append(final_val)
    
    # Ineficiencia 4: Conversiones innecesarias
    final_result = []
    for item in processed_data:
        final_result.append(str(int(float(item))))  # ❌ Conversiones innecesarias
    
    return result_string, final_result

def profile_function(func, *args, **kwargs):
    """Ejecuta función con profiling detallado"""
    profiler = cProfile.Profile()
    
    # Ejecutar con profiling
    profiler.enable()
    result = func(*args, **kwargs)
    profiler.disable()
    
    # Analizar resultados
    stream = io.StringIO()
    stats = pstats.Stats(profiler, stream=stream)
    stats.sort_stats('cumulative')
    stats.print_stats(10)  # Top 10 funciones
    
    profile_output = stream.getvalue()
    return result, profile_output

# Datos de prueba
test_data = list(range(1000)) * 2  # 2000 elementos con duplicados

print("🔍 Profiling función ineficiente...")
result, profile_output = profile_function(inefficient_data_processor, test_data)

print("📊 Resultados del profiling:")
print(profile_output)

# Medir tiempo total
start_time = time.time()
inefficient_data_processor(test_data)
slow_time = time.time() - start_time

print(f"⏱️ Tiempo total (versión lenta): {slow_time:.4f} segundos")
```
