## 🤔 ¿Qué es una Deque?

Una **deque** ("deck") es una **D**ouble-**e**nded **que**ue (cola de doble extremo). Es una generalización tanto de Stack como de Queue.

### 🌟 Característica Principal
**Permite inserción y eliminación eficiente en AMBOS extremos**

```
                DEQUE
    ←→ [A] [B] [C] [D] [E] ←→
      ↑                    ↑
    FRONT                REAR
   
Operaciones en AMBOS extremos:
- add_first() / delete_first()
- add_last() / delete_last()
```

### 🧠 Conexión con Estructuras Previas

| Estructura | Restricción | Deque puede simular |
|------------|-------------|--------------------|
| **Stack** | Solo un extremo (tope) | ✅ Usa solo first/last |
| **Queue** | Dos extremos fijos | ✅ add_last + delete_first |
| **Deque** | Ambos extremos libres | 🌟 ¡La más flexible! |

**❓ Si una deque puede hacer todo lo que hacen Stack y Queue, ¿por qué necesitamos las otras estructuras?**

## 🛠️ Operaciones Fundamentales de una Deque

### 📋 Operaciones Principales

| Operación | Descripción | Ejemplo con [A, B, C] |
|-----------|-------------|------------------------|
| 📥 **add_first(item)** | Insertar al frente | add_first(X) → [X, A, B, C] |
| 📥 **add_last(item)** | Insertar al final | add_last(X) → [A, B, C, X] |
| 📤 **delete_first()** | Remover del frente | delete_first() → A, queda [B, C] |
| 📤 **delete_last()** | Remover del final | delete_last() → C, queda [A, B] |
| 👁️ **first()** | Ver primer elemento | first() → A |
| 👁️ **last()** | Ver último elemento | last() → C |
| ❓ **is_empty()** | Verificar si está vacía | is_empty() → False |
| 📏 **len()** | Obtener número de elementos | len() → 3 |

### 🎯 Relación con Estructuras Conocidas

**🔄 Simular Stack:**
- `push(x)` ≡ `add_last(x)`
- `pop()` ≡ `delete_last()`
- `top()` ≡ `last()`

**🔄 Simular Queue:**
- `enqueue(x)` ≡ `add_last(x)`
- `dequeue()` ≡ `delete_first()`
- `first()` ≡ `first()`

## 💻 Implementación de Deque

⚠️🚨 IMPORTANTE: La implementación de la deque debe permitir operaciones eficientes en ambos extremos (⏱️ O(1) inserción/eliminación). Por ello utilizaremos una **lista doblemente enlazada** 🔗.  
Antes de avanzar, revisa los contenidos de **listas enlazadas simples** y **listas doblemente enlazadas** (📘 próximo módulo, cap 7 de [DSAP]) para comprender cómo los nodos con referencias `prev` ⬅️ y `next` ➡️ permiten estas operaciones sin recorrer toda la estructura. 💡

Implementaremos una deque usando una **lista doblemente enlazada** para garantizar O(1) en ambos extremos:

````python
class _DoublyLinkedBase:
    """🔗 Clase base para listas doblemente enlazadas"""
    
    class _Node:
        """🔗 Nodo individual para lista doblemente enlazada"""
        __slots__ = '_element', '_prev', '_next'
        
        def __init__(self, element, prev, next):
            self._element = element
            self._prev = prev
            self._next = next
    
    def __init__(self):
        """🚀 Crear lista vacía con sentinelas"""
        self._header = self._Node(None, None, None)
        self._trailer = self._Node(None, None, None)
        self._header._next = self._trailer
        self._trailer._prev = self._header
        self._size = 0
    
    def __len__(self):
        """📏 Retornar número de elementos"""
        return self._size
    
    def is_empty(self):
        """❓ Verificar si está vacía"""
        return self._size == 0
    
    def _insert_between(self, element, predecessor, successor):
        """🔗 Insertar elemento entre dos nodos existentes"""
        newest = self._Node(element, predecessor, successor)
        predecessor._next = newest
        successor._prev = newest
        self._size += 1
        return newest
    
    def _delete_node(self, node):
        """🗑️ Eliminar nodo y retornar su elemento"""
        predecessor = node._prev
        successor = node._next
        predecessor._next = successor
        successor._prev = predecessor
        self._size -= 1
        element = node._element
        node._prev = node._next = node._element = None  # Ayuda al GC
        return element

class LinkedDeque(_DoublyLinkedBase):
    """🔄 Implementación de Deque usando lista doblemente enlazada"""
    
    def first(self):
        """👁️ Retornar primer elemento sin removerlo"""
        if self.is_empty():
            raise IndexError("❌ first() en deque vacía")
        return self._header._next._element
    
    def last(self):
        """👁️ Retornar último elemento sin removerlo"""
        if self.is_empty():
            raise IndexError("❌ last() en deque vacía")
        return self._trailer._prev._element
    
    def add_first(self, element):
        """📥 Agregar elemento al frente - O(1)"""
        self._insert_between(element, self._header, self._header._next)
    
    def add_last(self, element):
        """📥 Agregar elemento al final - O(1)"""
        self._insert_between(element, self._trailer._prev, self._trailer)
    
    def delete_first(self):
        """📤 Remover y retornar primer elemento - O(1)"""
        if self.is_empty():
            raise IndexError("❌ delete_first() en deque vacía")
        return self._delete_node(self._header._next)
    
    def delete_last(self):
        """📤 Remover y retornar último elemento - O(1)"""
        if self.is_empty():
            raise IndexError("❌ delete_last() en deque vacía")
        return self._delete_node(self._trailer._prev)
    
    def __str__(self):
        """🖨️ Representación string de la deque"""
        if self.is_empty():
            return "LinkedDeque([])"
        
        elements = []
        current = self._header._next
        while current is not self._trailer:
            elements.append(str(current._element))
            current = current._next
        
        return f"LinkedDeque([{' ↔ '.join(elements)}])"

# 🧪 Probemos nuestra implementación
print("🔄 Creando deque y probando operaciones básicas:")
deque = LinkedDeque()
print(f"Deque vacía: {deque}")
print(f"¿Está vacía? {deque.is_empty()}")
````

### 🧠 Comprendiendo la Lista Doblemente Enlazada

**¿Por qué usamos nodos sentinela?**

```
Sin sentinelas (problemático):
[A] ↔ [B] ↔ [C]
↑             ↑
first        last

Con sentinelas (elegante):
HEADER ↔ [A] ↔ [B] ↔ [C] ↔ TRAILER
  ↑                        ↑
nunca NULL              nunca NULL
```

**Ventajas de los sentinelas:**
- ✅ **Simplifica el código**: No casos especiales para lista vacía
- ✅ **Evita verificaciones NULL**: Siempre hay nodos prev/next
- ✅ **Operaciones uniformes**: Mismo código para cualquier posición

### 🎨 1. Sistemas Undo/Redo Avanzados

Un editor que permite deshacer y rehacer operaciones:

````python
class AdvancedEditor:
    """✏️ Editor con undo/redo usando deque"""
    
    def __init__(self):
        self.content = ""
        self.history = LinkedDeque()  # Historial de operaciones
        self.current_position = 0     # Posición actual en el historial
    
    def add_text(self, text):
        """✍️ Agregar texto y guardar en historial"""
        # Eliminar operaciones "futuras" si estamos en el medio del historial
        while self.current_position < len(self.history):
            self.history.delete_last()
        
        # Agregar nueva operación
        operation = ("add", text, len(self.content))
        self.history.add_last(operation)
        self.current_position += 1
        
        # Aplicar la operación
        self.content += text
        print(f"✍️ Agregado: '{text}'")
        print(f"📄 Contenido: '{self.content}'")
    
    def undo(self):
        """↩️ Deshacer última operación"""
        if self.current_position == 0:
            print("⚠️ No hay operaciones para deshacer")
            return
        
        # Obtener la operación a deshacer (sin removerla)
        current = self.history._header._next
        for _ in range(self.current_position - 1):
            current = current._next
        
        operation, text, position = current._element
        
        if operation == "add":
            # Deshacer agregado = quitar texto
            self.content = self.content[:position]
        
        self.current_position -= 1
        print(f"↩️ Deshecho: '{text}'")
        print(f"📄 Contenido: '{self.content}'")
    
    def redo(self):
        """↪️ Rehacer operación"""
        if self.current_position >= len(self.history):
            print("⚠️ No hay operaciones para rehacer")
            return
        
        # Obtener la operación a rehacer
        current = self.history._header._next
        for _ in range(self.current_position):
            current = current._next
        
        operation, text, position = current._element
        
        if operation == "add":
            self.content += text
        
        self.current_position += 1
        print(f"↪️ Rehecho: '{text}'")
        print(f"📄 Contenido: '{self.content}'")
    
    def show_status(self):
        """📊 Mostrar estado del editor"""
        print(f"\n📊 Estado del editor:")
        print(f"📄 Contenido: '{self.content}'")
        print(f"📚 Operaciones en historial: {len(self.history)}")
        print(f"📍 Posición actual: {self.current_position}")
        print()

# 🧪 Demostración del editor avanzado
print("✏️ EDITOR AVANZADO CON UNDO/REDO")
print("=" * 40)

editor = AdvancedEditor()
editor.show_status()

# Escribir texto
editor.add_text("Hola ")
editor.add_text("mundo ")
editor.add_text("cruel")
editor.show_status()
````

````python
# 🎮 Controla el editor

# Deshacer algunas operaciones
editor.undo()  # Quita "cruel"
editor.undo()  # Quita "mundo "
editor.show_status()

# Rehacer una operación
editor.redo()  # Vuelve "mundo "
editor.show_status()

# Agregar texto nuevo (esto eliminará "cruel" del historial)
editor.add_text("hermoso!")
editor.show_status()

# Intentar rehacer (no debería poder)
editor.redo()

# 🖊️ Tu turno: Experimenta con más operaciones undo/redo
````

### 🪟 2. Algoritmo Sliding Window

Una aplicación común: encontrar el máximo en una ventana deslizante:

````python
def sliding_window_maximum(arr, window_size):
    """🪟 Encontrar el máximo en cada ventana deslizante usando deque"""
    if not arr or window_size <= 0:
        return []
    
    deque = LinkedDeque()  # Almacena índices, no valores
    result = []
    
    print(f"🪟 Analizando ventanas de tamaño {window_size} en {arr}")
    print("=" * 50)
    
    for i in range(len(arr)):
        current_value = arr[i]
        print(f"\n📍 Procesando índice {i}, valor: {current_value}")
        
        # Remover índices que están fuera de la ventana actual
        while not deque.is_empty() and deque.first() <= i - window_size:
            removed_idx = deque.delete_first()
            print(f"  🗑️ Removiendo índice fuera de ventana: {removed_idx}")
        
        # Remover índices de elementos menores que el actual
        # (nunca serán el máximo mientras el actual esté en la ventana)
        while not deque.is_empty() and arr[deque.last()] <= current_value:
            removed_idx = deque.delete_last()
            print(f"  📉 Removiendo índice de valor menor: {removed_idx} (valor: {arr[removed_idx]})")
        
        # Agregar el índice actual
        deque.add_last(i)
        print(f"  📥 Agregando índice actual: {i}")
        
        # Si hemos procesado al menos window_size elementos
        if i >= window_size - 1:
            max_idx = deque.first()  # El frente siempre tiene el máximo
            max_value = arr[max_idx]
            result.append(max_value)
            
            window_start = i - window_size + 1
            window_end = i
            window = arr[window_start:window_end + 1]
            print(f"  🪟 Ventana [{window_start}:{window_end + 1}] = {window}")
            print(f"  🌟 Máximo: {max_value} (índice {max_idx})")
    
    return result

# 🧪 Ejemplo práctico
print("🪟 ALGORITMO SLIDING WINDOW MAXIMUM")
print("=" * 40)

test_array = [1, 3, -1, -3, 5, 3, 6, 7]
window_size = 3

result = sliding_window_maximum(test_array, window_size)
print(f"\n🎯 RESULTADO FINAL: {result}")
print(f"Array original: {test_array}")
print(f"Tamaño ventana: {window_size}")
````

### 🔍 3. Verificador de Palíndromos Eficiente

Usando deque para verificar palíndromos de manera elegante:

````python
def is_palindrome_deque(text):
    """🔍 Verificar palíndromo usando deque"""
    # Limpiar texto: solo letras, sin espacios, todo minúscula
    clean_text = ''.join(char.lower() for char in text if char.isalnum())
    
    if len(clean_text) <= 1:
        return True
    
    deque = LinkedDeque()
    
    # Agregar todos los caracteres a la deque
    for char in clean_text:
        deque.add_last(char)
    
    print(f"🔍 Verificando: '{text}'")
    print(f"📝 Texto limpio: '{clean_text}'")
    print(f"🔄 Deque inicial: {deque}")
    
    # Comparar extremos mientras haya más de un elemento
    step = 1
    while len(deque) > 1:
        first_char = deque.delete_first()
        last_char = deque.delete_last()
        
        print(f"  Paso {step}: Comparando '{first_char}' vs '{last_char}'")
        
        if first_char != last_char:
            print(f"  ❌ No coinciden: NO es palíndromo")
            return False
        
        print(f"  ✅ Coinciden")
        print(f"     Deque restante: {deque}")
        step += 1
    
    print(f"  🎉 Todos los caracteres coinciden: SÍ es palíndromo")
    return True

# 🧪 Probando el verificador de palíndromos
print("🔍 VERIFICADOR DE PALÍNDROMOS CON DEQUE")
print("=" * 45)

test_cases = [
    "ana",
    "A man a plan a canal Panama",
    "race a car",
    "Anita lava la tina",
    "hello"
]

for test in test_cases:
    result = is_palindrome_deque(test)
    print(f"\n{'='*30}")
    print(f"Resultado: {'✅ PALÍNDROMO' if result else '❌ NO PALÍNDROMO'}")
    print()
````

## 🏆 Deque con collections.deque de Python

Python incluye una implementación optimizada de deque en el módulo `collections`:

````python
from collections import deque

# 🐍 Deque nativa de Python
print("🐍 EXPLORANDO collections.deque")
print("=" * 35)

# Crear deque con elementos iniciales
python_deque = deque([1, 2, 3, 4, 5])
print(f"Deque inicial: {python_deque}")

# Operaciones en ambos extremos
python_deque.appendleft(0)    # add_first
python_deque.append(6)        # add_last
print(f"Después de appendleft(0) y append(6): {python_deque}")

left = python_deque.popleft()  # delete_first
right = python_deque.pop()     # delete_last
print(f"Removido izq: {left}, der: {right}")
print(f"Deque final: {python_deque}")

# 🚀 Funcionalidades adicionales
print("\n🚀 Funcionalidades adicionales:")

# Rotación
python_deque.rotate(2)  # Rotar 2 posiciones a la derecha
print(f"Después de rotate(2): {python_deque}")

python_deque.rotate(-1)  # Rotar 1 posición a la izquierda
print(f"Después de rotate(-1): {python_deque}")

# Deque con tamaño máximo
limited_deque = deque(maxlen=3)
for i in range(5):
    limited_deque.append(i)
    print(f"Agregando {i}: {limited_deque}")

print("\n💡 ¡Cuando está llena, agregar un elemento expulsa del otro extremo!")
````

## 🎊 ¡Resumen!

### ✅ Stack, Queue y Deque - Lo que Dominamos:

#### 📚 **Stack (LIFO)**
- **Concepto**: Last In, First Out
- **Operaciones**: push, pop, top (solo un extremo)
- **Aplicaciones**: Undo/redo, navegación atrás, paréntesis balanceados
- **Complejidad**: O(1) para todas las operaciones

#### 🚶‍♂️ **Queue (FIFO)**
- **Concepto**: First In, First Out
- **Operaciones**: enqueue, dequeue, first (dos extremos fijos)
- **Implementación**: Lista circular para eficiencia
- **Aplicaciones**: Turnos, impresión, BFS, simulaciones
- **Complejidad**: O(1) con implementación correcta

#### 🔄 **Deque (Double-ended)**
- **Concepto**: Inserción/eliminación en ambos extremos
- **Operaciones**: add_first/last, delete_first/last (ambos extremos)
- **Implementación**: Lista doblemente enlazada
- **Aplicaciones**: Sliding window, palindromes, undo/redo avanzado
- **Complejidad**: O(1) para todas las operaciones de extremos

### 🎯 **Criterios de Selección**

| Necesidad | Estructura Recomendada | Razón |
|-----------|----------------------|-------|
| Acceso al más reciente | Stack | LIFO natural |
| Procesamiento en orden | Queue | FIFO justo |
| Máxima flexibilidad | Deque | Ambos extremos |
| Memoria limitada | Stack/Queue | Menos overhead |
| Algoritmos específicos | Según algoritmo | Cada uno tiene su lugar |

### 🧠 **Conceptos Fundamentales Integrados**
- **Abstracción**: Interfaces claras independientes de implementación
- **Eficiencia**: Importancia de la complejidad temporal
- **Trade-offs**: Flexibilidad vs simplicidad vs memoria
- **Aplicabilidad**: Reconocer patrones en problemas reales
