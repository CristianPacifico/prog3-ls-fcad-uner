## 🤔 ¿Qué es una Pila (Stack)?

Una **pila** es una estructura de datos lineal que sigue el principio **LIFO** (**L**ast **I**n, **F**irst **O**ut).

### 🌍 Analogías del Mundo Real

![Plate Stack](https://cdn-icons-png.flaticon.com/256/16376/16376688.png "Fuente: Flaticon")

- Colocas un plato **arriba** de la pila
- Tomas un plato **desde arriba** de la pila
- ¡El último plato que pusiste es el primero que sacas!

- 📚 **Pila de libros**

## 🛠️ El TDA PILA y Operaciones Fundamentales

Las pilas son las estructuras de datos más simples, pero también se encuentran entre las más importantes. Se utilizan en una gran variedad de aplicaciones y como herramienta para muchas estructuras de datos y algoritmos más sofisticados. Formalmente, una pila es un tipo de datos abstracto (TDA) tal que una instancia **S** admite los métodos siguientes:


| Método | Descripción | Ejemplo con [1, 2, 3] |
|-----------|-------------|------------------------|
| 📥 **S.push(item)** | Agregar elemento al tope de S | S.push(4) → [1, 2, 3, 4] |
| 📤 **S.pop()** | Remover y retornar el elemento del tope de S | S.pop() → 3, queda [1, 2] |
| 👁️ **S.top()** / **S.peek()** | Ver el elemento del tope de S sin removerlo | S.top() → 3, queda [1, 2, 3] |
| ❓ **S.is_empty()** | Verificar si la pila S está vacía | S.is_empty() → False |
| 📏 **len(S)** | Obtener el número de elementos de S | len(S) → 3 |

Por convención, asumimos que una pila recién creada está vacía y que no hay un límite a priori en la capacidad de la pila. Los elementos añadidos a la pila pueden ser de cualquier tipo.

## 💻 Implementación de Stack en Python

Podemos implementar una pila con bastante facilidad almacenando sus elementos en una lista de Python 🐍. La clase `list` ya admite añadir un elemento al final con el método `append` y eliminar el último elemento con el método `pop` 📥📤, por lo que es natural alinear la parte superior de la pila al final de la lista.

![figure-6.2.png](images/figure-6.2.png)

Aunque un programador podría utilizar directamente la clase `list` en lugar de una clase _stack_ formal 🤔, las listas también incluyen comportamientos (por ejemplo, añadir o eliminar elementos de posiciones arbitrarias) que romperían la abstracción que representa el TDA pila ⚠️.

Además, la terminología utilizada por la clase `list` no se ajusta con precisión a la nomenclatura tradicional de un TDA de pila 📚, en particular la distinción entre `append` y `push`. En su lugar, demostramos cómo utilizar una lista para el almacenamiento interno 🏗️, al tiempo que proporcionamos una interfaz pública coherente con una pila ✨.



### 🔧 El patrón `adapter`

El patrón de diseño `adapter` se aplica a cualquier contexto en el que queramos modificar eficazmente una clase existente para que sus métodos coincidan con los de una clase o interfaz relacionada, pero diferente. 

![Adapter Pattern](https://refactoring.guru/images/patterns/content/adapter/adapter-en.png "Patrón Adapter")

Una forma general de aplicar el patrón adaptador es definir una nueva clase de tal manera que contenga una instancia de la clase existente como un atributo oculto y, a continuación, implementar cada método de la nueva clase utilizando los métodos de esta variable de instancia oculta. 

🔌 Al aplicar el patrón `adapter` de esta manera, hemos creado una nueva clase que realiza algunas de las mismas funciones que una clase existente, pero reempaquetadas de una manera más conveniente. 

En el contexto del TDA Pila/Stack, podemos adaptar la clase `list` de Python utilizando las correspondencias que se muestran en la tabla:

| Stack Method | Realization with Python list | 🎯 Descripción |
|--------------|-------------------------------|--------------------|
| S.push(e) | L.append(e) | 📥 Agregar al final |
| S.pop() | L.pop() | 📤 Remover del final |
| S.top() | L[-1] | 👁️ Ver último elemento |
| S.is_empty() | len(L) == 0 | ❓ Verificar si está vacía |
| len(S) | len(L) | 📏 Obtener tamaño |

🧩 **¡El patrón adapter nos permite usar la funcionalidad de `list` con la interfaz semántica de `Stack`!**

## Implementación de un TDA Stack utilizando un `list` Python

Utilizamos el patrón de diseño adapter para definir una clase `Stack` que utiliza una lista Python subyacente para el almacenamiento. 

````python
class Stack:
    """🏗️ Implementación de una Pila usando lista de Python"""
    
    def __init__(self):
        """🚀 Inicializar una pila vacía"""
        self._data = []  # Lista para almacenar los elementos
    
    def __len__(self):
        """📏 Retornar el número de elementos en la pila"""
        return len(self._data)
    
    def is_empty(self):
        """❓ Verificar si la pila está vacía"""
        return len(self._data) == 0
    
    def push(self, item):
        """📥 Agregar un elemento al tope de la pila"""
        self._data.append(item)
    
    def pop(self):
        """📤 Remover y retornar el elemento del tope"""
        if self.is_empty():
            raise IndexError("❌ pop() en pila vacía")
        return self._data.pop()
    
    def top(self):
        """👁️ Retornar el elemento del tope sin removerlo"""
        if self.is_empty():
            raise IndexError("❌ top() en pila vacía")
        return self._data[-1]
    
    def __str__(self):
        """🖨️ Representación string de la pila"""
        return f"Stack({self._data})"

# 🧪 ¡Vamos a probar nuestra implementación!
print("🎪 Creando una pila y probando operaciones:")
stack = Stack()
print(f"Pila vacía: {stack}")
print(f"¿Está vacía? {stack.is_empty()}")
````

## ⚡ Análisis de Complejidad Temporal

| Operación | Complejidad | Explicación |
|-----------|-------------|-------------|
| **push()** | O(1) | Agregar al final de la lista es constante |
| **pop()** | O(1) | Remover del final de la lista es constante |
| **top()** | O(1) | Acceder al último elemento es constante |
| **is_empty()** | O(1) | Verificar longitud es constante |
| **len()** | O(1) | Python mantiene el tamaño internamente |

## 🎯 Aplicación Práctica: Verificación de Paréntesis Balanceados

Una de las aplicaciones más comunes de las pilas es verificar si los paréntesis, corchetes y llaves están balanceados en una expresión.

### 🤔 El Problema
Dada una cadena con caracteres `()`, `[]`, `{}`, determinar si están correctamente balanceados.

**✅ Ejemplos válidos:**
- `"()"`
- `"()[]{}"` 
- `"([{}])"`
- `"{[()]}"`

**❌ Ejemplos inválidos:**
- `"("`
- `"()]"`
- `"([)]"`
- `"((("`

### 💡 La Estrategia
1. 🔍 Recorremos cada carácter
2. 📥 Si es un símbolo de apertura `(`, `[`, `{` → lo apilamos
3. 📤 Si es un símbolo de cierre `)`, `]`, `}` → verificamos que coincida con el tope
4. ✅ Al final, la pila debe estar vacía

````python
def is_balanced(expression):
    """🔍 Verificar si los símbolos están balanceados usando una pila"""
    stack = Stack()
    
    # 🗺️ Mapeo de símbolos de cierre a apertura
    closing_to_opening = {')': '(', ']': '[', '}': '{'}
    opening_symbols = set(['(', '[', '{'])
    
    for char in expression:
        if char in opening_symbols:
            # 📥 Símbolo de apertura: lo apilamos
            stack.push(char)
            print(f"📥 Apilando '{char}': {stack}")
            
        elif char in closing_to_opening:
            # 📤 Símbolo de cierre: verificamos coincidencia
            if stack.is_empty():
                print(f"❌ Error: '{char}' sin apertura correspondiente")
                return False
            
            top_symbol = stack.pop()
            expected = closing_to_opening[char]
            print(f"📤 Desapilando '{top_symbol}' para cerrar '{char}'")
            
            if top_symbol != expected:
                print(f"❌ Error: esperaba '{expected}' pero encontré '{top_symbol}'")
                return False
    
    # ✅ Al final, la pila debe estar vacía
    result = stack.is_empty()
    if not result:
        print(f"❌ Error: símbolos sin cerrar: {stack}")
    else:
        print(f"✅ ¡Expresión balanceada!")
    
    return result

# 🧪 Probemos con algunos ejemplos
test_cases = [
    "()",      # ✅ Válido
    "([{}])",  # ✅ Válido  
    "([)]",    # ❌ Inválido
    "(((",     # ❌ Inválido
]

for test in test_cases:
    print(f"\n🧪 Probando: '{test}'")
    is_balanced(test)
    print("-" * 40)
````
