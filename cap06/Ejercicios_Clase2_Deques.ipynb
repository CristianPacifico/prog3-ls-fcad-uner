## 🪟 Ejercicio 2: Algoritmo Sliding Window Maximum Optimizado

### 📋 Consigna
Implementen el algoritmo **Sliding Window Maximum** usando deque para mantener candidatos al máximo en O(1) por elemento.

**Problema:**
Dado un array y un tamaño de ventana k, encontrar el máximo en cada ventana deslizante.

**Ejemplo:**
- Array: `[1, 3, -1, -3, 5, 3, 6, 7]`, k=3
- Resultado: `[3, 3, 5, 5, 6, 7]`

### 🎯 Objetivo
Aplicar deque en un algoritmo de optimización avanzado.

```python
from collections import deque as PythonDeque  # Usaremos la implementación de Python

def sliding_window_maximum_optimized(arr, k):
    """🪟 Algoritmo optimizado de Sliding Window Maximum"""
    
    if not arr or k <= 0 or k > len(arr):
        return []
    
    # 🖊️ Paso 1: Inicializar estructuras
    dq = PythonDeque()  # Almacena ÍNDICES, no valores
    result = []
    
    print(f"🪟 Array: {arr}, Tamaño ventana: {k}")
    print("=" * 50)
    
    # 🖊️ Paso 2: Procesar cada elemento
    for i in range(len(arr)):
        current_value = arr[i]
        print(f"\n📍 Procesando índice {i}: valor {current_value}")
        
        # 🖊️ Paso 2a: Remover índices fuera de la ventana actual
        # 💡 Si el frente de la deque está fuera de [i-k+1, i], removerlo
        while dq and dq[0] <= i - k:
            removed_idx = dq.popleft()
            print(f"  🗑️ Removiendo índice fuera de ventana: {removed_idx}")
        
        # 🖊️ Paso 2b: Mantener orden decreciente en la deque
        # 💡 Remover del final todos los índices con valores menores al actual
        while dq and arr[dq[-1]] <= current_value:
            removed_idx = dq.pop()
            print(f"  📉 Removiendo índice con valor menor: {removed_idx} (valor: {arr[removed_idx]})")
        
        # 🖊️ Paso 2c: Agregar índice actual
        dq.append(i)
        print(f"  📥 Agregando índice actual: {i}")
        print(f"  🔄 Deque de índices: {list(dq)}")
        print(f"  🔄 Valores correspondientes: {[arr[idx] for idx in dq]}")
        
        # 🖊️ Paso 2d: Si hemos formado una ventana completa, agregar resultado
        if i >= k - 1:
            max_idx = dq[0]  # El frente siempre tiene el máximo
            max_value = arr[max_idx]
            result.append(max_value)
            
            # Mostrar información de la ventana
            window_start = i - k + 1
            window_end = i
            window_values = arr[window_start:window_end + 1]
            print(f"  🪟 Ventana [{window_start}:{window_end + 1}]: {window_values}")
            print(f"  🌟 Máximo: {max_value} (índice {max_idx})")
    
    return result

def compare_with_naive(arr, k):
    """🔍 Comparar con algoritmo naive para verificar correctitud"""
    
    # Algoritmo naive O(n*k)
    naive_result = []
    for i in range(len(arr) - k + 1):
        window = arr[i:i + k]
        naive_result.append(max(window))
    
    # Algoritmo optimizado O(n)
    optimized_result = sliding_window_maximum_optimized(arr, k)
    
    print(f"\n🔍 COMPARACIÓN DE RESULTADOS:")
    print(f"Algoritmo naive (O(n*k)):     {naive_result}")
    print(f"Algoritmo optimizado (O(n)):  {optimized_result}")
    print(f"¿Coinciden? {'✅ SÍ' if naive_result == optimized_result else '❌ NO'}")
    
    return naive_result == optimized_result

# 🧪 Casos de prueba
test_cases = [
    ([1, 3, -1, -3, 5, 3, 6, 7], 3),
    ([1, 2, 3, 4, 5], 2),
    ([5, 4, 3, 2, 1], 3),
    ([1], 1),
    ([9, 10, 9, -7, -4, -8, 2, -6], 5)
]

print("🧪 TESTING SLIDING WINDOW MAXIMUM")
print("=" * 40)

# 🎮 Descomenten para probar cada caso
# for i, (arr, k) in enumerate(test_cases, 1):
#     print(f"\n\n🧪 CASO DE PRUEBA {i}")
#     print("=" * 25)
#     compare_with_naive(arr, k)
#     input("\n⏸️ Presiona Enter para continuar...")
```

## 🔍 Ejercicio 3: Detector Avanzado de Palíndromos

### 📋 Consigna
Implementen un detector de palíndromos que maneje múltiples casos:

**Funcionalidades:**
- 🔤 **Palíndromos básicos**: "ana", "oso"
- 📝 **Frases con espacios**: "anita lava la tina"
- 🔢 **Alfanuméricos**: "A man a plan a canal Panama"
- 🎯 **Palíndromos parciales**: Encontrar la subsecuencia palindrómica más larga

### 🎯 Objetivo
Aplicar deque en algoritmos de procesamiento de texto avanzado.

```python
import re
from collections import deque as PythonDeque

class AdvancedPalindromeDetector:
    """🔍 Detector avanzado de palíndromos usando deque"""
    
    @staticmethod
    def clean_text(text, mode='alphanumeric'):
        """🧹 Limpiar texto según el modo especificado"""
        if mode == 'basic':
            return text.lower()
        elif mode == 'no_spaces':
            return text.replace(' ', '').lower()
        elif mode == 'alphanumeric':
            return re.sub(r'[^a-zA-Z0-9]', '', text).lower()
        return text
    
    @staticmethod
    def is_palindrome_deque(text, mode='alphanumeric', verbose=True):
        """🔍 Verificar palíndromo usando deque con modo configurable"""
        
        # 🖊️ Paso 1: Limpiar texto
        clean = AdvancedPalindromeDetector.clean_text(text, mode)
        
        if verbose:
            print(f"🔍 Verificando: '{text}'")
            print(f"🧹 Texto limpio ({mode}): '{clean}'")
        
        if len(clean) <= 1:
            if verbose:
                print(f"✅ Palíndromo trivial")
            return True
        
        # 🖊️ Paso 2: Llenar deque con caracteres
        dq = PythonDeque(clean)
        
        if verbose:
            print(f"🔄 Deque inicial: {list(dq)}")
        
        # 🖊️ Paso 3: Comparar extremos iterativamente
        step = 1
        while len(dq) > 1:
            left = dq.popleft()
            right = dq.pop()
            
            if verbose:
                print(f"  Paso {step}: '{left}' vs '{right}' | Restante: {list(dq)}")
            
            if left != right:
                if verbose:
                    print(f"  ❌ No coinciden → NO es palíndromo")
                return False
            
            step += 1
        
        if verbose:
            print(f"  ✅ Todos los pares coinciden → SÍ es palíndromo")
        return True
    
    @staticmethod
    def find_longest_palindromic_substring(text, verbose=True):
        """🎯 Encontrar la subsecuencia palindrómica más larga"""
        
        clean = AdvancedPalindromeDetector.clean_text(text)
        n = len(clean)
        
        if n <= 1:
            return clean
        
        longest = ""
        
        if verbose:
            print(f"🎯 Buscando palíndromo más largo en: '{text}'")
            print(f"🧹 Texto limpio: '{clean}'")
            print("🔍 Probando todas las subsecuencias...")
        
        # 🖊️ Probar todas las subsecuencias posibles
        for i in range(n):
            for j in range(i + 1, n + 1):
                substring = clean[i:j]
                
                # Usar deque para verificar si es palíndromo
                if AdvancedPalindromeDetector.is_palindrome_deque(substring, 'basic', False):
                    if len(substring) > len(longest):
                        longest = substring
                        if verbose:
                            print(f"  🌟 Nuevo máximo: '{longest}' (longitud {len(longest)})")
        
        if verbose:
            print(f"\n🏆 Palíndromo más largo: '{longest}' (longitud {len(longest)})")
        
        return longest
    
    @staticmethod
    def palindrome_analysis(text):
        """📊 Análisis completo de palíndromos"""
        
        print(f"📊 ANÁLISIS COMPLETO DE PALÍNDROMOS")
        print(f"Texto: '{text}'")
        print("=" * 50)
        
        # Diferentes modos de verificación
        modes = {
            'basic': 'Básico (solo minúsculas)',
            'no_spaces': 'Sin espacios',
            'alphanumeric': 'Solo alfanumérico'
        }
        
        for mode, description in modes.items():
            print(f"\n🔍 Modo: {description}")
            is_pal = AdvancedPalindromeDetector.is_palindrome_deque(text, mode, verbose=False)
            clean = AdvancedPalindromeDetector.clean_text(text, mode)
            print(f"   Texto procesado: '{clean}'")
            print(f"   Resultado: {'✅ PALÍNDROMO' if is_pal else '❌ NO PALÍNDROMO'}")
        
        # Encontrar palíndromo más largo
        print(f"\n🎯 Palíndromo más largo:")
        longest = AdvancedPalindromeDetector.find_longest_palindromic_substring(text, verbose=False)
        print(f"   '{longest}' (longitud: {len(longest)})")

# 🧪 Casos de prueba avanzados
test_cases = [
    "ana",
    "A man a plan a canal Panama",
    "race a car",
    "Anita lava la tina",
    "Was it a car or a cat I saw?",
    "No 'x' in Nixon",
    "Madam, I'm Adam",
    "Hello, World!"
]

print("🧪 TESTING ADVANCED PALINDROME DETECTOR")
print("=" * 45)

# 🎮 Descomenten para probar casos individuales
# for i, test_case in enumerate(test_cases, 1):
#     print(f"\n\n🧪 CASO DE PRUEBA {i}")
#     print("=" * 25)
#     AdvancedPalindromeDetector.palindrome_analysis(test_case)
#     input("\n⏸️ Presiona Enter para continuar...")

# 🎮 O ejecutar análisis detallado de un caso específico
# print("\n🔍 ANÁLISIS DETALLADO")
# AdvancedPalindromeDetector.is_palindrome_deque("A man a plan a canal Panama", 'alphanumeric', True)
```
