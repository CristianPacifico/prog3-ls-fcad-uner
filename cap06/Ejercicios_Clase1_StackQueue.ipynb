## 🎮 Ejercicio 4: Simulador de Procesos del Sistema

### 📋 Consigna
Implementen un simulador simple que modele cómo un sistema operativo maneja procesos usando **Queue para scheduling** y **Stack para call stack**.

**Funcionalidades:**
- 📋 **Cola de procesos**: Procesos esperando ser ejecutados
- 📚 **Call stack**: Funciones que está ejecutando el proceso actual
- ⚡ **Ejecución paso a paso**: Simular tiempo de CPU

### 🎯 Objetivo
Aplicar ambas estructuras en un problema del mundo real.

```python
import time
import random

class Process:
    """💻 Representa un proceso del sistema"""
    
    def __init__(self, pid, name, functions):
        self.pid = pid
        self.name = name
        self.functions = functions.copy()  # Lista de funciones a ejecutar
        self.call_stack = []  # Stack para llamadas de función
        self.completed = False
    
    def __str__(self):
        return f"Process({self.pid}, {self.name}, {len(self.functions)} funcs)"

class OSSimulator:
    """🖥️ Simulador simple de Sistema Operativo"""
    
    def __init__(self):
        # 🖊️ Inicializar estructuras de datos
        self.ready_queue = []  # Queue de procesos listos
        self.current_process = None
        self.completed_processes = []
        self.time_slice = 2  # Quantum de tiempo por proceso
    
    def add_process(self, process):
        """📥 Agregar proceso a la cola de ready"""
        # 🖊️ Tu código aquí
        print(f"📥 Proceso agregado: {process}")
    
    def schedule_next_process(self):
        """⚡ Seleccionar siguiente proceso para ejecutar"""
        # 🖊️ Tu código aquí
        # 💡 Sacar proceso de ready_queue y asignarlo a current_process
        if len(self.ready_queue) > 0:
            # Usar dequeue (¿cómo lo implementarían?)
            pass
        
        print(f"⚡ Ejecutando: {self.current_process}")
    
    def execute_function(self, process, function_name):
        """🔧 Ejecutar una función del proceso"""
        # 🖊️ Agregar función al call stack del proceso
        print(f"   📚 Llamando función: {function_name}")
        
        # Simular tiempo de ejecución
        execution_time = random.uniform(0.5, 1.5)
        print(f"   ⏱️ Ejecutando por {execution_time:.1f}s...")
        time.sleep(execution_time)  # Simulación real
        
        # 🖊️ Remover función del call stack
        print(f"   ✅ Función completada: {function_name}")
    
    def run_time_slice(self):
        """⏰ Ejecutar process por un quantum de tiempo"""
        if self.current_process is None:
            return
        
        print(f"\n⏰ Quantum de tiempo para {self.current_process.name}")
        
        # Ejecutar funciones por el tiempo asignado
        start_time = time.time()
        while (time.time() - start_time < self.time_slice and 
               len(self.current_process.functions) > 0):
            
            # 🖊️ Tomar siguiente función y ejecutar
            function_name = self.current_process.functions.pop(0)
            self.execute_function(self.current_process, function_name)
        
        # 🖊️ Verificar si el proceso terminó
        if len(self.current_process.functions) == 0:
            print(f"   🎉 Proceso {self.current_process.name} COMPLETADO")
            self.completed_processes.append(self.current_process)
            self.current_process = None
        else:
            print(f"   ⏸️ Quantum agotado, regresando a cola")
            # 🖊️ Devolver proceso a ready_queue (¿al final?)
            self.current_process = None
    
    def show_status(self):
        """📊 Mostrar estado del sistema"""
        print(f"\n📊 ESTADO DEL SISTEMA")
        print(f"   Procesos en cola: {len(self.ready_queue)}")
        if self.current_process:
            print(f"   Proceso actual: {self.current_process.name}")
            print(f"   Funciones restantes: {len(self.current_process.functions)}")
        print(f"   Procesos completados: {len(self.completed_processes)}")
    
    def run_simulation(self):
        """🚀 Ejecutar simulación completa"""
        print("🚀 INICIANDO SIMULACIÓN DE SO")
        print("=" * 35)
        
        step = 1
        while len(self.ready_queue) > 0 or self.current_process is not None:
            print(f"\n--- PASO {step} ---")
            
            # Asignar proceso si no hay uno ejecutando
            if self.current_process is None:
                self.schedule_next_process()
            
            # Ejecutar quantum
            if self.current_process:
                self.run_time_slice()
            
            self.show_status()
            step += 1
            
            # Pausa para observar
            input("\n⏸️ Presiona Enter para continuar...")
        
        print("\n🎉 SIMULACIÓN COMPLETADA")
        print(f"Procesos ejecutados: {[p.name for p in self.completed_processes]}")

# 🧪 Configurar y ejecutar simulación
def setup_simulation():
    """🛠️ Configurar procesos para la simulación"""
    os_sim = OSSimulator()
    
    # Crear procesos con diferentes funciones
    processes = [
        Process(1, "WebBrowser", ["init()", "load_page()", "render()", "cleanup()"]),
        Process(2, "TextEditor", ["start()", "load_file()", "parse()", "display()"]),
        Process(3, "Calculator", ["init()", "show_ui()", "wait_input()"])
    ]
    
    # Agregar procesos al simulador
    for process in processes:
        os_sim.add_process(process)
    
    return os_sim

# 🎮 Descomenten para ejecutar la simulación
# simulator = setup_simulation()
# simulator.run_simulation()
```
