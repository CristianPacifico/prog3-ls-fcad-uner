## 🧪 Funciones de Utilidad

Definimos funciones que nos ayudarán en los ejercicios:

````python
def test_exercise(exercise_name, test_function, expected_result=None):
    """
    Función para probar ejercicios de manera uniforme.
    
    Args:
        exercise_name: Nombre del ejercicio
        test_function: Función que ejecuta el test
        expected_result: Resultado esperado (opcional)
    """
    print(f"\n🧪 Probando: {exercise_name}")
    print("-" * 50)
    
    try:
        start_time = time.time()
        result = test_function()
        end_time = time.time()
        
        print(f"✅ Resultado: {result}")
        print(f"⏱️  Tiempo: {(end_time - start_time)*1000:.2f}ms")
        
        if expected_result is not None:
            if result == expected_result:
                print(f"✅ Test PASADO")
            else:
                print(f"❌ Test FALLIDO - Esperado: {expected_result}")
        
    except Exception as e:
        print(f"❌ Error: {e}")

def generate_test_data(size, data_type="list"):
    """
    Genera datos de prueba de diferentes tipos.
    
    Args:
        size: Tamaño de los datos
        data_type: Tipo de datos ("list", "random", "sorted", "reverse")
    
    Returns:
        Datos de prueba generados
    """
    if data_type == "list" or data_type == "random":
        return [random.randint(1, 1000) for _ in range(size)]
    elif data_type == "sorted":
        return list(range(1, size + 1))
    elif data_type == "reverse":
        return list(range(size, 0, -1))
    else:
        return list(range(size))

def measure_performance(func, data_sizes, iterations=10):
    """
    Mide el rendimiento de una función.
    
    Args:
        func: Función a medir
        data_sizes: Lista de tamaños de datos
        iterations: Número de iteraciones
    
    Returns:
        Lista de tiempos promedio
    """
    times = []
    
    for size in data_sizes:
        total_time = 0
        
        for _ in range(iterations):
            test_data = generate_test_data(size)
            
            start_time = time.time()
            func(test_data.copy())  # Usar copia para evitar efectos secundarios
            end_time = time.time()
            
            total_time += (end_time - start_time)
        
        avg_time = total_time / iterations
        times.append(avg_time)
    
    return times

print("🛠️  Funciones de utilidad definidas")
````

### 📊 Comparación recursiva vs. iterativa para flatten


````python
from collections import deque
import time

def flatten_iterativo(lst: list) -> list:
    """Versión iterativa usando una pila explícita (deque)."""
    result = []
    stack = deque(reversed(lst))  # empujar en orden inverso para procesar de izq a der
    while stack:
        item = stack.pop()
        if isinstance(item, list):
            stack.extend(reversed(item))  # empujar sublista en orden inverso
        else:
            result.append(item)
    return result


# Verificar equivalencia
import random
def make_random_nested(depth, max_items=5):
    if depth == 0 or random.random() < 0.4:
        return random.randint(0, 99)
    return [make_random_nested(depth-1, max_items) for _ in range(random.randint(1, max_items))]

random.seed(42)
for _ in range(50):
    nested = [make_random_nested(3) for _ in range(5)]
    r1 = flatten(nested)
    r2 = flatten_iterativo(nested)
    assert r1 == r2, f"Discrepancia: {nested}"

print("✓ flatten recursivo == flatten iterativo para 50 listas aleatorias")

# Comparación de rendimiento
import matplotlib.pyplot as plt
sizes = [100, 500, 1000, 2000, 3000]
t_rec, t_it = [], []
for s in sizes:
    nested = [[random.randint(0, 99) for _ in range(10)] for _ in range(s)]
    t0 = time.perf_counter(); flatten(nested); t_rec.append(time.perf_counter()-t0)
    t0 = time.perf_counter(); flatten_iterativo(nested); t_it.append(time.perf_counter()-t0)

plt.figure(figsize=(8, 4))
plt.plot(sizes, t_rec, 'o-', label='recursivo', color='steelblue', lw=2)
plt.plot(sizes, t_it,  's-', label='iterativo', color='green',    lw=2)
plt.title("flatten: recursivo vs iterativo"); plt.xlabel("# de sub-listas"); plt.ylabel("tiempo (s)")
plt.legend(); plt.grid(True, alpha=0.4); plt.tight_layout(); plt.show()
````

| # | Ejercicio | Tipo recursión | T(n) recurrencia | Solución |
|---|-----------|---------------|-----------------|---------|
| 1 | `max_list` | Lineal | T(n)=T(n-1)+O(1) | O(n) |
| 2 | `hanoi_traza` | Binaria c/ estado | T(n)=2T(n-1)+O(1) | O(2^n) |
| 3 | `flatten` | Múltiple sobre árbol | T(n)=T(#hijos)+O(n) | O(n) nodos |
| 4 | `contar_caminos` | Binaria | T=2T+O(1) → sin memo O(2^(m+n)) | O(2^(m+n)) / O(m·n) memo |
| ★ | `combinaciones` | Binaria (Pascal) | C(n,k)=C(n-1,k-1)+C(n-1,k) | O(2^n) / O(n·k) memo |

### 🎯 Conexiones clave descubiertas
- `contar_caminos(0,0,m,n)` = `math.comb(m+n-2, m-1)` — la grilla y los combinatorios son el mismo problema
- Torres de Hanói es **O(2^n) inevitable** porque el output tiene ese tamaño
- `flatten` aplica el mismo patrón de DFS que `disk_usage` del notebook 02
