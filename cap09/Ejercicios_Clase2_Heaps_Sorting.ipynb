## 🟡 Ejercicio 1 — Traza de Up-Heap y Down-Heap

### Parte A: Up-Heap

Dado el siguiente heap (como array):

```
Heap array: [2, 5, 6, 9, 7, 11, 10]
```

Se inserta el valor **3** al final. Dibujar el árbol y trazar los swaps del up-heap bubbling.

```
Árbol inicial:
         2
       /   \
      5     6
     / \   / \
    9   7 11  10

Después de insertar 3 al final:
         2
       /   \
      5     6
     / \   / \
    9   7 11  10
   /
  3   ← nuevo (índice 7)
```

**¿Cuántos swaps se necesitan? ¿Cuál es el estado final del heap?**

### Parte B: Down-Heap

Del heap `[2, 5, 6, 9, 7, 11, 10]`, se llama a `remove_min()`.

1. El 2 (mínimo) se guarda
2. El último elemento (10) se mueve a la raíz
3. Se aplica down-heap

**Trazar los swaps del down-heap bubbling hasta el estado final.**


````python
class Empty(Exception): pass
class PriorityQueueBase:
    class _Item:
        __slots__ = '_key', '_value'
        def __init__(self, k, v): self._key = k; self._value = v
        def __lt__(self, other): return self._key < other._key
        def __repr__(self): return f'({self._key},{self._value!r})'
    def is_empty(self): return len(self) == 0


class HeapTracer(PriorityQueueBase):
    """HeapPriorityQueue con traza de swaps para visualización pedagógica."""
    
    def _parent(self, j): return (j-1)//2
    def _left(self, j): return 2*j+1
    def _right(self, j): return 2*j+2
    def _has_left(self, j): return self._left(j) < len(self._data)
    def _has_right(self, j): return self._right(j) < len(self._data)
    
    def __init__(self):
        self._data = []
        self._swaps = []   # Para registrar swaps
    
    def _swap(self, i, j):
        print(f'    SWAP [{i}]={self._data[i]} ↔ [{j}]={self._data[j]}')
        self._data[i], self._data[j] = self._data[j], self._data[i]
        self._swaps.append((i, j))
    
    def _upheap(self, j):
        parent = self._parent(j)
        if j > 0 and self._data[j] < self._data[parent]:
            self._swap(j, parent)
            self._upheap(parent)
    
    def _downheap(self, j):
        if self._has_left(j):
            l = self._left(j); s = l
            if self._has_right(j):
                r = self._right(j)
                if self._data[r] < self._data[l]: s = r
            if self._data[s] < self._data[j]:
                self._swap(j, s); self._downheap(s)
    
    def add(self, k, v):
        self._data.append(self._Item(k, v))
        print(f'  add({k}) → array antes upheap: {[x._key for x in self._data]}')
        self._upheap(len(self._data)-1)
        print(f'         → array después: {[x._key for x in self._data]}')
    
    def remove_min(self):
        if self.is_empty(): raise Empty()
        self._swap(0, len(self._data)-1)
        item = self._data.pop()
        print(f'  remove_min() → extrae {item._key}, array antes downheap: {[x._key for x in self._data]}')
        self._downheap(0)
        print(f'               → array después: {[x._key for x in self._data]}')
        return (item._key, item._value)


# ── Traza Up-Heap ──
print('=== Traza UP-HEAP ===')
H = HeapTracer()
for k in [2, 5, 6, 9, 7, 11, 10]:   # Construir heap inicial
    H._data.append(H._Item(k, k))
print(f'Heap inicial: {[x._key for x in H._data]}')
print()
H.add(3, 3)

print()
print('=== Traza DOWN-HEAP ===')
H2 = HeapTracer()
for k in [2, 5, 6, 9, 7, 11, 10]:
    H2._data.append(H2._Item(k, k))
print(f'Heap inicial: {[x._key for x in H2._data]}')
print()
H2.remove_min()
````
