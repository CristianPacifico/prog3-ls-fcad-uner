## 🔄 Listas Enlazadas Circulares

Una **lista enlazada circular** es una variación donde el último nodo apunta de vuelta al primero, formando un ciclo cerrado.

### 🌍 Analogías del Mundo Real

![Roundabout](https://cdn-icons-png.flaticon.com/256/5727/5727723.png "Fuente: Flaticon")

**🎠 Carrusel/Calesita:**

- Los caballos están en un círculo continuo
- No hay "primer" o "último" caballo específico
- Puedes empezar desde cualquier caballo y dar vueltas infinitas

**🕐 Reloj:**

- Las horas van de 1 a 12 y luego vuelven a 1
- No hay "inicio" o "fin" absoluto
- Puedes contar desde cualquier hora

```python
# Primero necesitamos la clase Node de la clase anterior
class Node:
    """🧱 Nodo individual para lista enlazada"""
    
    def __init__(self, element, next_node=None):
        """🚀 Inicializar nodo con elemento y referencia"""
        self.element = element      # 📦 Datos almacenados
        self.next = next_node       # 🔗 Referencia al siguiente nodo
    
    def __str__(self):
        """🖨️ Representación del nodo"""
        next_info = f" → {self.next.element}" if self.next else " → None"
        return f"[{self.element}]{next_info}"

print("✅ Clase Node definida")
```

### 🏗️ Estructura Básica de Lista Circular


* **Estructura:** En el caso de las listas enlazadas, hay una noción más tangible de una lista circularmente enlazada, ya que podemos hacer que la cola de la lista use su referencia `next` para apuntar al `head` de la lista.

* **Propósito:** Modelan conjuntos de datos cíclicos que no tienen un principio o fin específico.

* **Referencia:** Se mantiene una referencia a un nodo particular, a menudo llamado `current` o `tail` (si se usa para una cola), para navegar la lista. Utilizamos el identificador `current` para describir dicho nodo designado. Al asignar `current = current.next`, podemos avanzar de manera efectiva a través de los nodos de la lista.

```python
class CircularlyLinkedList:
    """🔄 Lista enlazada circular"""
    
    def __init__(self):
        """🚀 Inicializar lista circular vacía"""
        self._tail = None      # 🎯 Solo mantenemos referencia al tail
        self._size = 0         # 📏 Número de elementos
    
    def __len__(self):
        """📏 Retornar número de elementos"""
        return self._size
    
    def is_empty(self):
        """❓ Verificar si está vacía"""
        return self._size == 0
    
    def first(self):
        """👑 Retornar (sin remover) el primer elemento"""
        if self.is_empty():
            raise ValueError("❌ List is empty")
        return self._tail.next.element  # El primero está después del tail
    
    def __str__(self):
        """🖨️ Representación string de la lista circular"""
        if self.is_empty():
            return "Lista circular vacía: ∅"
        
        elements = []
        current = self._tail.next  # Empezar desde el head
        for _ in range(self._size):
            elements.append(str(current.element))
            current = current.next
        
        return f"Lista circular: ({' → '.join(elements)} → ⤴️)"

# 🧪 Crear lista circular básica
print("🎪 Creando lista circular:")
circular_list = CircularlyLinkedList()
print(f"Lista inicial: {circular_list}")
```

### 🔄 Operación Rotate - La Magia Circular

```python
# Agregamos los métodos de manipulación a la clase CircularlyLinkedList

def rotate(self):
    """🔄 Rotar la lista: mover el primer elemento al final - O(1)"""
    if not self.is_empty():
        self._tail = self._tail.next  # ¡Increíblemente simple!
        print(f"🔄 Lista rotada: {self}")

def add_first(self, element):
    """📥 Agregar elemento al inicio - O(1)"""
    new_node = Node(element)
    
    if self.is_empty():
        # 🌟 Primer nodo: apunta a sí mismo
        new_node.next = new_node
        self._tail = new_node
    else:
        # 🔗 Insertar después del tail (que es antes del head)
        new_node.next = self._tail.next
        self._tail.next = new_node
    
    self._size += 1
    print(f"📥 Agregado '{element}' al inicio: {self}")

def add_last(self, element):
    """📥 Agregar elemento al final - O(1)"""
    self.add_first(element)  # Agregar al inicio...
    self.rotate()            # ...y luego rotar para moverlo al final

# Agregamos los métodos a la clase
CircularlyLinkedList.rotate = rotate
CircularlyLinkedList.add_first = add_first
CircularlyLinkedList.add_last = add_last

print("✅ Métodos de manipulación agregados")
```

```python
# 🧪 ¡Probemos las operaciones circulares!
print("🎪 Probando operaciones circulares:")

circular_list = CircularlyLinkedList()
circular_list.add_first("A")
circular_list.add_first("B") 
circular_list.add_first("C")

print("\n🔄 Rotaciones:")
for i in range(5):
    print(f"Rotación {i+1}: ", end="")
    circular_list.rotate()
```

⏱️ Complejidad: todas las operaciones son O(1) en el peor caso (first, enqueue, dequeue, rotate).


### 💡 ¿Por qué Solo Tail?

**🤔 Pregunta**: ¿Por qué mantenemos solo `_tail` y no `_head`?

**💡 Respuesta**: ¡Porque el head siempre es `_tail.next`! Esto nos ahorra memoria y simplifica las operaciones.

## 🎯 Aplicación: Planificador Round-Robin

* Un ejemplo de aplicación donde las listas circulares son útiles. Un planificador round-robin itera a través de una colección de elementos de forma circular y "atiende" cada elemento.

* Una cola circular implementa eficientemente el método `rotate()`, que mueve el primer elemento al final, optimizando el ciclo de "desencolar, servir, encolar".

```python
class RoundRobinScheduler:
    """🎯 Planificador round-robin usando lista circular"""
    
    def __init__(self):
        """🚀 Inicializar planificador vacío"""
        self._queue = CircularlyLinkedList()
        self._current_process = None
    
    def add_process(self, process_name, time_slice=10):
        """📥 Agregar proceso al planificador"""
        process = {"name": process_name, "time": time_slice}
        self._queue.add_last(process)
        print(f"📥 Proceso '{process_name}' agregado con {time_slice}ms")
    
    def schedule_next(self):
        """⏭️ Programar siguiente proceso - O(1)"""
        if self._queue.is_empty():
            print("❌ No hay procesos para programar")
            return None
        
        # 🎯 Obtener proceso actual
        current_node = self._queue._tail.next if self._queue._tail else None
        if current_node:
            self._current_process = current_node.element
            print(f"🏃 Ejecutando: {self._current_process['name']} "
                  f"({self._current_process['time']}ms)")
            
            # 🔄 Rotar para el próximo ciclo
            self._queue.rotate()
            return self._current_process
        
        return None
    
    def remove_process(self, process_name):
        """🗑️ Remover proceso completado"""
        # Implementación simplificada
        print(f"✅ Proceso '{process_name}' completado y removido")

print("✅ RoundRobinScheduler definido")
```

![Los tres pasos iterativos para la programación round-robin utilizando una cola.](images/figure-7.9.png)

```python
# 🧪 ¡Simulemos un sistema operativo!
print("🎪 Simulador de Planificador Round-Robin:")

scheduler = RoundRobinScheduler()

# 📥 Agregar procesos
scheduler.add_process("Chrome", 15)
scheduler.add_process("VSCode", 20)
scheduler.add_process("Python", 10)
scheduler.add_process("Spotify", 5)

print("\n⏭️ Ciclos de planificación:")
for cycle in range(8):
    print(f"\n--- Ciclo {cycle + 1} ---")
    scheduler.schedule_next()
```

### 🎮 Mini-Challenge: Queue Circular

**💪 Desafío**: Implementa una Queue usando lista circular.

```python
class CircularQueue:
    """🔄 Queue circular con operación rotate eficiente"""
    
    def __init__(self):
        self._list = CircularlyLinkedList()
    
    def enqueue(self, element):
        """📥 Agregar al final de la cola"""
        # 🖊️ Tu implementación aquí
        self._list.add_last(element)
        print(f"📥 Enqueued: {element}")
    
    def dequeue(self):
        """📤 Remover del frente de la cola"""
        # 🖊️ Tu implementación aquí
        if self._list.is_empty():
            raise ValueError("❌ Queue is empty")
        
        element = self._list.first()
        # Implementación simplificada - en una versión completa
        # tendríamos un método remove_first()
        print(f"📤 Dequeued: {element}")
        return element
    
    def rotate(self):
        """🔄 Mover frente al final - O(1)"""
        # 🖊️ Tu implementación aquí
        self._list.rotate()
    
    def __str__(self):
        return f"CircularQueue: {self._list}"

# Probemos la CircularQueue
print("🧪 Probando CircularQueue:")
cq = CircularQueue()
cq.enqueue("Tarea 1")
cq.enqueue("Tarea 2")
cq.enqueue("Tarea 3")
print(f"Queue: {cq}")

print("\n🔄 Rotaciones:")
cq.rotate()
cq.rotate()
```

## ⚡ Listas Doblemente Enlazadas y Nodos Centinelas

Una **lista doblemente enlazada** permite navegación bidireccional con nodos que tienen referencias tanto al siguiente como al anterior.

* **Estructura:** Para proporcionar una mayor simetría, definimos una lista enlazada en la que cada nodo mantiene una referencia explícita al nodo anterior y una referencia al nodo posterior. Esta estructura se conoce como lista doblemente enlazada.

* **Nodos:** Cada nodo tiene un `element`, una referencia `next` y una referencia `prev` (anterior).

* **Ventaja:** Permiten una mayor variedad de operaciones de actualización en tiempo O(1), incluyendo inserciones y eliminaciones en posiciones arbitrarias de la lista.


### 🧱 Nodo Doblemente Enlazado

```python
class DoublyNode:
    """🧱 Nodo para lista doblemente enlazada"""
    
    def __init__(self, element, prev_node=None, next_node=None):
        """🚀 Inicializar nodo con elemento y referencias bidireccionales"""
        self.element = element      # 📦 Datos
        self.prev = prev_node       # ⬅️ Referencia al anterior
        self.next = next_node       # ➡️ Referencia al siguiente
    
    def __str__(self):
        """🖨️ Representación del nodo"""
        prev_info = f"{self.prev.element} ← " if self.prev else "None ← "
        next_info = f" → {self.next.element}" if self.next else " → None"
        return f"{prev_info}[{self.element}]{next_info}"

print("✅ DoublyNode definido")
```

### 🛡️ Nodos Centinela (Sentinels)

Los **nodos centinela** son nodos "dummy" que simplifican las operaciones eliminando casos especiales.

* **Propósito:** Para evitar algunos casos especiales al operar cerca de los límites de una lista doblemente enlazada, resulta útil añadir nodos especiales en ambos extremos de la lista: un nodo de encabezado al principio de la lista y un nodo de cola al final de la lista. Estos nodos "dummy" no almacenan elementos reales de la secuencia.

* **Simplificación:** Los centinelas simplifican la lógica de las operaciones al asegurar que las inserciones siempre ocurren entre dos nodos existentes y que los elementos a eliminar siempre tienen vecinos en ambos lados, eliminando casos especiales para listas vacías o inserciones/eliminaciones en los extremos.

![figure-7.10.png](images/figure-7.10.png)

```python
class DoublyLinkedList:
    """⚡ Lista doblemente enlazada con nodos centinela"""
    
    def __init__(self):
        """🚀 Crear lista vacía con nodos centinela"""
        
        # 🛡️ Crear nodos centinela
        self._header = DoublyNode(None)      # 🏁 Centinela inicial
        self._trailer = DoublyNode(None)     # 🏁 Centinela final
        
        # 🔗 Conectar centinelas entre sí
        self._header.next = self._trailer
        self._trailer.prev = self._header
        
        self._size = 0
        
        print("🛡️ Lista creada con nodos centinela")
    
    def __len__(self):
        """📏 Número de elementos"""
        return self._size
    
    def is_empty(self):
        """❓ Verificar si está vacía"""
        return self._size == 0
    
    def __str__(self):
        """🖨️ Representación de la lista"""
        if self.is_empty():
            return "Lista doblemente enlazada vacía: [header] ↔ [trailer]"
        
        elements = []
        current = self._header.next
        while current != self._trailer:
            elements.append(str(current.element))
            current = current.next
        
        return f"DLL: [header] ↔ {' ↔ '.join(elements)} ↔ [trailer]"

# 🧪 Crear lista doblemente enlazada
print("🎪 Creando lista doblemente enlazada:")
dll = DoublyLinkedList()
print(f"Lista inicial: {dll}")
```

### 🛠️ Operaciones Bidireccionales

```python
def _insert_between(self, element, predecessor, successor):
    """🔧 Método auxiliar: insertar entre dos nodos - O(1)"""
    
    # 🏗️ Crear nuevo nodo
    new_node = DoublyNode(element, predecessor, successor)
    
    # 🔗 Actualizar enlaces
    predecessor.next = new_node
    successor.prev = new_node
    
    self._size += 1
    return new_node

def _delete_node(self, node):
    """🗑️ Método auxiliar: eliminar nodo dado - O(1)"""
    
    # 🔗 Reconectar vecinos
    predecessor = node.prev
    successor = node.next
    
    predecessor.next = successor
    successor.prev = predecessor
    
    self._size -= 1
    
    # 📦 Guardar elemento y limpiar nodo
    element = node.element
    node.prev = node.next = node.element = None  # Garbage collection
    
    return element

def add_first(self, element):
    """📥 Agregar al inicio - O(1)"""
    self._insert_between(element, self._header, self._header.next)
    print(f"📥 Agregado '{element}' al inicio: {self}")

def add_last(self, element):
    """📥 Agregar al final - O(1)"""
    self._insert_between(element, self._trailer.prev, self._trailer)
    print(f"📥 Agregado '{element}' al final: {self}")

def remove_first(self):
    """📤 Remover del inicio - O(1)"""
    if self.is_empty():
        raise ValueError("❌ Cannot remove from empty list")
    
    element = self._delete_node(self._header.next)
    print(f"📤 Removido '{element}' del inicio: {self}")
    return element

def remove_last(self):
    """📤 Remover del final - O(1)"""
    if self.is_empty():
        raise ValueError("❌ Cannot remove from empty list")
    
    element = self._delete_node(self._trailer.prev)
    print(f"📤 Removido '{element}' del final: {self}")
    return element

# Agregamos los métodos a la clase
DoublyLinkedList._insert_between = _insert_between
DoublyLinkedList._delete_node = _delete_node
DoublyLinkedList.add_first = add_first
DoublyLinkedList.add_last = add_last
DoublyLinkedList.remove_first = remove_first
DoublyLinkedList.remove_last = remove_last

print("✅ Métodos bidireccionales agregados")
```

```python
# 🧪 ¡Probemos todas las operaciones!
print("🎪 Probando operaciones bidireccionales:")

dll = DoublyLinkedList()
dll.add_first("B")
dll.add_first("A")
dll.add_last("C")
dll.add_last("D")

print("\n📤 Removiendo elementos:")
dll.remove_first()
dll.remove_last()
```

### 🎯 Ventajas de Listas Doblemente Enlazadas

| Operación | Simplemente Enlazada | Doblemente Enlazada |
|-----------|---------------------|---------------------|
| **add_first()** | O(1) ✅ | O(1) ✅ |
| **remove_first()** | O(1) ✅ | O(1) ✅ |
| **add_last()** | O(1) ✅ | O(1) ✅ |
| **remove_last()** | O(n) 🐌 | O(1) ⚡ |
| **navegación bidireccional** | ❌ | ✅ |
| **eliminación con referencia** | O(n) 🐌 | O(1) ⚡ |

## 🎯 Aplicación: Deque Eficiente

```python
class LinkedDeque:
    """🎯 Deque implementado con lista doblemente enlazada"""
    
    def __init__(self):
        """🚀 Inicializar deque vacío"""
        self._list = DoublyLinkedList()
    
    def add_first(self, element):
        """📥 Agregar al frente - O(1)"""
        self._list.add_first(element)
    
    def add_last(self, element):
        """📥 Agregar al final - O(1)"""
        self._list.add_last(element)
    
    def remove_first(self):
        """📤 Remover del frente - O(1)"""
        return self._list.remove_first()
    
    def remove_last(self):
        """📤 Remover del final - O(1)"""
        return self._list.remove_last()
    
    def first(self):
        """👁️ Ver primer elemento"""
        if self._list.is_empty():
            raise ValueError("❌ Deque is empty")
        return self._list._header.next.element
    
    def last(self):
        """👁️ Ver último elemento"""
        if self._list.is_empty():
            raise ValueError("❌ Deque is empty")
        return self._list._trailer.prev.element
    
    def is_empty(self):
        """❓ Verificar si está vacío"""
        return self._list.is_empty()
    
    def __len__(self):
        """📏 Número de elementos"""
        return len(self._list)
    
    def __str__(self):
        """🖨️ Representación del deque"""
        return f"Deque: {self._list}"

print("✅ LinkedDeque definido")
```

```python
# 🧪 ¡Probemos nuestro LinkedDeque!
print("🎪 Probando LinkedDeque:")

deque = LinkedDeque()

# 📥 Agregar en ambos extremos
deque.add_first("🍎")
deque.add_last("🍌")
deque.add_first("🍇")
deque.add_last("🍊")

print(f"Deque lleno: {deque}")
print(f"Primer elemento: {deque.first()}")
print(f"Último elemento: {deque.last()}")

# 📤 Remover de ambos extremos
print("\nRemoción alternada:")
print(f"Remove first: {deque.remove_first()}")
print(f"Remove last: {deque.remove_last()}")
print(f"Deque actual: {deque}")
```

## 📊 Comparación Final: Tipos de Listas Enlazadas
