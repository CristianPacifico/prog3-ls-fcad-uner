## Puntos clave
- Listas circulares: permiten `rotate()` en O(1) y recorridos cíclicos con una única referencia a `tail`.
- Listas doblemente enlazadas con centinelas: simplifican inserciones/eliminaciones uniformes en O(1).
- Lista Posicional: usa `Position` para operar localmente sin índices; base de estructuras flexibles.
- Implementación: se apoya en una base doblemente enlazada y valida posiciones para robustez.
