## 🤔 ¿Qué son las Listas Enlazadas?

Una **lista enlazada** es una estructura de datos lineal donde los elementos no se almacenan en posiciones contiguas de memoria. En su lugar, cada elemento (llamado **nodo**) contiene:

1. **Datos**: El valor o elemento almacenado
2. **Referencia**: Un "puntero" al siguiente nodo en la secuencia

## 🆚 Arrays vs. Listas Enlazadas

| Aspecto | Arrays (Python Lists) | Listas Enlazadas |
|---------|----------------------|------------------|
| **Acceso por índice** | O(1) ⚡ | O(n) 🐌 |
| **Inserción al inicio** | O(n) 🐌 | O(1) ⚡ |
| **Inserción al final** | O(1)* 🟡 | O(1) ⚡ |
| **Inserción en posición k** | O(n-k) 🐌 | O(1)** ⚡ |
| **Eliminación** | Similar a inserción | Similar a inserción |
| **Uso de memoria** | Menos overhead | Más overhead (refs) |
| **Localidad espacial** | Excelente 🎯 | Pobre 📍 |

*\* Amortizado (puede ser O(n) por redimensionamiento)*  
*\*\* Si tienes referencia al nodo*

### ⚠️ Problemas de los Arrays Dinámicos

1. **🏗️ Overhead de memoria**: Array puede ser más grande que elementos actuales
2. **⏰ Límites amortizados**: Inaceptables en sistemas de tiempo real
3. **🔄 Inserciones/eliminaciones costosas**: O(n) en posiciones intermedias

### En Resumen:

Las listas enlazadas son una estructura de datos fundamental que ofrece una alternativa a las secuencias basadas en arreglos (como las listas de Python). Ambas estructuras mantienen elementos en un orden determinado, pero utilizan estilos de representación muy diferentes.

**Desventajas de los Arreglos Dinámicos (Python lists):**

1. La longitud de un arreglo dinámico puede ser mayor que el número real de elementos que almacena.
2. Los límites amortizados para las operaciones pueden ser inaceptables en sistemas de tiempo real.
3. Las inserciones y eliminaciones en posiciones intermedias de un arreglo son costosas.

**Ventajas y Desventajas de las Listas Enlazadas frente a Arreglos:**

* **Representación:** Un arreglo ofrece una representación más centralizada con un gran bloque de memoria. En contraste, una lista enlazada, se basa en una representación más distribuida en la que se asigna un objeto liviano, conocido como nodo, para cada elemento. Cada nodo mantiene una referencia a su elemento y una o más referencias a nodos vecinos para representar colectivamente el orden lineal.
* **Acceso por índice:** Los elementos de una lista enlazada no pueden ser accedidos de manera eficiente mediante un índice numérico $k$. Esto contrasta con los arreglos que permiten acceso $O(1)$ por índice.
* **Desventajas evitadas:** Las listas enlazadas evitan las tres desventajas mencionadas para las secuencias basadas en arreglos, especialmente las inserciones y eliminaciones costosas en posiciones intermedias.

## 🧱 Anatomía de un Nodo - Building Blocks

Cada nodo almacena una referencia a un objeto que es un elemento de la secuencia, así como una referencia al siguiente nodo de la lista.

![Nodo - Figura 7.1](images/figure-7.1.png)

### 🔧 Estructura Básica

```python
class Node:
    """🏗️ Nodo básico para lista simplemente enlazada"""
    
    def __init__(self, element, next_node=None):
        """🚀 Inicializar nodo con elemento y referencia"""
        self.element = element      # 📦 Datos almacenados
        self.next = next_node       # 🔗 Referencia al siguiente nodo
    
    def __str__(self):
        """🖨️ Representación string del nodo"""
        next_info = f" -> {self.next.element}" if self.next else " -> None"
        return f"[{self.element}]{next_info}"

# 🧪 ¡Vamos a crear algunos nodos!
print("🎪 Creando nodos individuales:")

# Crear nodos individuales
nodo1 = Node("A")
nodo2 = Node("B") 
nodo3 = Node("C")

print(f"Nodo 1: {nodo1}")
print(f"Nodo 2: {nodo2}")  
print(f"Nodo 3: {nodo3}")
```

### 🔗 Conectando Nodos

```python
# 🔗 Conectar los nodos para formar una lista
nodo1.next = nodo2
nodo2.next = nodo3

print(f"\n🔗 Después de conectar:")
print(f"Nodo 1: {nodo1}")
print(f"Nodo 2: {nodo2}")
print(f"Nodo 3: {nodo3}")

# 🌟 Recorrer la lista completa
print(f"\n🚶 Recorriendo la lista desde nodo1:")
current = nodo1
while current is not None:
    print(f"  📍 Visitando: {current.element}")
    current = current.next
print("  ✅ Fin de la lista")
```

## 🏗️ Lista Simplemente Enlazada - Implementación Completa

* **Estructura:** Una lista simplemente enlazada, en su forma más simple, es una colección de nodos 🧩 que forman colectivamente una secuencia lineal ➡️. Cada nodo almacena una referencia a un objeto que es un elemento (carga útil) de la secuencia 📦, así como una referencia al siguiente nodo de la lista 🔗. 

* **Nodos:** Cada nodo tiene un elemento 🟦 y una referencia al siguiente nodo (next) 👉. Si no hay más nodos, next es `None` 🚫.

* **Cabeza y Cola (Head and Tail):** El primer 🥇 y último 🏁 nodo de una lista enlazada se conoce como _cabeza (head)_ 👑 y _cola (tail)_ 🎯 de la lista, respectivamente. La instancia de la lista enlazada `L` generalmente mantiene una referencia a `head` 👑 y, por conveniencia, a `tail` 🎯 y al `size` 📏.

* **Recorrido (Traversal):** El proceso de seguir las referencias `next` de los nodos, desde `L.head` 👑 hasta `L.tail` 🎯, se conoce como _recorrido de lista (traversing the linked list)_ 🚶‍♂️ o "_Link Hopping/Pointer Hopping_" 🔀.

![Diagrama de una lista simplemente enlazada - Figura 7.2](images/figure-7.2.png)

### 🧩 Estructura de la Clase

```python
class SinglyLinkedList:
    """🔗 Lista simplemente enlazada con operaciones básicas"""
    
    def __init__(self):
        """🚀 Inicializar lista vacía"""
        self._head = None           # 👑 Primer nodo (cabeza)
        self._tail = None           # 🎯 Último nodo (cola)  
        self._size = 0              # 📏 Número de elementos
    
    def __len__(self):
        """📏 Retornar número de elementos"""
        return self._size
    
    def is_empty(self):
        """❓ Verificar si la lista está vacía"""           
        return self._size == 0
    
    def __str__(self):
        """🖨️ Representación string de la lista"""
        if self.is_empty():
            return "Lista vacía: []"
        
        elements = []
        current = self._head
        while current is not None:
            elements.append(str(current.element))
            current = current.next
        
        return f"Lista: [{' -> '.join(elements)}]"
```

## Definición de SinglyLinkedList

`SinglyLinkedList` modela una lista simplemente enlazada con punteros a cabeza y cola.

Estado interno:
- `_head`: referencia al primer nodo (Node) o None si la lista está vacía.
- `_tail`: referencia al último nodo (Node) o None si la lista está vacía.
- `_size`: cantidad de elementos almacenados (int).

Invariantes:
- Si `_size == 0` => `_head` is None y `_tail` is None.
- Si `_size > 0` => `_head` y `_tail` no son None y `_tail.next` es None.
- El recorrido desde `_head` siguiendo `.next` visita exactamente `_size` nodos.

Métodos principales:
- `__len__(self)`: retorna `_size`. Complejidad O(1).
- `is_empty(self)`: retorna (`_size == 0`). Complejidad O(1).
- `__str__(self)`: si está vacía, muestra "Lista vacía: []".
    Si no, recorre desde `_head` acumulando `element` de cada Node y forma:
    "Lista: [e1 -> e2 -> ... -> ek]". Complejidad O(n).

Notas de diseño:
- Mantener `_tail` permite inserción eficiente al final (add_last) en O(1).
- Mantener `_head` permite inserción/eliminación al inicio (add_first/remove_first) en O(1).
- El almacenamiento real se hace en `Node(element, next)`, donde `next` apunta al siguiente nodo.

### 📥 Inserción al Inicio (add_first)

Vamos a definir, fuera de `SinglyLinkedList` los métodos `add_first`, `remove_first`, `add_last``remove_last`.

> ### Nota
> 🧭 Para claridad en la secuencia de la explicación, se implementa uno a uno estos métodos, y se los agrega a la definición de la clase. ➕📦
>
> 🐒 Esta técnica se llama *"monkey patching"*. Consiste en modificar clases, objetos o módulos en tiempo de ejecución ⏱️, agregando o cambiando métodos fuera de su definición original ✍️. ✅ Es válida en Python y funciona tanto en notebooks como en scripts y proyectos, pero ⚠️ no se recomienda para proyectos grandes o producción, porque puede dificultar el mantenimiento 🧩, la depuración 🐞 y la comprensión del código 🧠. Es útil para prototipos 🧪, pruebas rápidas ⚡, o cuando necesitas modificar el comportamiento de una clase sin editar su código fuente 🧰 (por ejemplo, en notebooks educativos como es este caso 📓).
>

> - Python Docs — Data Model (atributos y enlace dinámico): https://docs.python.org/3/reference/datamodel.html


```python
# Agregar método add_first a la clase SinglyLinkedList
def add_first(self, element):
    """📥 Agregar elemento al inicio de la lista - O(1)"""
    
    # 🏗️ Crear nuevo nodo
    new_node = Node(element)
    
    if self.is_empty():
        # 🌟 Lista vacía: nuevo nodo es head Y tail
        self._head = new_node
        self._tail = new_node
    else:
        # 🔗 Conectar nuevo nodo al head actual
        new_node.next = self._head
        self._head = new_node
    
    self._size += 1
    print(f"📥 Agregado '{element}' al inicio. Tamaño: {self._size}")

# Agregar el método a la clase - monkey patching
SinglyLinkedList.add_first = add_first

# 🧪 ¡Probemos la inserción al inicio!
print("🎪 Creando lista y agregando elementos al inicio:")

lista = SinglyLinkedList()
print(f"Lista inicial: {lista}")
print(f"¿Está vacía? {lista.is_empty()}")

lista.add_first("C")
print(f"Después de agregar 'C': {lista}")

lista.add_first("B")
print(f"Después de agregar 'B': {lista}")

lista.add_first("A")
print(f"Después de agregar 'A': {lista}")
```

### 📤 Eliminación del Inicio (remove_first)

```python
def remove_first(self):
    """📤 Remover y retornar el primer elemento - O(1)"""
    
    if self.is_empty():
        raise ValueError("❌ Cannot remove from empty list")
    
    # 📦 Guardar elemento a retornar
    element = self._head.element
    
    # 🔄 Actualizar head
    self._head = self._head.next
    self._size -= 1
    
    # 🎯 Si la lista quedó vacía, actualizar tail
    if self.is_empty():
        self._tail = None
    
    print(f"📤 Removido '{element}'. Tamaño: {self._size}")
    return element

# Agregar el método a la clase
SinglyLinkedList.remove_first = remove_first

# 🧪 ¡Probemos la eliminación!
print(f"\n🎪 Removiendo elementos del inicio:")
print(f"Lista actual: {lista}")

try:
    removed = lista.remove_first()
    print(f"Removido: {removed}")
    print(f"Lista después: {lista}")
    
    removed = lista.remove_first()
    print(f"Removido: {removed}")
    print(f"Lista después: {lista}")
    
    removed = lista.remove_first()
    print(f"Removido: {removed}")
    print(f"Lista después: {lista}")
    
    # Intentar remover de lista vacía
    removed = lista.remove_first()
    
except ValueError as e:
    print(f"💥 Error esperado: {e}")
```

### 📥 Inserción al Final (add_last)

```python
def add_last(self, element):
    """📥 Agregar elemento al final de la lista - O(1)"""
    
    # 🏗️ Crear nuevo nodo
    new_node = Node(element)
    
    if self.is_empty():
        # 🌟 Lista vacía: nuevo nodo es head Y tail
        self._head = new_node
        self._tail = new_node
    else:
        # 🔗 Conectar tail actual al nuevo nodo
        self._tail.next = new_node
        self._tail = new_node
    
    self._size += 1
    print(f"📥 Agregado '{element}' al final. Tamaño: {self._size}")

# Agregar el método a la clase
SinglyLinkedList.add_last = add_last

# 🧪 ¡Probemos inserción al final!
print(f"\n🎪 Agregando elementos al final:")

lista = SinglyLinkedList()
lista.add_last("🍎")
lista.add_last("🍌") 
lista.add_last("🍇")
print(f"Lista con elementos al final: {lista}")
```

## 📤 Eliminación del Final (remove_last)

Desafortunadamente, no podemos eliminar fácilmente el último nodo de una lista enlazada simple. 

Incluso si mantenemos una referencia `_tail` al último nodo de la lista, debemos poder acceder al nodo anterior al último nodo para poder eliminar este último nodo. Pero no podemos llegar al nodo anterior a `_tail` siguiendo los enlaces siguientes desde `_tail`. La única forma de acceder a este nodo es comenzar desde `_head` y buscar a lo largo de toda la lista. Pero esa secuencia de operaciones de salto de enlaces podría llevar un tiempo lineal en relación a la cantidad de elementos $O(n)$. Si queremos admitir esa operación de manera eficiente, tendremos que hacer que nuestra lista sea doblemente enlazada.


## ⚡ Análisis de Complejidad

| Operación | Complejidad | Explicación |
|-----------|-------------|--------------|
| **add_first()** | O(1) | Solo actualizamos head |
| **remove_first()** | O(1) | Solo actualizamos head |
| **add_last()** | O(1) | Tenemos referencia a tail |
| **remove_last()** | O(n) 🐌 | Necesitamos encontrar penúltimo |
| **search(element)** | O(n) | Recorrido lineal necesario |
| **access by index** | O(n) | Recorrido desde head |

## 🎯 Aplicación Práctica: Stack con Lista Enlazada

Las listas enlazadas son perfectas para implementar un Stack porque todas las operaciones de pila ocurren en un extremo.

### 🧠 Explicación del Código

- Se construye un Stack (pila) sobre una lista simplemente enlazada usando **composición**: la clase de pila mantiene una instancia interna de la lista.
- Operaciones:
  - push(x) delega en add_first(x) de la lista. El tope de la pila coincide con la cabeza (head) de la lista.
  - pop() delega en remove_first(), removiendo el elemento del tope en O(1).
  - top() retorna el elemento en head sin removerlo, también en O(1).
  - is_empty() y len() se delegan a la lista subyacente.
- Complejidad: todas las operaciones fundamentales del stack (push, pop, top, is_empty, len) son O(1) gracias al acceso directo a la cabeza y al mantenimiento de size en la lista.

Ventajas de esta construcción:

- Sin redimensionamientos ni copias como en arrays dinámicos (`ArrayList`).
- Crecimiento y reducción estrictamente O(1).
- Encapsula detalles de nodos y punteros, exponiendo una API simple de pila.

```python
class LinkedStack:
    """📚 Stack implementado con lista simplemente enlazada"""
    
    def __init__(self):
        """🚀 Inicializar stack vacío"""
        self._list = SinglyLinkedList()
    
    def push(self, element):
        """📥 Agregar elemento al tope del stack - O(1)"""
        self._list.add_first(element)  # El tope está en el head
        print(f"📥 Push: {element}")
    
    def pop(self):
        """📤 Remover y retornar el tope del stack - O(1)"""
        if self.is_empty():
            raise ValueError("❌ Pop from empty stack")
        
        element = self._list.remove_first()
        print(f"📤 Pop: {element}")
        return element
    
    def top(self):
        """👁️ Ver el tope sin removerlo - O(1)"""
        if self.is_empty():
            raise ValueError("❌ Top of empty stack")
        return self._list._head.element
    
    def is_empty(self):
        """❓ Verificar si está vacío"""
        return self._list.is_empty()
    
    def __len__(self):
        """📏 Número de elementos"""
        return len(self._list)
    
    def __str__(self):
        """🖨️ Representación del stack"""
        return f"Stack: {self._list}"

# 🧪 ¡Probemos nuestro LinkedStack!
print("🎪 Probando LinkedStack:")

stack = LinkedStack()
print(f"Stack vacío: {stack}")

# 📥 Push elementos
stack.push("🍎")
stack.push("🍌")
stack.push("🍇")
print(f"Stack después de pushes: {stack}")

# 👁️ Ver el tope
print(f"👁️ Tope actual: {stack.top()}")

# 📤 Pop elementos
while not stack.is_empty():
    stack.pop()
    if not stack.is_empty():
        print(f"Stack después de pop: {stack}")
```

#### 🧩 Se usó el Patrón Adapter

Recordando el Patron Adapter, visto en anteriormente. En el patron cada participante tiene su rol:
- Target (interfaz esperada por el cliente): Stack con operaciones push, pop, top, is_empty y len.
- Adaptee (clase existente a reutilizar): SinglyLinkedList con add_first, remove_first, acceso a head, is_empty y __len__.
- Adapter: LinkedStack, que traduce la interfaz de Stack a las operaciones de la lista enlazada.

Mapa de adaptación:
- push(x) → add_first(x)
- pop() → remove_first()
- top() → head.element
- is_empty() → is_empty()
- len() → \_\_len\_\_()

Beneficios del Adapter aquí:
- Desacopla a los clientes de los detalles de implementación de la estructura subyacente.
- Permite intercambiar el “adaptee” (por ejemplo, otra estructura con operaciones equivalentes O(1)) sin cambiar el código cliente.
- Centraliza validaciones y manejo de errores (por ejemplo, excepciones al hacer pop/top en vacío) en un único punto.

En síntesis, LinkedStack actúa como un Adapter que presenta una interfaz de pila LIFO sobre una lista simplemente enlazada, aprovechando sus operaciones O(1) en la cabeza para cumplir eficientemente con los requisitos de la pila.
