## 🔧 Configuración Inicial

Primero definimos las clases base que necesitaremos para los ejercicios:

```python
# Clases base necesarias para los ejercicios

class Node:
    """🧱 Nodo individual para lista enlazada simple"""
    
    def __init__(self, element, next_node=None):
        """🚀 Inicializar nodo con elemento y referencia"""
        self.element = element      # 📦 Datos almacenados
        self.next = next_node       # 🔗 Referencia al siguiente nodo
    
    def __str__(self):
        """🖨️ Representación del nodo"""
        next_info = f" → {self.next.element}" if self.next else " → None"
        return f"[{self.element}]{next_info}"


class DoublyNode:
    """🧱 Nodo para lista doblemente enlazada"""
    
    def __init__(self, element, prev_node=None, next_node=None):
        """🚀 Inicializar nodo con elemento y referencias bidireccionales"""
        self.element = element      # 📦 Datos
        self.prev = prev_node       # ⬅️ Referencia al anterior
        self.next = next_node       # ➡️ Referencia al siguiente


class CircularlyLinkedList:
    """🔄 Lista enlazada circular básica"""
    
    def __init__(self):
        self._tail = None
        self._size = 0
    
    def __len__(self):
        return self._size
    
    def is_empty(self):
        return self._size == 0
    
    def first(self):
        if self.is_empty():
            raise ValueError("❌ List is empty")
        return self._tail.next.element
    
    def add_last(self, element):
        """📥 Agregar elemento al final - O(1)"""
        new_node = Node(element)
        
        if self.is_empty():
            new_node.next = new_node
            self._tail = new_node
        else:
            new_node.next = self._tail.next
            self._tail.next = new_node
            self._tail = new_node
        
        self._size += 1
    
    def rotate(self):
        """🔄 Rotar la lista - O(1)"""
        if not self.is_empty():
            self._tail = self._tail.next


class DoublyLinkedList:
    """⚡ Lista doblemente enlazada con nodos centinela"""
    
    def __init__(self):
        self._header = DoublyNode(None)
        self._trailer = DoublyNode(None)
        self._header.next = self._trailer
        self._trailer.prev = self._header
        self._size = 0
    
    def __len__(self):
        return self._size
    
    def is_empty(self):
        return self._size == 0
    
    def _insert_between(self, element, predecessor, successor):
        new_node = DoublyNode(element, predecessor, successor)
        predecessor.next = new_node
        successor.prev = new_node
        self._size += 1
        return new_node
    
    def _delete_node(self, node):
        predecessor = node.prev
        successor = node.next
        predecessor.next = successor
        successor.prev = predecessor
        self._size -= 1
        element = node.element
        node.prev = node.next = node.element = None
        return element
    
    def add_first(self, element):
        return self._insert_between(element, self._header, self._header.next)
    
    def add_last(self, element):
        return self._insert_between(element, self._trailer.prev, self._trailer)


print("✅ Clases base configuradas correctamente")
```

## 📝 Ejercicio 1: LinkedQueue Completa

**🎯 Objetivo**: Completar la implementación de una Queue usando lista enlazada simple.

### 📋 Especificaciones

Complete los métodos `enqueue()` y `dequeue()` en la siguiente clase:

```python
class LinkedQueue:
    """📦 Queue implementada con lista enlazada simple"""
    
    def __init__(self):
        """🚀 Inicializar queue vacía"""
        self._head = None      # 👑 Frente de la cola
        self._tail = None      # 🎯 Final de la cola  
        self._size = 0         # 📏 Número de elementos
    
    def __len__(self):
        """📏 Retornar número de elementos"""
        return self._size
    
    def is_empty(self):
        """❓ Verificar si está vacía"""
        return self._size == 0
    
    def first(self):
        """👁️ Ver primer elemento sin removerlo"""
        if self.is_empty():
            raise ValueError("❌ Queue is empty")
        return self._head.element
    
    def enqueue(self, element):
        """📥 Agregar elemento al final de la cola - O(1)"""
        # 🖊️ TODO: Implementar aquí
        # Pistas:
        # - Crear nuevo nodo
        # - Si está vacía, head y tail apuntan al nuevo nodo
        # - Si no está vacía, enlazar tail actual al nuevo nodo
        # - Actualizar tail y size
        
        new_node = Node(element)
        
        if self.is_empty():
            # Primera inserción: head y tail apuntan al nuevo nodo
            self._head = new_node
            self._tail = new_node
        else:
            # Enlazar tail actual al nuevo nodo y actualizar tail
            self._tail.next = new_node
            self._tail = new_node
        
        self._size += 1
        print(f"📥 Enqueued: {element}")
    
    def dequeue(self):
        """📤 Remover y retornar primer elemento - O(1)"""
        # 🖊️ TODO: Implementar aquí
        # Pistas:
        # - Verificar si está vacía
        # - Guardar elemento del head
        # - Mover head al siguiente nodo
        # - Si queda vacía, tail debe ser None
        # - Actualizar size y retornar elemento
        
        if self.is_empty():
            raise ValueError("❌ Cannot dequeue from empty queue")
        
        # Guardar elemento a retornar
        element = self._head.element
        
        # Mover head al siguiente nodo
        self._head = self._head.next
        
        # Si la queue queda vacía, actualizar tail
        if self._head is None:
            self._tail = None
        
        self._size -= 1
        print(f"📤 Dequeued: {element}")
        return element
    
    def __str__(self):
        """🖨️ Representación string de la queue"""
        if self.is_empty():
            return "Queue vacía: ∅"
        
        elements = []
        current = self._head
        while current:
            elements.append(str(current.element))
            current = current.next
        
        return f"Queue: [{' ← '.join(elements)}] (← enqueue, dequeue →)"


print("✅ LinkedQueue implementada")
```

## 🎯 Ejercicio 2: Playlist Circular

**🎯 Objetivo**: Crear un reproductor de música con playlist circular.

### 📋 Especificaciones

Complete la implementación de un sistema de playlist que nunca termine:

```python
class Song:
    """🎵 Clase para representar una canción"""
    
    def __init__(self, title, artist, duration):
        """🚀 Inicializar canción"""
        self.title = title
        self.artist = artist
        self.duration = duration  # En segundos
    
    def __str__(self):
        """🖨️ Representación de la canción"""
        minutes = self.duration // 60
        seconds = self.duration % 60
        return f"🎵 {self.title} - {self.artist} ({minutes:02d}:{seconds:02d})"


class CircularPlaylist:
    """🎶 Playlist circular para reproductor de música"""
    
    def __init__(self, name="Mi Playlist"):
        """🚀 Inicializar playlist"""
        self.name = name
        self._list = CircularlyLinkedList()
        self._current_song = None
    
    def add_song(self, song):
        """📥 Agregar canción a la playlist"""
        # 🖊️ TODO: Implementar
        # - Agregar la canción a la lista circular
        # - Si es la primera canción, establecerla como actual
        
        self._list.add_last(song)
        
        # Si es la primera canción, establecerla como actual
        if len(self._list) == 1:
            self._current_song = song
        
        print(f"📥 Agregada a '{self.name}': {song}")
    
    def next_song(self):
        """⏭️ Avanzar a la siguiente canción"""
        # 🖊️ TODO: Implementar
        # - Rotar la lista circular
        # - Actualizar canción actual
        # - Retornar la nueva canción actual
        
        if not self._list.is_empty():
            self._list.rotate()
            self._current_song = self._list.first()
            print(f"⏭️  Siguiente: {self._current_song}")
            return self._current_song
        return None
    
    def current(self):
        """🎵 Obtener canción actual"""
        return self._current_song
    
    def play_for(self, num_songs):
        """▶️ Reproducir durante un número específico de canciones"""
        # 🖊️ TODO: Implementar
        # - Reproducir num_songs canciones
        # - Mostrar cada canción que se reproduce
        # - Avanzar automáticamente
        
        if self._list.is_empty():
            print("❌ No hay canciones en la playlist")
            return
        
        print(f"▶️  Reproduciendo {num_songs} canciones:")
        for i in range(num_songs):
            print(f"  🎵 Reproduciendo: {self._current_song}")
            if i < num_songs - 1:  # No avanzar en la última iteración
                self.next_song()
    
    def total_duration(self):
        """⏱️ Calcular duración total de la playlist"""
        # 🖊️ TODO: Implementar (BONUS)
        # - Iterar por todas las canciones una sola vez
        # - Sumar todas las duraciones
        # - Retornar tiempo total en formato mm:ss
        
        if self._list.is_empty():
            return "00:00"
        
        total_seconds = 0
        current = self._list._tail.next  # Empezar desde el head
        
        # Iterar una vez por toda la lista circular
        for _ in range(len(self._list)):
            total_seconds += current.element.duration
            current = current.next
        
        # Convertir a formato mm:ss
        minutes = total_seconds // 60
        seconds = total_seconds % 60
        return f"{minutes:02d}:{seconds:02d}"


print("✅ CircularPlaylist implementada")
```

```python
# 🧪 Crear playlist de prueba
def test_circular_playlist():
    """🧪 Probar playlist circular"""
    
    playlist = CircularPlaylist("🎶 Rock Classics")
    
    # Agregar canciones
    songs = [
        Song("Bohemian Rhapsody", "Queen", 355),
        Song("Stairway to Heaven", "Led Zeppelin", 482),
        Song("Hotel California", "Eagles", 391),
        Song("Sweet Child O' Mine", "Guns N' Roses", 356)
    ]
    
    for song in songs:
        playlist.add_song(song)
    
    print(f"\n🎵 Canción actual: {playlist.current()}")
    
    # Reproducir varias canciones
    print(f"\n▶️ Reproduciendo 8 canciones (2 ciclos completos):")
    playlist.play_for(8)
    
    # Duración total
    print(f"\n⏱️ Duración total de la playlist: {playlist.total_duration()}")

# Ejecutar pruebas
test_circular_playlist()
```

### 🌐 Opción A: Navegador Web Simple

Complete la implementación de un navegador con historial bidireccional:

```python
class WebPage:
    """🌐 Página web simple"""
    def __init__(self, url, title):
        self.url = url
        self.title = title
        self.visit_time = time.time()
    
    def __str__(self):
        return f"🌐 {self.title} - {self.url}"


class WebBrowser:
    """🌐 Navegador con historial bidireccional"""
    
    def __init__(self):
        self.history = DoublyLinkedList()
        self.current_page = None
        self.current_node = None  # Nodo actual en el historial
    
    def visit(self, url, title):
        """🔗 Visitar nueva página"""
        page = WebPage(url, title)
        
        # Si hay una página actual, agregamos después de ella
        if self.current_node is None:
            # Primera página
            self.current_node = self.history.add_last(page)
        else:
            # Agregar después de la página actual y limpiar historial hacia adelante
            # Implementación simplificada: solo agregamos al final
            self.current_node = self.history.add_last(page)
        
        self.current_page = page
        print(f"🔗 Visitando: {page}")
    
    def back(self):
        """⬅️ Ir a página anterior"""
        if self.current_node and self.current_node.prev != self.history._header:
            self.current_node = self.current_node.prev
            self.current_page = self.current_node.element
            print(f"⬅️ Atrás: {self.current_page}")
        else:
            print("❌ No hay páginas anteriores")
    
    def forward(self):
        """➡️ Ir a página siguiente"""
        if self.current_node and self.current_node.next != self.history._trailer:
            self.current_node = self.current_node.next
            self.current_page = self.current_node.element
            print(f"➡️ Adelante: {self.current_page}")
        else:
            print("❌ No hay páginas siguientes")
    
    def show_history(self):
        """📚 Mostrar historial completo"""
        print("📚 Historial del navegador:")
        current = self.history._header.next
        index = 1
        while current != self.history._trailer:
            marker = " ← ACTUAL" if current == self.current_node else ""
            print(f"  {index}. {current.element}{marker}")
            current = current.next
            index += 1


# 🧪 Probar navegador
def test_web_browser():
    """🧪 Probar navegador web"""
    
    browser = WebBrowser()
    
    # Visitar páginas
    browser.visit("https://google.com", "Google")
    browser.visit("https://github.com", "GitHub")
    browser.visit("https://stackoverflow.com", "Stack Overflow")
    browser.visit("https://python.org", "Python.org")
    
    print("\n📚 Historial después de navegación:")
    browser.show_history()
    
    print("\n⬅️ Navegando hacia atrás:")
    browser.back()
    browser.back()
    
    print("\n➡️ Navegando hacia adelante:")
    browser.forward()
    
    print("\n📚 Historial final:")
    browser.show_history()

# Ejecutar pruebas
test_web_browser()
```

### 🎵 Opción B: Sistema de Recomendaciones

Complete la implementación de un sistema de recomendaciones musicales:

```python
class MusicRecommendationSystem:
    """🎵 Sistema de recomendaciones circulares"""
    
    def __init__(self):
        self.liked_songs = CircularlyLinkedList()
        self.current_recommendation = None
    
    def like_song(self, song):
        """👍 Marcar canción como favorita"""
        self.liked_songs.add_last(song)
        
        # Si es la primera canción, establecerla como recomendación actual
        if len(self.liked_songs) == 1:
            self.current_recommendation = song
        
        print(f"👍 Agregada a favoritas: {song}")
    
    def next_recommendation(self):
        """⏭️ Siguiente recomendación basada en favoritas"""
        if not self.liked_songs.is_empty():
            self.liked_songs.rotate()
            self.current_recommendation = self.liked_songs.first()
            print(f"🎯 Recomendación: {self.current_recommendation}")
            return self.current_recommendation
        return None
    
    def generate_playlist(self, length):
        """🎶 Generar playlist de longitud específica"""
        if self.liked_songs.is_empty():
            print("❌ No hay canciones favoritas para generar playlist")
            return []
        
        playlist = []
        print(f"🎶 Generando playlist de {length} canciones:")
        
        for i in range(length):
            playlist.append(self.current_recommendation)
            print(f"  {i+1}. {self.current_recommendation}")
            self.next_recommendation()
        
        return playlist
    
    def show_favorites(self):
        """💖 Mostrar canciones favoritas"""
        if self.liked_songs.is_empty():
            print("💔 No hay canciones favoritas")
            return
        
        print("💖 Canciones favoritas:")
        current = self.liked_songs._tail.next
        for i in range(len(self.liked_songs)):
            marker = " ← ACTUAL" if current.element == self.current_recommendation else ""
            print(f"  {i+1}. {current.element}{marker}")
            current = current.next


# 🧪 Probar sistema de recomendaciones
def test_recommendation_system():
    """🧪 Probar sistema de recomendaciones"""
    
    recommender = MusicRecommendationSystem()
    
    # Agregar canciones favoritas
    favorite_songs = [
        Song("Imagine", "John Lennon", 183),
        Song("Billie Jean", "Michael Jackson", 294),
        Song("Like a Rolling Stone", "Bob Dylan", 369),
        Song("Purple Haze", "Jimi Hendrix", 182)
    ]
    
    for song in favorite_songs:
        recommender.like_song(song)
    
    print("\n💖 Estado inicial:")
    recommender.show_favorites()
    
    print(f"\n🎯 Recomendación actual: {recommender.current_recommendation}")
    
    print(f"\n⏭️ Obteniendo siguientes recomendaciones:")
    for i in range(6):  # Más que el número de canciones para mostrar la rotación
        recommender.next_recommendation()
    
    print(f"\n🎶 Generando playlist personalizada:")
    playlist = recommender.generate_playlist(5)

# Ejecutar pruebas
test_recommendation_system()
```
